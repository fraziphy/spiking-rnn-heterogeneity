```python
# Single Trial Simulation and Raster Plot
# Import necessary modules and run a single simulation

import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# Add project root to path
project_root = os.getcwd()  # If notebook is in project root
# project_root = os.path.dirname(os.getcwd())  # If notebook is in subdirectory

sys.path.insert(0, project_root)

# Import using FULL PACKAGE PATHS - this makes relative imports work
from src.spiking_network import SpikingRNN
from src.rng_utils import get_rng
from analysis.spontaneous_analysis import analyze_spontaneous_activity

# =============================================================================
# SIMULATION PARAMETERS
# =============================================================================

# Network parameters
n_neurons = 1000
dt = 0.1  # ms
synaptic_mode = "dynamic"  # or "immediate"

# Heterogeneity parameters
session_id = 10
v_th_std = 0  # Threshold heterogeneity
g_std = 1  # Synaptic weight heterogeneity
trial_id = 1

# Simulation parameters
duration = 500.0  # 2 seconds in milliseconds
static_input_rate = 30.0  # Hz

# Additional network parameters
v_th_distribution = "normal"
network_params = {
    'v_th_distribution': v_th_distribution,
    'static_input_strength': 10.0,  # Enhanced connectivity strength
    'dynamic_input_strength': 1.0,
    'readout_weight_scale': 1.0
}

print(f"Simulation Parameters:")
print(f"  Network size: {n_neurons} neurons")
print(f"  Duration: {duration} ms")
print(f"  v_th_std: {v_th_std}")
print(f"  g_std: {g_std}")
print(f"  Static input rate: {static_input_rate} Hz")
print(f"  Synaptic mode: {synaptic_mode}")

# =============================================================================
# CREATE AND INITIALIZE NETWORK
# =============================================================================

static_input_mode="common_tonic"
# static_input_mode="independent"

# Create network
network = SpikingRNN(n_neurons=n_neurons, dt=dt, synaptic_mode="filter",static_input_mode=static_input_mode)

# Initialize network structure (depends on session + parameters)
network.initialize_network(session_id, v_th_std, g_std, **network_params)

print(f"\nNetwork initialized successfully!")

# Get network information
# network_info = network.get_network_info()
# print(f"  Threshold mean: {network_info['threshold_mean']:.2f} mV")
# print(f"  Threshold std: {network_info['threshold_std']:.3f} mV")
# print(f"  Weight connections: {network_info['weight_matrix_nnz']}")
# print(f"  Weight mean: {network_info['weight_mean']:.3f}")
# print(f"  Weight std: {network_info['weight_std']:.3f}")

# =============================================================================
# RUN SIMULATION
# =============================================================================

print(f"\nRunning simulation...")

# Run the simulation
spike_times = network.simulate_network_dynamics(
    session_id=session_id,
    v_th_std=v_th_std,
    g_std=g_std,
    trial_id=trial_id,
    duration=duration,
    static_input_rate=static_input_rate
)

print(f"Simulation completed!")
print(f"  Total spikes: {len(spike_times)}")
print(f"  Average firing rate: {len(spike_times) / (duration/1000) / n_neurons:.2f} Hz")

# =============================================================================
# CREATE RASTER PLOT
# =============================================================================

def plot_raster(spike_times, duration, n_neurons, title_suffix=""):
    """Create a raster plot from spike times"""
    
    if not spike_times:
        print("No spikes to plot!")
        return
    
    # Extract spike times and neuron IDs
    times = [spike[0] for spike in spike_times]
    neuron_ids = [spike[1] for spike in spike_times]
    
    # Create the plot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), 
                                   gridspec_kw={'height_ratios': [3, 1]})
    
    # Main raster plot
    ax1.scatter(times, neuron_ids, s=0.5, alpha=0.6, c='black')
    ax1.set_ylabel('Neuron ID')
    ax1.set_title(f'Raster Plot - {title_suffix}')
    ax1.set_xlim(0, duration)
    ax1.set_ylim(0, n_neurons)
    ax1.grid(True, alpha=0.3)
    
    # Population firing rate (binned)
    bin_size = 10.0  # ms
    bins = np.arange(0, duration + bin_size, bin_size)
    spike_counts, _ = np.histogram(times, bins=bins)
    firing_rate = spike_counts / (bin_size / 1000.0) / n_neurons  # Hz
    bin_centers = bins[:-1] + bin_size/2
    
    ax2.plot(bin_centers, firing_rate, 'b-', linewidth=1)
    ax2.fill_between(bin_centers, firing_rate, alpha=0.3)
    ax2.set_xlabel('Time (ms)')
    ax2.set_ylabel('Population Rate (Hz)')
    ax2.set_title('Population Firing Rate')
    ax2.set_xlim(0, duration)
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    return fig

# Create raster plot
title_suffix = f"v_th_std={v_th_std}, g_std={g_std}, Input={static_input_rate}Hz"
fig = plot_raster(spike_times, duration, n_neurons, title_suffix)

# =============================================================================
# DETAILED ANALYSIS
# =============================================================================

def analyze_spike_statistics(spike_times, duration, n_neurons):
    """Analyze basic spike statistics"""
    
    if not spike_times:
        return
    
    times = [spike[0] for spike in spike_times]
    neuron_ids = [spike[1] for spike in spike_times]
    
    # Per-neuron spike counts
    spike_counts = np.zeros(n_neurons)
    for neuron_id in neuron_ids:
        spike_counts[neuron_id] += 1
    
    # Firing rates per neuron
    firing_rates = spike_counts / (duration / 1000.0)  # Hz
    
    # Statistics
    total_spikes = len(spike_times)
    mean_rate = np.mean(firing_rates)
    std_rate = np.std(firing_rates)
    silent_neurons = np.sum(firing_rates == 0)
    active_neurons = n_neurons - silent_neurons
    
    print(f"\nSpike Statistics:")
    print(f"  Total spikes: {total_spikes}")
    print(f"  Mean firing rate: {mean_rate:.2f}  {std_rate:.2f} Hz")
    print(f"  Silent neurons: {silent_neurons} ({silent_neurons/n_neurons*100:.1f}%)")
    print(f"  Active neurons: {active_neurons} ({active_neurons/n_neurons*100:.1f}%)")
    print(f"  Min firing rate: {np.min(firing_rates):.2f} Hz")
    print(f"  Max firing rate: {np.max(firing_rates):.2f} Hz")
    
    # Plot firing rate distribution
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.hist(firing_rates, bins=50, alpha=0.7, edgecolor='black')
    ax.axvline(mean_rate, color='red', linestyle='--', linewidth=2, 
               label=f'Mean: {mean_rate:.2f} Hz')
    ax.set_xlabel('Firing Rate (Hz)')
    ax.set_ylabel('Number of Neurons')
    ax.set_title('Distribution of Firing Rates')
    ax.legend()
    ax.grid(True, alpha=0.3)
    plt.show()
    
    return {
        'firing_rates': firing_rates,
        'mean_rate': mean_rate,
        'std_rate': std_rate,
        'silent_neurons': silent_neurons,
        'active_neurons': active_neurons
    }

# Analyze the results
stats = analyze_spike_statistics(spike_times, duration, n_neurons)

# =============================================================================
# PARAMETER EXPLORATION FUNCTION
# =============================================================================

def run_parameter_comparison(v_th_stds, g_stds, input_rate=200.0, duration=1000.0):
    """Compare different parameter combinations"""
    
    fig, axes = plt.subplots(len(v_th_stds), len(g_stds), figsize=(15, 12))
    if len(v_th_stds) == 1:
        axes = axes.reshape(1, -1)
    if len(g_stds) == 1:
        axes = axes.reshape(-1, 1)
    
    results = []
    
    for i, v_th in enumerate(v_th_stds):
        for j, g in enumerate(g_stds):
            print(f"Running v_th_std={v_th}, g_std={g}...")
            
            # Create new network
            net = SpikingRNN(n_neurons=500, dt=dt, synaptic_mode=synaptic_mode)  # Smaller for speed
            net.initialize_network(session_id, v_th, g, **network_params)
            
            # Run simulation
            spikes = net.simulate_network_dynamics(
                session_id=session_id, v_th_std=v_th, g_std=g, trial_id=trial_id,
                duration=duration, static_input_rate=input_rate
            )
            
            # Extract data for plotting
            times = [spike[0] for spike in spikes]
            neuron_ids = [spike[1] for spike in spikes]
            
            # Plot mini raster
            ax = axes[i, j]
            if spikes:
                ax.scatter(times, neuron_ids, s=0.3, alpha=0.6, c='black')
            ax.set_title(f'v_th={v_th:.1f}, g={g:.1f}')
            ax.set_xlim(0, duration)
            ax.set_ylim(0, 500)
            
            if i == len(v_th_stds) - 1:
                ax.set_xlabel('Time (ms)')
            if j == 0:
                ax.set_ylabel('Neuron ID')
            
            # Store results
            firing_rate = len(spikes) / (duration/1000) / 500
            results.append({
                'v_th_std': v_th,
                'g_std': g,
                'n_spikes': len(spikes),
                'firing_rate': firing_rate
            })
    
    plt.tight_layout()
    plt.show()
    
    return results

# Example parameter comparison (uncomment to run)
# print("\nParameter Comparison:")
# comparison_results = run_parameter_comparison([0.1, 0.5], [0.1, 0.5], input_rate=200.0, duration=1000.0)
# for result in comparison_results:
#     print(f"v_th_std={result['v_th_std']}, g_std={result['g_std']}: {result['firing_rate']:.2f} Hz")

print(f"\nSimulation complete! You can now:")
print(f"  1. Modify parameters above and re-run")
print(f"  2. Try different synaptic modes ('immediate' vs 'dynamic')")
print(f"  3. Explore different input rates")
print(f"  4. Uncomment parameter comparison section")
print(f"  5. Add perturbations using network.inject_perturbation(neuron_id)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyArrowPatch
from matplotlib.lines import Line2D
import sys
import os

# Add project root to path
project_root = os.getcwd()
sys.path.insert(0, project_root)

# Import your network modules
from src.spiking_network import SpikingRNN
from src.rng_utils import get_rng

# Set random seed for reproducibility
np.random.seed(10)

# =============================================================================
# HELPER FUNCTIONS
# =============================================================================

def compute_participation_ratio(spike_times, n_neurons, duration, transient_time, bin_size=20.0):
    """Compute participation ratio from spike times"""
    if not spike_times:
        return 0.0
    
    # Filter out transient period
    spike_times_filtered = [(t, n) for t, n in spike_times if t >= transient_time]
    
    if not spike_times_filtered:
        return 0.0
    
    # Create time bins
    analysis_duration = duration - transient_time
    n_bins = int(analysis_duration / bin_size)
    
    # Create firing rate matrix (neurons x time_bins)
    firing_rate_matrix = np.zeros((n_neurons, n_bins))
    
    for t, neuron_id in spike_times_filtered:
        bin_idx = int((t - transient_time) / bin_size)
        if bin_idx < n_bins:
            firing_rate_matrix[neuron_id, bin_idx] += 1
    
    # Convert to rates (Hz)
    firing_rate_matrix = firing_rate_matrix / (bin_size / 1000.0)
    
    # Compute covariance matrix
    mean_rates = np.mean(firing_rate_matrix, axis=1, keepdims=True)
    centered_rates = firing_rate_matrix - mean_rates
    
    # Covariance matrix
    cov_matrix = np.cov(centered_rates)
    
    # Compute eigenvalues
    eigenvalues = np.linalg.eigvalsh(cov_matrix)
    eigenvalues = eigenvalues[eigenvalues > 0]
    
    if len(eigenvalues) == 0:
        return 0.0
    
    # Participation ratio
    pr = (np.sum(eigenvalues) ** 2) / np.sum(eigenvalues ** 2)
    
    return pr

def get_firing_rates(spike_times, n_neurons, duration, transient_time):
    """Get per-neuron firing rates excluding transient period"""
    spike_times_filtered = [(t, n) for t, n in spike_times if t >= transient_time]
    
    spike_counts = np.zeros(n_neurons)
    for _, neuron_id in spike_times_filtered:
        spike_counts[neuron_id] += 1
    
    analysis_duration = (duration - transient_time) / 1000.0
    firing_rates = spike_counts / analysis_duration
    
    return firing_rates

# =============================================================================
# STEP 1: RUN ALL SIMULATIONS AND STORE DATA
# =============================================================================

print("="*70)
print("STEP 1: RUNNING SIMULATIONS")
print("="*70)

# Simulation parameters
n_neurons = 1000
dt = 0.1
duration = 500.0
transient_time = 200.0
session_id = 10
v_th_std = 0
g_std = 1
trial_id = 1

network_params = {
    'v_th_distribution': 'normal',
    'static_input_strength': 10.0,
    'dynamic_input_strength': 1.0,
    'readout_weight_scale': 1.0
}

# Dictionary to store all simulation results
simulation_data = {
    'n_neurons': n_neurons,
    'duration': duration,
    'transient_time': transient_time,
    'main_simulation': {},
    'dimensionality_simulations': []
}

# Main simulation for raster plot (static_input_rate = 35 Hz)
print("\n1. Running main simulation (30 Hz static input)...")
static_input_rate_main = 30.0

network_main = SpikingRNN(n_neurons=n_neurons, dt=dt, synaptic_mode="filter", 
                          static_input_mode="common_tonic")
network_main.initialize_network(session_id, v_th_std, g_std, **network_params)

spike_times_main = network_main.simulate_network_dynamics(
    session_id=session_id,
    v_th_std=v_th_std,
    g_std=g_std,
    trial_id=trial_id,
    duration=duration,
    static_input_rate=static_input_rate_main
)

simulation_data['main_simulation'] = {
    'static_input_rate': static_input_rate_main,
    'spike_times': spike_times_main,
    'n_spikes': len(spike_times_main)
}

print(f"   Complete. Total spikes: {len(spike_times_main)}")

# Dimensionality simulations (30-40 Hz)
print("\n2. Running dimensionality simulations (30-40 Hz)...")
static_rates = np.arange(30, 41, 1)

for i, rate in enumerate(static_rates):
    print(f"   [{i+1}/{len(static_rates)}] Static input rate: {rate} Hz")
    
    net_temp = SpikingRNN(n_neurons=n_neurons, dt=dt, synaptic_mode="filter",
                         static_input_mode="common_tonic")
    net_temp.initialize_network(session_id, v_th_std, g_std, **network_params)
    
    spikes_temp = net_temp.simulate_network_dynamics(
        session_id=session_id, v_th_std=v_th_std, g_std=g_std,
        trial_id=trial_id, duration=duration, static_input_rate=rate
    )
    
    simulation_data['dimensionality_simulations'].append({
        'static_input_rate': rate,
        'spike_times': spikes_temp,
        'n_spikes': len(spikes_temp)
    })
    
    print(f"       Total spikes: {len(spikes_temp)}")

print("\n" + "="*70)
print("ALL SIMULATIONS COMPLETE!")
print("="*70)




# =============================================================================
# STEP 2: PROCESS DATA AND CREATE PLOTS
# =============================================================================

print("\n" + "="*70)
print("STEP 2: CREATING PLOTS")
print("="*70)

# Extract main simulation data
spike_times_main = simulation_data['main_simulation']['spike_times']
times = [s[0] for s in spike_times_main]
neuron_ids = [s[1] for s in spike_times_main]

# Compute firing rates for main simulation
firing_rates_main = get_firing_rates(spike_times_main, n_neurons, duration, transient_time)

# Compute participation ratios with chosen bin size
bin_size_pr = 2.0  # YOU CAN CHANGE THIS WITHOUT RERUNNING SIMULATIONS
print(f"\nComputing participation ratios with bin_size={bin_size_pr} ms...")

static_rates = []
participation_ratios = []

for sim in simulation_data['dimensionality_simulations']:
    rate = sim['static_input_rate']
    spikes = sim['spike_times']
    
    pr = compute_participation_ratio(spikes, n_neurons, duration, 
                                    transient_time, bin_size=bin_size_pr)
    
    static_rates.append(rate)
    participation_ratios.append(pr)
    print(f"   Rate {rate} Hz: PR = {pr:.2f}")
```

```python
import numpy as np
import pickle
from scipy import stats

# =============================================================================
# DATA PROCESSING FOR ENCODING FIDELITY AND TASK PERFORMANCE
# =============================================================================

# Parameters
sessions = range(10)  # 0 to 24
hdin_values = range(1, 6)  # 1 to 5

# =============================================================================
# 1. ENCODING FIDELITY DATA
# =============================================================================

encoding_results = {
    'hdin': [],
    'mean_r2': [],
    'r2_ci_lower': [],
    'r2_ci_upper': [],
    'mean_pr': [],
    'pr_ci_lower': [],
    'pr_ci_upper': [],
}

for hdin in hdin_values:
    r2_values = []
    pr_values = []
    
    for sess in sessions:
        try:
            results_file = f'results/autoencoding/data/task_autoencoding_session_{sess}_vth_0.000_g_1.000_rate_30_hd_{hdin}.pkl'
            with open(results_file, 'rb') as f:
                results = pickle.load(f)
            r2_values.append(results[0]['test_r2_mean'])
            pr_values.append(results[0]["dimensionality_summary"]["bin_2.0ms"]["participation_ratio_mean"])
        except FileNotFoundError:
            print(f"Warning: File not found for encoding - session {sess}, hdin {hdin}")
        except Exception as e:
            print(f"Error loading encoding data - session {sess}, hdin {hdin}: {e}")
    
    if len(r2_values) > 0:
        r2_array = np.array(r2_values)
        mean_r2 = np.mean(r2_array)
        
        # Calculate 95% confidence interval using t-distribution
        n = len(r2_array)
        std_err = stats.sem(r2_array)  # Standard error of the mean
        ci = stats.t.interval(0.95, n-1, loc=mean_r2, scale=std_err)
        
        encoding_results['hdin'].append(hdin)
        encoding_results['mean_r2'].append(mean_r2)
        encoding_results['r2_ci_lower'].append(ci[0])
        encoding_results['r2_ci_upper'].append(ci[1])
        
        print(f"Encoding - hdin={hdin}: mean R={mean_r2:.4f}, 95% CI=[{ci[0]:.4f}, {ci[1]:.4f}], n={n}")

    if len(pr_values) > 0:
        pr_array = np.array(pr_values)
        mean_pr = np.mean(pr_array)
        
        # Calculate 95% confidence interval using t-distribution
        n = len(pr_array)
        std_err = stats.sem(pr_array)  # Standard error of the mean
        ci = stats.t.interval(0.95, n-1, loc=mean_pr, scale=std_err)
        
        encoding_results['mean_pr'].append(mean_pr)
        encoding_results['pr_ci_lower'].append(ci[0])
        encoding_results['pr_ci_upper'].append(ci[1])
        
        print(f"Encoding - hdin={hdin}: mean PR={mean_pr:.4f}, 95% CI=[{ci[0]:.4f}, {ci[1]:.4f}], n={n}")

# =============================================================================
# 2. TASK PERFORMANCE DATA
# =============================================================================

transformation_task_results = {
    'hdin': [],
    'mean_r2': [],
    'ci_lower': [],
    'ci_upper': []
}

for hdin in hdin_values:
    r2_values = []
    
    for sess in sessions:
        try:
            results_file = f'results/temporal/data/task_temporal_session_{sess}_vth_0.000_g_1.000_rate_30_hdin_{hdin}_hdout_1.pkl'
            with open(results_file, 'rb') as f:
                results = pickle.load(f)
            r2_values.append(results[0]['test_r2_mean'])
        except FileNotFoundError:
            print(f"Warning: File not found for task - session {sess}, hdin {hdin}")
        except Exception as e:
            print(f"Error loading task data - session {sess}, hdin {hdin}: {e}")
    
    if len(r2_values) > 0:
        r2_array = np.array(r2_values)
        mean_r2 = np.mean(r2_array)
        
        # Calculate 95% confidence interval using t-distribution
        n = len(r2_array)
        std_err = stats.sem(r2_array)
        ci = stats.t.interval(0.95, n-1, loc=mean_r2, scale=std_err)
        
        transformation_task_results['hdin'].append(hdin)
        transformation_task_results['mean_r2'].append(mean_r2)
        transformation_task_results['ci_lower'].append(ci[0])
        transformation_task_results['ci_upper'].append(ci[1])
        
        print(f"Transformation Task - hdin={hdin}: mean R={mean_r2:.4f}, 95% CI=[{ci[0]:.4f}, {ci[1]:.4f}], n={n}")





classification_task_results = {
    'hdin': [],
    'mean_accuracy': [],
    'ci_lower': [],
    'ci_upper': []
}

sessions = [0,1,2,3,4, 5]
for hdin in hdin_values:
    accuracy_values = []
    
    for sess in sessions:
        try:
            results_file = f'results/categorical/data/task_categorical_session_{sess}_vth_0.000_g_1.000_rate_30_hdin_{hdin}_hdout_4.pkl'
            with open(results_file, 'rb') as f:
                results = pickle.load(f)
            accuracy_values.append(results[0]['test_accuracy_mean'])
        except FileNotFoundError:
            print(f"Warning: File not found for task - session {sess}, hdin {hdin}")
        except Exception as e:
            print(f"Error loading task data - session {sess}, hdin {hdin}: {e}")
    
    if len(accuracy_values) > 0:
        accuracy_array = np.array(accuracy_values)
        mean_accuracy = np.mean(accuracy_array)
        
        # Calculate 95% confidence interval using t-distribution
        n = len(accuracy_array)
        std_err = stats.sem(accuracy_array)
        ci = stats.t.interval(0.95, n-1, loc=mean_accuracy, scale=std_err)
        
        classification_task_results['hdin'].append(hdin)
        classification_task_results['mean_accuracy'].append(mean_accuracy)
        classification_task_results['ci_lower'].append(ci[0])
        classification_task_results['ci_upper'].append(ci[1])
        
        print(f"Classification Task - hdin={hdin}: mean R={mean_accuracy:.4f}, 95% CI=[{ci[0]:.4f}, {ci[1]:.4f}], n={n}")
        
print("\n" + "="*70)
print("DATA PROCESSING COMPLETE")
print("="*70)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, FancyArrowPatch
from matplotlib.lines import Line2D
import sys
import os

# Add project root to path
project_root = os.getcwd()
sys.path.insert(0, project_root)

# Import your network modules
from src.spiking_network import SpikingRNN
from src.rng_utils import get_rng

# Set random seed for reproducibility
np.random.seed(10)

# =============================================================================
# HELPER FUNCTIONS
# =============================================================================

def compute_participation_ratio(spike_times, n_neurons, duration, transient_time, bin_size=20.0):
    """Compute participation ratio from spike times"""
    if not spike_times:
        return 0.0
    
    # Filter out transient period
    spike_times_filtered = [(t, n) for t, n in spike_times if t >= transient_time]
    
    if not spike_times_filtered:
        return 0.0
    
    # Create time bins
    analysis_duration = duration - transient_time
    n_bins = int(analysis_duration / bin_size)
    
    # Create firing rate matrix (neurons x time_bins)
    firing_rate_matrix = np.zeros((n_neurons, n_bins))
    
    for t, neuron_id in spike_times_filtered:
        bin_idx = int((t - transient_time) / bin_size)
        if bin_idx < n_bins:
            firing_rate_matrix[neuron_id, bin_idx] += 1
    
    # Convert to rates (Hz)
    firing_rate_matrix = firing_rate_matrix / (bin_size / 1000.0)
    
    # Compute covariance matrix
    mean_rates = np.mean(firing_rate_matrix, axis=1, keepdims=True)
    centered_rates = firing_rate_matrix - mean_rates
    
    # Covariance matrix
    cov_matrix = np.cov(centered_rates)
    
    # Compute eigenvalues
    eigenvalues = np.linalg.eigvalsh(cov_matrix)
    eigenvalues = eigenvalues[eigenvalues > 0]
    
    if len(eigenvalues) == 0:
        return 0.0
    
    # Participation ratio
    pr = (np.sum(eigenvalues) ** 2) / np.sum(eigenvalues ** 2)
    
    return pr

def get_firing_rates(spike_times, n_neurons, duration, transient_time):
    """Get per-neuron firing rates excluding transient period"""
    spike_times_filtered = [(t, n) for t, n in spike_times if t >= transient_time]
    
    spike_counts = np.zeros(n_neurons)
    for _, neuron_id in spike_times_filtered:
        spike_counts[neuron_id] += 1
    
    analysis_duration = (duration - transient_time) / 1000.0
    firing_rates = spike_counts / analysis_duration
    
    return firing_rates

# =============================================================================
# STEP 1: RUN ALL SIMULATIONS AND STORE DATA
# =============================================================================

print("="*70)
print("STEP 1: RUNNING SIMULATIONS")
print("="*70)

# Simulation parameters
n_neurons = 1000
dt = 0.1
duration = 2000.0
transient_time = 200.0
session_id = 10
v_th_std = 0
g_std = 1
trial_id = 1

network_params = {
    'v_th_distribution': 'normal',
    'static_input_strength': 10.0,
    'dynamic_input_strength': 1.0,
    'readout_weight_scale': 1.0
}

# Dictionary to store all simulation results
simulation_data = {
    'n_neurons': n_neurons,
    'duration': duration,
    'transient_time': transient_time,
    'main_simulation': {},
    'dimensionality_simulations': []
}

# Main simulation for raster plot (static_input_rate = 35 Hz)
print("\n1. Running main simulation (30 Hz static input)...")
static_input_rate_main = 30.0

network_main = SpikingRNN(n_neurons=n_neurons, dt=dt, synaptic_mode="filter", 
                          static_input_mode="common_tonic")
network_main.initialize_network(session_id, v_th_std, g_std, **network_params)

spike_times_main = network_main.simulate_network_dynamics(
    session_id=session_id,
    v_th_std=v_th_std,
    g_std=g_std,
    trial_id=trial_id,
    duration=duration,
    static_input_rate=static_input_rate_main
)

simulation_data['main_simulation'] = {
    'static_input_rate': static_input_rate_main,
    'spike_times': spike_times_main,
    'n_spikes': len(spike_times_main)
}

print(f"   Complete. Total spikes: {len(spike_times_main)}")

# Dimensionality simulations (30-40 Hz)
print("\n2. Running dimensionality simulations (30-40 Hz)...")
static_rates = np.arange(30, 41, 1)

for i, rate in enumerate(static_rates):
    print(f"   [{i+1}/{len(static_rates)}] Static input rate: {rate} Hz")
    
    net_temp = SpikingRNN(n_neurons=n_neurons, dt=dt, synaptic_mode="filter",
                         static_input_mode="common_tonic")
    net_temp.initialize_network(session_id, v_th_std, g_std, **network_params)
    
    spikes_temp = net_temp.simulate_network_dynamics(
        session_id=session_id, v_th_std=v_th_std, g_std=g_std,
        trial_id=trial_id, duration=duration, static_input_rate=rate
    )
    
    simulation_data['dimensionality_simulations'].append({
        'static_input_rate': rate,
        'spike_times': spikes_temp,
        'n_spikes': len(spikes_temp)
    })
    
    print(f"       Total spikes: {len(spikes_temp)}")

print("\n" + "="*70)
print("ALL SIMULATIONS COMPLETE!")
print("="*70)




# =============================================================================
# STEP 2: PROCESS DATA AND CREATE PLOTS
# =============================================================================

print("\n" + "="*70)
print("STEP 2: CREATING PLOTS")
print("="*70)

# Extract main simulation data
spike_times_main = simulation_data['main_simulation']['spike_times']
times = [s[0] for s in spike_times_main]
neuron_ids = [s[1] for s in spike_times_main]

# Compute firing rates for main simulation
firing_rates_main = get_firing_rates(spike_times_main, n_neurons, duration, transient_time)

# Compute participation ratios with chosen bin size
bin_size_pr = 2.0  # YOU CAN CHANGE THIS WITHOUT RERUNNING SIMULATIONS
print(f"\nComputing participation ratios with bin_size={bin_size_pr} ms...")

static_rates = []
participation_ratios = []

for sim in simulation_data['dimensionality_simulations']:
    rate = sim['static_input_rate']
    spikes = sim['spike_times']
    
    pr = compute_participation_ratio(spikes, n_neurons, duration, 
                                    transient_time, bin_size=bin_size_pr)
    
    static_rates.append(rate)
    participation_ratios.append(pr)
    print(f"   Rate {rate} Hz: PR = {pr:.2f}")
```

```python
#!/usr/bin/env python3
"""
Generate conference-ready figure showing methods and results.
Combines network architecture, workflow, and performance comparison.
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch, Circle, Rectangle
from matplotlib.patches import ConnectionPatch
import pickle
import os
from collections import defaultdict

# Use a clean, professional style
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = ['Arial', 'Helvetica']
plt.rcParams['font.size'] = 9
plt.rcParams['axes.linewidth'] = 1.5


def load_results(filename):
    """Load pickle results file."""
    with open(filename, 'rb') as f:
        return pickle.load(f)


def collect_results(base_dir, pattern, sessions, hd_dims):
    """Collect results across sessions and HD dims."""
    results_by_hd = defaultdict(list)
    
    for hd_dim in hd_dims:
        for sess in sessions:
            filename = pattern.format(sess=sess, hdin=hd_dim)
            filepath = os.path.join(base_dir, filename)
            
            if os.path.exists(filepath):
                try:
                    data = load_results(filepath)
                    result = data[0] if isinstance(data, list) else data
                    test_r2 = result.get('test_r2_mean', np.nan)
                    results_by_hd[hd_dim].append(test_r2)
                except Exception as e:
                    pass
    
    return results_by_hd


def compute_statistics(results_by_hd):
    """Compute mean and std across sessions."""
    hd_dims = sorted(results_by_hd.keys())
    means = []
    stds = []
    
    for hd_dim in hd_dims:
        values = np.array(results_by_hd[hd_dim])
        values = values[~np.isnan(values)]
        
        if len(values) > 0:
            means.append(np.mean(values))
            stds.append(np.std(values))
        else:
            means.append(np.nan)
            stds.append(np.nan)
    
    return np.array(hd_dims), np.array(means), np.array(stds)


def draw_network_architecture(ax):
    """Draw spiking network architecture diagram."""
    
    # Colors
    color_input = '#3498db'  # Blue
    color_rnn = '#e74c3c'    # Red
    color_output = '#2ecc71' # Green
    color_readout = '#9b59b6' # Purple
    
    # Title
    ax.text(0.5, 0.95, 'Network Architecture', 
            ha='center', va='top', fontsize=11, fontweight='bold',
            transform=ax.transAxes)
    
    # HD Input
    input_box = FancyBboxPatch((0.05, 0.65), 0.15, 0.15,
                               boxstyle="round,pad=0.01", 
                               facecolor=color_input, edgecolor='black',
                               linewidth=2, alpha=0.7)
    ax.add_patch(input_box)
    ax.text(0.125, 0.725, 'HD Input\n$d$-dim', ha='center', va='center',
            fontsize=9, fontweight='bold', color='white')
    
    # Spiking RNN
    rnn_box = FancyBboxPatch((0.35, 0.55), 0.3, 0.35,
                            boxstyle="round,pad=0.02",
                            facecolor=color_rnn, edgecolor='black',
                            linewidth=2, alpha=0.7)
    ax.add_patch(rnn_box)
    ax.text(0.5, 0.80, 'Spiking RNN', ha='center', va='center',
            fontsize=10, fontweight='bold', color='white')
    ax.text(0.5, 0.725, '$N=1000$ neurons', ha='center', va='center',
            fontsize=8, color='white')
    
    # Add spike symbols inside RNN
    np.random.seed(42)
    for _ in range(15):
        x = 0.37 + np.random.rand() * 0.26
        y = 0.58 + np.random.rand() * 0.28
        ax.plot([x, x], [y, y+0.02], 'w-', linewidth=1.5, alpha=0.6)
    
    # Heterogeneity labels
    ax.text(0.5, 0.62, r'Heterogeneity: $\sigma_{v_{th}}, \sigma_g$',
            ha='center', va='center', fontsize=7, color='white',
            style='italic')
    
    # Readout (Linear decoder)
    readout_box = FancyBboxPatch((0.75, 0.65), 0.18, 0.15,
                                 boxstyle="round,pad=0.01",
                                 facecolor=color_readout, edgecolor='black',
                                 linewidth=2, alpha=0.7)
    ax.add_patch(readout_box)
    ax.text(0.84, 0.725, 'Linear\nReadout', ha='center', va='center',
            fontsize=9, fontweight='bold', color='white')
    
    # Arrows
    # Input -> RNN
    arrow1 = FancyArrowPatch((0.21, 0.725), (0.34, 0.725),
                            arrowstyle='->', mutation_scale=25,
                            linewidth=2.5, color='black')
    ax.add_patch(arrow1)
    
    # RNN -> Readout (filtered spikes)
    arrow2 = FancyArrowPatch((0.66, 0.725), (0.74, 0.725),
                            arrowstyle='->', mutation_scale=25,
                            linewidth=2.5, color='black')
    ax.add_patch(arrow2)
    ax.text(0.70, 0.765, 'Filtered\nSpikes', ha='center', va='bottom',
            fontsize=7, style='italic')
    
    # Exponential filter formula
    ax.text(0.5, 0.48, r'$x(t) = x(t-\Delta t) \cdot e^{-\Delta t/\tau} + s(t)$',
            ha='center', va='top', fontsize=8,
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.8, edgecolor='gray'))
    
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')


def draw_task_comparison(ax):
    """Draw comparison of temporal vs auto-encoding tasks."""
    
    # Colors
    color_temporal = '#ff6b6b'
    color_auto = '#4ecdc4'
    
    # Title
    ax.text(0.5, 0.95, 'Task Comparison', 
            ha='center', va='top', fontsize=11, fontweight='bold',
            transform=ax.transAxes)
    
    # Temporal Task (Top)
    ax.text(0.05, 0.75, 'A) Temporal Task:', ha='left', va='top',
            fontsize=10, fontweight='bold')
    
    # Input box
    input_temp = Rectangle((0.05, 0.55), 0.12, 0.12,
                           facecolor=color_temporal, edgecolor='black',
                           linewidth=1.5, alpha=0.7)
    ax.add_patch(input_temp)
    ax.text(0.11, 0.61, 'Input\n$d$-dim', ha='center', va='center',
            fontsize=8, fontweight='bold', color='white')
    
    # Arrow
    ax.annotate('', xy=(0.25, 0.61), xytext=(0.18, 0.61),
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))
    
    # Network
    net_temp = Rectangle((0.25, 0.55), 0.15, 0.12,
                         facecolor='#e74c3c', edgecolor='black',
                         linewidth=1.5, alpha=0.7)
    ax.add_patch(net_temp)
    ax.text(0.325, 0.61, 'Network', ha='center', va='center',
            fontsize=8, fontweight='bold', color='white')
    
    # Arrow
    ax.annotate('', xy=(0.48, 0.61), xytext=(0.41, 0.61),
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))
    
    # Output box (different!)
    output_temp = Rectangle((0.48, 0.55), 0.12, 0.12,
                            facecolor='#f39c12', edgecolor='black',
                            linewidth=1.5, alpha=0.7)
    ax.add_patch(output_temp)
    ax.text(0.54, 0.61, 'Output\n1-dim', ha='center', va='center',
            fontsize=8, fontweight='bold', color='white')
    
    # Label
    ax.text(0.65, 0.61, 'Different\ntarget!', ha='left', va='center',
            fontsize=8, style='italic', color=color_temporal,
            fontweight='bold')
    
    # Auto-Encoding Task (Bottom)
    ax.text(0.05, 0.42, 'B) Auto-Encoding:', ha='left', va='top',
            fontsize=10, fontweight='bold')
    
    # Input box
    input_auto = Rectangle((0.05, 0.22), 0.12, 0.12,
                           facecolor=color_auto, edgecolor='black',
                           linewidth=1.5, alpha=0.7)
    ax.add_patch(input_auto)
    ax.text(0.11, 0.28, 'Input\n$d$-dim', ha='center', va='center',
            fontsize=8, fontweight='bold', color='white')
    
    # Arrow
    ax.annotate('', xy=(0.25, 0.28), xytext=(0.18, 0.28),
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))
    
    # Network
    net_auto = Rectangle((0.25, 0.22), 0.15, 0.12,
                         facecolor='#e74c3c', edgecolor='black',
                         linewidth=1.5, alpha=0.7)
    ax.add_patch(net_auto)
    ax.text(0.325, 0.28, 'Network', ha='center', va='center',
            fontsize=8, fontweight='bold', color='white')
    
    # Arrow
    ax.annotate('', xy=(0.48, 0.28), xytext=(0.41, 0.28),
                arrowprops=dict(arrowstyle='->', lw=2, color='black'))
    
    # Output box (SAME as input!)
    output_auto = Rectangle((0.48, 0.22), 0.12, 0.12,
                            facecolor=color_auto, edgecolor='black',
                            linewidth=1.5, alpha=0.7)
    ax.add_patch(output_auto)
    ax.text(0.54, 0.28, 'Output\n$d$-dim', ha='center', va='center',
            fontsize=8, fontweight='bold', color='white')
    
    # Curved arrow showing same
    arc = mpatches.FancyBboxPatch((0.11, 0.17), 0.43, 0.28,
                                  boxstyle="round,pad=0.02",
                                  fill=False, edgecolor=color_auto,
                                  linewidth=2, linestyle='--', alpha=0.6)
    ax.add_patch(arc)
    
    ax.text(0.33, 0.13, 'Same dimensionality!\n(reconstruct input)',
            ha='center', va='top', fontsize=8, style='italic',
            color=color_auto, fontweight='bold')
    
    ax.set_xlim(0, 0.85)
    ax.set_ylim(0, 1)
    ax.axis('off')


def plot_results(ax, temporal_data, autoencoding_data):
    """Plot performance comparison results."""
    
    temp_hd, temp_mean, temp_std = temporal_data
    auto_hd, auto_mean, auto_std = autoencoding_data
    
    # Title
    ax.text(0.5, 1.05, 'Performance Comparison', 
            ha='center', va='bottom', fontsize=11, fontweight='bold',
            transform=ax.transAxes)
    
    # Plot with nice styling
    line1 = ax.errorbar(temp_hd, temp_mean, yerr=temp_std,
                        marker='o', linestyle='-', linewidth=2.5, markersize=9,
                        capsize=6, capthick=2.5, 
                        label='Temporal (1D output)',
                        color='#ff6b6b', ecolor='#ff6b6b', alpha=0.85,
                        markeredgecolor='white', markeredgewidth=1.5)
    
    line2 = ax.errorbar(auto_hd, auto_mean, yerr=auto_std,
                        marker='s', linestyle='--', linewidth=2.5, markersize=9,
                        capsize=6, capthick=2.5,
                        label='Auto-Encoding (reconstruct input)',
                        color='#4ecdc4', ecolor='#4ecdc4', alpha=0.85,
                        markeredgecolor='white', markeredgewidth=1.5)
    
    # Styling
    ax.set_xlabel('Input HD Dimensionality ($d$)', fontsize=10, fontweight='bold')
    ax.set_ylabel('Test $R^2$ (mean  std across 25 sessions)', fontsize=10, fontweight='bold')
    
    ax.legend(fontsize=9, frameon=True, shadow=True, loc='lower right',
             framealpha=0.95, edgecolor='black')
    ax.grid(True, alpha=0.25, linestyle='--', linewidth=0.8)
    
    # Set x-axis to integer ticks
    ax.set_xticks(np.arange(1, 6))
    ax.set_xlim(0.5, 5.5)
    
    # Add horizontal line at R=0
    ax.axhline(y=0, color='gray', linestyle=':', linewidth=1.5, alpha=0.5)
    
    # Add percentage improvement annotations
    for i in range(len(temp_hd)):
        if i < len(auto_hd) and not np.isnan(temp_mean[i]) and not np.isnan(auto_mean[i]):
            improvement = ((auto_mean[i] - temp_mean[i]) / abs(temp_mean[i])) * 100
            if improvement > 5:  # Only show if improvement > 5%
                y_pos = max(auto_mean[i], temp_mean[i]) + max(auto_std[i], temp_std[i]) + 0.03
                ax.text(auto_hd[i], y_pos, f'+{improvement:.0f}%',
                       ha='center', va='bottom', fontsize=7,
                       fontweight='bold', color='#2ecc71',
                       bbox=dict(boxstyle='round,pad=0.3', 
                                facecolor='white', alpha=0.8,
                                edgecolor='#2ecc71', linewidth=1))
    
    # Spine styling
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(1.5)
    ax.spines['bottom'].set_linewidth(1.5)
    
    ax.tick_params(width=1.5, length=5)


def create_conference_figure(temporal_data, autoencoding_data, save_path='conference_figure.png'):
    """Create complete conference figure."""
    
    # Create figure with custom layout
    fig = plt.figure(figsize=(14, 6))
    
    # Create grid
    gs = fig.add_gridspec(2, 3, hspace=0.4, wspace=0.3,
                         left=0.05, right=0.97, top=0.93, bottom=0.08)
    
    # Panel A: Network Architecture (spans 2 rows, 1 column)
    ax1 = fig.add_subplot(gs[:, 0])
    draw_network_architecture(ax1)
    ax1.text(-0.05, 0.98, 'A', transform=ax1.transAxes,
            fontsize=16, fontweight='bold', va='top', ha='right')
    
    # Panel B: Task Comparison (top right)
    ax2 = fig.add_subplot(gs[:, 1])
    draw_task_comparison(ax2)
    ax2.text(-0.05, 0.98, 'B', transform=ax2.transAxes,
            fontsize=16, fontweight='bold', va='top', ha='right')
    
    # Panel C: Results (bottom right, spans 2 rows)
    ax3 = fig.add_subplot(gs[:, 2])
    plot_results(ax3, temporal_data, autoencoding_data)
    ax3.text(-0.15, 1.05, 'C', transform=ax3.transAxes,
            fontsize=16, fontweight='bold', va='top', ha='right')
    
    # Add overall title
    fig.suptitle('Effect of Network Heterogeneity on HD Signal Reconstruction:\nTemporal Transformation vs Auto-Encoding',
                fontsize=13, fontweight='bold', y=0.98)
    
    # Add footer with key parameters
    footer_text = ('Parameters: $N=1000$ neurons, $\\sigma_{v_{th}}=0$, $\\sigma_g=1.0$, '
                  'input rate = 30 Hz, stimulus = 300 ms, 20-fold CV, $n=100$ trials/pattern')
    fig.text(0.5, 0.01, footer_text, ha='center', va='bottom',
            fontsize=7, style='italic', color='gray')
    
    plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
    print(f"\n{'='*70}")
    print(f"Conference figure saved to: {save_path}")
    print(f"{'='*70}")
    
    return fig


def main():
    """Main execution function."""
    
    # Configuration
    sessions = list(range(0, 25))  # 0 to 24
    hd_dims = list(range(1, 6))    # 1 to 5
    
    # File patterns
    temporal_pattern = "task_temporal_session_{sess}_vth_0.000_g_1.000_rate_30.0_hdin_{hdin}_hdout_1.pkl"
    autoencoding_pattern = "task_autoencoding_session_{sess}_vth_0.000_g_1.000_rate_30.0_hd_{hdin}.pkl"
    
    print("Loading temporal task results...")
    temporal_results = collect_results(
        "results/temporal/data",
        temporal_pattern,
        sessions,
        hd_dims
    )
    
    print("Loading auto-encoding results...")
    autoencoding_results = collect_results(
        "results/autoencoding/data",
        autoencoding_pattern,
        sessions,
        hd_dims
    )
    
    print("Computing statistics...")
    temporal_data = compute_statistics(temporal_results)
    autoencoding_data = compute_statistics(autoencoding_results)
    
    print("\nGenerating conference figure...")
    fig = create_conference_figure(
        temporal_data,
        autoencoding_data,
        save_path='conference_figure.png'
    )
    
    # Also save as PDF for publication
    fig.savefig('conference_figure.pdf', dpi=300, bbox_inches='tight', facecolor='white')
    print(f"PDF version saved to: conference_figure.pdf")
    
    plt.show()
    
    # Print summary statistics
    print("\n" + "="*70)
    print("SUMMARY STATISTICS")
    print("="*70)
    temp_hd, temp_mean, temp_std = temporal_data
    auto_hd, auto_mean, auto_std = autoencoding_data
    
    for i, hd in enumerate(temp_hd):
        print(f"\nHD Dimension = {int(hd)}:")
        print(f"  Temporal:      R = {temp_mean[i]:.4f}  {temp_std[i]:.4f}")
        if i < len(auto_hd):
            print(f"  Auto-Encoding: R = {auto_mean[i]:.4f}  {auto_std[i]:.4f}")
            improvement = ((auto_mean[i] - temp_mean[i]) / abs(temp_mean[i])) * 100
            print(f"  Improvement:   {improvement:+.1f}%")


if __name__ == "__main__":
    main()
```
