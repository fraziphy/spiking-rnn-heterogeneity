```python
"""
Jupyter notebook to run encoding experiment and visualize decoded vs ground truth signals.
Run this in a Jupyter notebook after setting up your project paths.
"""

import numpy as np
import matplotlib.pyplot as plt
import pickle
import sys
import os

# Add project paths
project_root = '/Users/frazi/github_projects/spiking-rnn-heterogeneity'  # CHANGE THIS
sys.path.insert(0, os.path.join(project_root, 'experiments'))
sys.path.insert(0, os.path.join(project_root, 'src'))
sys.path.insert(0, os.path.join(project_root, 'analysis'))

from encoding_experiment import EncodingExperiment
from hd_signal_manager import HDSignalManager

#%% ==========================================================================
# PARAMETERS - Customize these
# =============================================================================

# Network parameters
session_id = 1
v_th_std = 2.0
g_std = 2.0
hd_dim = 1
embed_dim = 1
n_neurons = 1000

# Input parameters
static_input_rate = 30.0
v_th_distribution = "uniform"
synaptic_mode = "filter"
static_input_mode = "common_tonic"
hd_input_mode = "common_tonic"

# Paths
signal_cache_dir = os.path.join(project_root, "hd_signals")

#%% ==========================================================================
# RUN EXPERIMENT
# =============================================================================

print("Initializing experiment...")
experiment = EncodingExperiment(
    n_neurons=n_neurons,
    dt=0.1,
    synaptic_mode=synaptic_mode,
    static_input_mode=static_input_mode,
    hd_input_mode=hd_input_mode,
    embed_dim=embed_dim
)

# Initialize signal manager
print("Loading/generating HD signal...")
signal_manager = HDSignalManager(signal_cache_dir)

print(f"Running experiment:")
print(f"  Session: {session_id}, v_th_std: {v_th_std}, g_std: {g_std}")
print(f"  hd_dim: {hd_dim}, embed_dim: {embed_dim}")
print(f"  This will take several minutes...")

result = experiment.run_parameter_combination(
    session_id=session_id,
    v_th_std=v_th_std,
    g_std=g_std,
    hd_dim=hd_dim,
    v_th_distribution=v_th_distribution,
    static_input_rate=static_input_rate,
    signal_manager=signal_manager
)

print(f"✓ Experiment complete!")
print(f"  Mean spikes per trial: {result['n_spikes_mean']:.0f}")
print(f"  Test RMSE: {result['decoding']['test_rmse_mean']:.4f}")
print(f"  Test R²: {result['decoding']['test_r2_mean']:.4f}")
print(f"  Test correlation: {result['decoding']['test_correlation_mean']:.4f}")

#%% ==========================================================================
# EXTRACT DATA FOR VISUALIZATION
# =============================================================================

# Ground truth HD input
ground_truth = experiment.hd_generator.Y_base  # shape (n_timesteps, embed_dim)
n_timesteps = ground_truth.shape[0]
time_axis = np.arange(n_timesteps) * 0.1  # dt = 0.1 ms

# Get all trial data
trial_results = result['trial_results']

# Get filtered spikes for all trials
from encoding_analysis import LinearDecoder

decoder = LinearDecoder(dt=0.1, tau=10.0, lambda_reg=1e-3)
spikes_trials_all = [trial['spike_times'] for trial in trial_results]
filtered_spikes = decoder.preprocess_data(
    spikes_trials_all, n_neurons, experiment.encoding_time
)

# Decode each fold's test trial using that fold's weights
n_folds = len(result['decoding']['decoder_weights'])
decoded_per_fold = []

for fold_idx in range(n_folds):
    # Get the test trial index for this fold
    test_idx = result['decoding']['test_indices'][fold_idx][0]
    
    # Get weights from this fold
    decoder.w = result['decoding']['decoder_weights'][fold_idx]
    
    # Decode only the test trial
    decoded_test = decoder.predict(filtered_spikes[test_idx:test_idx+1])
    decoded_per_fold.append((test_idx, decoded_test[0]))  # Store (trial_idx, decoded_signal)

# Sort by trial index for easier plotting
decoded_per_fold.sort(key=lambda x: x[0])
decoded_all_trials = np.array([d[1] for d in decoded_per_fold])  # (n_trials, n_steps, embed_dim)

# Verify RMSE calculation
print("\nVerifying RMSE matches stored values:")
for fold_idx in range(min(3, n_folds)):
    test_idx = result['decoding']['test_indices'][fold_idx][0]
    for ch_idx in range(min(2, embed_dim)):
        # Find where this trial ended up in decoded_per_fold
        sorted_position = [i for i, (tidx, _) in enumerate(decoded_per_fold) if tidx == test_idx][0]
        manual_rmse = np.sqrt(((decoded_all_trials[sorted_position, :, ch_idx] - 
                                ground_truth[:, ch_idx])**2).mean())
        stored_rmse = result['decoding']['test_rmse_per_fold'][fold_idx, ch_idx]
        print(f"Fold {fold_idx} (trial {test_idx}), Ch {ch_idx}: manual={manual_rmse:.6f}, stored={stored_rmse:.6f}")
print()

#%% ==========================================================================
# VISUALIZATION
# =============================================================================

# Choose which channels to plot (plot first 3 channels)
n_channels_to_plot = min(3, embed_dim)
trial_to_plot = 0  # Plot first trial

fig, axes = plt.subplots(n_channels_to_plot, 1, figsize=(12, 3*n_channels_to_plot))
if n_channels_to_plot == 1:
    axes = [axes]

for ch_idx in range(n_channels_to_plot):
    ax = axes[ch_idx]
    
    # Ground truth
    ax.plot(time_axis, ground_truth[:, ch_idx], 
            'k-', linewidth=2, label='Ground Truth', alpha=0.7)
    
    # Decoded signal
    ax.plot(time_axis, decoded_all_trials[trial_to_plot, :, ch_idx],
            'r-', linewidth=1.5, label='Decoded', alpha=0.8)
    
    # Formatting
    ax.set_xlabel('Time (ms)', fontsize=12)
    ax.set_ylabel(f'Channel {ch_idx+1}\nActivity', fontsize=12)
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.set_xlim([0, time_axis[-1]])

axes[0].set_title(f'Decoded vs Ground Truth HD Input (Trial {trial_to_plot+1})\n' + 
                  f'v_th_std={v_th_std}, g_std={g_std}, hd_dim={hd_dim}',
                  fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

#%% ==========================================================================
# PLOT ALL TRIALS OVERLAY
# =============================================================================

fig, axes = plt.subplots(n_channels_to_plot, 1, figsize=(12, 3*n_channels_to_plot))
if n_channels_to_plot == 1:
    axes = [axes]

for ch_idx in range(n_channels_to_plot):
    ax = axes[ch_idx]
    
    # Ground truth (thick black)
    ax.plot(time_axis, ground_truth[:, ch_idx], 
            'k-', linewidth=3, label='Ground Truth', alpha=0.8, zorder=100)
    
    # All decoded trials (thin, transparent)
    for trial_idx in range(decoded_all_trials.shape[0]):
        ax.plot(time_axis, decoded_all_trials[trial_idx, :, ch_idx],
                'r-', linewidth=0.5, alpha=0.3)
    
    # Mean decoded (thick red)
    mean_decoded = decoded_all_trials[:, :, ch_idx].mean(axis=0)
    ax.plot(time_axis, mean_decoded,
            'r-', linewidth=2, label='Decoded (mean)', alpha=0.9, zorder=99)
    
    # Formatting
    ax.set_xlabel('Time (ms)', fontsize=12)
    ax.set_ylabel(f'Channel {ch_idx+1}\nActivity', fontsize=12)
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.set_xlim([0, time_axis[-1]])

axes[0].set_title(f'All Trials Overlay (n={decoded_all_trials.shape[0]} trials)\n' + 
                  f'v_th_std={v_th_std}, g_std={g_std}, hd_dim={hd_dim}',
                  fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

#%% ==========================================================================
# COMPUTE PER-CHANNEL METRICS
# =============================================================================

print("\nPer-channel decoding performance:")
print("=" * 60)

test_rmse_per_channel = result['decoding']['test_rmse_per_fold'].mean(axis=0)
test_r2_per_channel = result['decoding']['test_r2_per_fold'].mean(axis=0)
test_corr_per_channel = result['decoding']['test_correlation_per_fold'].mean(axis=0)

for ch_idx in range(embed_dim):
    print(f"Channel {ch_idx+1}:")
    print(f"  RMSE: {test_rmse_per_channel[ch_idx]:.4f}")
    print(f"  R²:   {test_r2_per_channel[ch_idx]:.4f}")
    print(f"  Corr: {test_corr_per_channel[ch_idx]:.4f}")
    print()

#%% ==========================================================================
# PLOT DECODING QUALITY METRICS
# =============================================================================

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

channel_labels = [f'Ch{i+1}' for i in range(embed_dim)]

# RMSE
axes[0].bar(range(embed_dim), test_rmse_per_channel, color='steelblue', alpha=0.7)
axes[0].set_xlabel('Channel', fontsize=12)
axes[0].set_ylabel('RMSE', fontsize=12)
axes[0].set_title('Root Mean Square Error', fontsize=13)
axes[0].set_xticks(range(embed_dim))
axes[0].set_xticklabels(channel_labels)
axes[0].grid(True, alpha=0.3, axis='y')

# R²
axes[1].bar(range(embed_dim), test_r2_per_channel, color='forestgreen', alpha=0.7)
axes[1].axhline(y=0, color='k', linestyle='--', linewidth=1, alpha=0.5)
axes[1].set_xlabel('Channel', fontsize=12)
axes[1].set_ylabel('R²', fontsize=12)
axes[1].set_title('R² Score', fontsize=13)
axes[1].set_xticks(range(embed_dim))
axes[1].set_xticklabels(channel_labels)
axes[1].grid(True, alpha=0.3, axis='y')

# Correlation
axes[2].bar(range(embed_dim), test_corr_per_channel, color='coral', alpha=0.7)
axes[2].axhline(y=0, color='k', linestyle='--', linewidth=1, alpha=0.5)
axes[2].set_xlabel('Channel', fontsize=12)
axes[2].set_ylabel('Correlation', fontsize=12)
axes[2].set_title('Pearson Correlation', fontsize=13)
axes[2].set_xticks(range(embed_dim))
axes[2].set_xticklabels(channel_labels)
axes[2].set_ylim([-0.1, 1.0])
axes[2].grid(True, alpha=0.3, axis='y')

plt.suptitle(f'Decoding Performance by Channel\n' + 
             f'hd_dim={hd_dim}, v_th_std={v_th_std}, g_std={g_std}',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

#%% ==========================================================================
# OPTIONAL: SAVE RESULTS
# =============================================================================

# Uncomment to save
# output_file = f'encoding_viz_session{session_id}_hd{hd_dim}_vth{v_th_std}_g{g_std}.pkl'
# with open(output_file, 'wb') as f:
#     pickle.dump(result, f)
# print(f"Results saved to: {output_file}")
```

```python
"""
Jupyter notebook to run encoding experiment and visualize decoded vs ground truth signals.
Run this in a Jupyter notebook after setting up your project paths.
"""

import numpy as np
import matplotlib.pyplot as plt
import pickle
import sys
import os

# Add project paths
project_root = '/Users/frazi/github_projects/spiking-rnn-heterogeneity'  # CHANGE THIS
sys.path.insert(0, os.path.join(project_root, 'experiments'))
sys.path.insert(0, os.path.join(project_root, 'src'))
sys.path.insert(0, os.path.join(project_root, 'analysis'))

from encoding_experiment import EncodingExperiment
from hd_signal_manager import HDSignalManager

#%% ==========================================================================
# PARAMETERS - Customize these
# =============================================================================

# Network parameters
session_id = 40
v_th_std = 2.0
g_std = 2.0
hd_dim = 9
embed_dim = 10
n_neurons = 1000

# Input parameters
static_input_rate = 30.0
v_th_distribution = "uniform"
synaptic_mode = "filter"
static_input_mode = "common_tonic"
hd_input_mode = "common_tonic"

# Paths
signal_cache_dir = os.path.join(project_root, "hd_signals")

#%% ==========================================================================
# RUN EXPERIMENT
# =============================================================================

print("Initializing experiment...")
experiment = EncodingExperiment(
    n_neurons=n_neurons,
    dt=0.1,
    synaptic_mode=synaptic_mode,
    static_input_mode=static_input_mode,
    hd_input_mode=hd_input_mode,
    embed_dim=embed_dim
)

# Initialize signal manager
print("Loading/generating HD signal...")
signal_manager = HDSignalManager(signal_cache_dir)

print(f"Running experiment:")
print(f"  Session: {session_id}, v_th_std: {v_th_std}, g_std: {g_std}")
print(f"  hd_dim: {hd_dim}, embed_dim: {embed_dim}")
print(f"  This will take several minutes...")

result = experiment.run_parameter_combination(
    session_id=session_id,
    v_th_std=v_th_std,
    g_std=g_std,
    hd_dim=hd_dim,
    v_th_distribution=v_th_distribution,
    static_input_rate=static_input_rate,
    signal_manager=signal_manager
)

print(f"✓ Experiment complete!")
print(f"  Mean spikes per trial: {result['n_spikes_mean']:.0f}")
print(f"  Test RMSE: {result['decoding']['test_rmse_mean']:.4f}")
print(f"  Test R²: {result['decoding']['test_r2_mean']:.4f}")
print(f"  Test correlation: {result['decoding']['test_correlation_mean']:.4f}")

#%% ==========================================================================
# EXTRACT DATA FOR VISUALIZATION
# =============================================================================

# Ground truth HD input
ground_truth = experiment.hd_generator.Y_base  # shape (n_timesteps, embed_dim)
n_timesteps = ground_truth.shape[0]
time_axis = np.arange(n_timesteps) * 0.1  # dt = 0.1 ms

# Get decoded signals from first CV fold
decoder_weights = result['decoding']['decoder_weights'][0]  # First fold
trial_results = result['trial_results']

# Get filtered spikes for all trials
from encoding_analysis import LinearDecoder

decoder = LinearDecoder(dt=0.1, tau=10.0, lambda_reg=1e-3)
spikes_trials_all = [trial['spike_times'] for trial in trial_results]
filtered_spikes = decoder.preprocess_data(
    spikes_trials_all, n_neurons, experiment.encoding_time
)

# Decode using first fold weights
# Manually set weights and predict
decoder.w = decoder_weights
decoded_all_trials = decoder.predict(filtered_spikes)  # shape (n_trials, n_steps, embed_dim)

#%% ==========================================================================
# VISUALIZATION
# =============================================================================

# Choose which channels to plot (plot first 3 channels)
n_channels_to_plot = min(3, embed_dim)
trial_to_plot = 0  # Plot first trial

fig, axes = plt.subplots(n_channels_to_plot, 1, figsize=(12, 3*n_channels_to_plot))
if n_channels_to_plot == 1:
    axes = [axes]

for ch_idx in range(n_channels_to_plot):
    ax = axes[ch_idx]
    
    # Ground truth
    ax.plot(time_axis, ground_truth[:, ch_idx], 
            'k-', linewidth=2, label='Ground Truth', alpha=0.7)
    
    # Decoded signal
    ax.plot(time_axis, decoded_all_trials[trial_to_plot, :, ch_idx],
            'r-', linewidth=1.5, label='Decoded', alpha=0.8)
    
    # Formatting
    ax.set_xlabel('Time (ms)', fontsize=12)
    ax.set_ylabel(f'Channel {ch_idx+1}\nActivity', fontsize=12)
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.set_xlim([0, time_axis[-1]])

axes[0].set_title(f'Decoded vs Ground Truth HD Input (Trial {trial_to_plot+1})\n' + 
                  f'v_th_std={v_th_std}, g_std={g_std}, hd_dim={hd_dim}',
                  fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

#%% ==========================================================================
# PLOT ALL TRIALS OVERLAY
# =============================================================================

fig, axes = plt.subplots(n_channels_to_plot, 1, figsize=(12, 3*n_channels_to_plot))
if n_channels_to_plot == 1:
    axes = [axes]

for ch_idx in range(n_channels_to_plot):
    ax = axes[ch_idx]
    
    # Ground truth (thick black)
    ax.plot(time_axis, ground_truth[:, ch_idx], 
            'k-', linewidth=3, label='Ground Truth', alpha=0.8, zorder=100)
    
    # All decoded trials (thin, transparent)
    for trial_idx in range(decoded_all_trials.shape[0]):
        ax.plot(time_axis, decoded_all_trials[trial_idx, :, ch_idx],
                'r-', linewidth=0.5, alpha=0.3)
    
    # Mean decoded (thick red)
    mean_decoded = decoded_all_trials[:, :, ch_idx].mean(axis=0)
    ax.plot(time_axis, mean_decoded,
            'r-', linewidth=2, label='Decoded (mean)', alpha=0.9, zorder=99)
    
    # Formatting
    ax.set_xlabel('Time (ms)', fontsize=12)
    ax.set_ylabel(f'Channel {ch_idx+1}\nActivity', fontsize=12)
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.set_xlim([0, time_axis[-1]])

axes[0].set_title(f'All Trials Overlay (n={decoded_all_trials.shape[0]} trials)\n' + 
                  f'v_th_std={v_th_std}, g_std={g_std}, hd_dim={hd_dim}',
                  fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

#%% ==========================================================================
# COMPUTE PER-CHANNEL METRICS
# =============================================================================

print("\nPer-channel decoding performance:")
print("=" * 60)

test_rmse_per_channel = result['decoding']['test_rmse_per_fold'].mean(axis=0)
test_r2_per_channel = result['decoding']['test_r2_per_fold'].mean(axis=0)
test_corr_per_channel = result['decoding']['test_correlation_per_fold'].mean(axis=0)

for ch_idx in range(embed_dim):
    print(f"Channel {ch_idx+1}:")
    print(f"  RMSE: {test_rmse_per_channel[ch_idx]:.4f}")
    print(f"  R²:   {test_r2_per_channel[ch_idx]:.4f}")
    print(f"  Corr: {test_corr_per_channel[ch_idx]:.4f}")
    print()

#%% ==========================================================================
# PLOT DECODING QUALITY METRICS
# =============================================================================

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

channel_labels = [f'Ch{i+1}' for i in range(embed_dim)]

# RMSE
axes[0].bar(range(embed_dim), test_rmse_per_channel, color='steelblue', alpha=0.7)
axes[0].set_xlabel('Channel', fontsize=12)
axes[0].set_ylabel('RMSE', fontsize=12)
axes[0].set_title('Root Mean Square Error', fontsize=13)
axes[0].set_xticks(range(embed_dim))
axes[0].set_xticklabels(channel_labels)
axes[0].grid(True, alpha=0.3, axis='y')

# R²
axes[1].bar(range(embed_dim), test_r2_per_channel, color='forestgreen', alpha=0.7)
axes[1].axhline(y=0, color='k', linestyle='--', linewidth=1, alpha=0.5)
axes[1].set_xlabel('Channel', fontsize=12)
axes[1].set_ylabel('R²', fontsize=12)
axes[1].set_title('R² Score', fontsize=13)
axes[1].set_xticks(range(embed_dim))
axes[1].set_xticklabels(channel_labels)
axes[1].grid(True, alpha=0.3, axis='y')

# Correlation
axes[2].bar(range(embed_dim), test_corr_per_channel, color='coral', alpha=0.7)
axes[2].axhline(y=0, color='k', linestyle='--', linewidth=1, alpha=0.5)
axes[2].set_xlabel('Channel', fontsize=12)
axes[2].set_ylabel('Correlation', fontsize=12)
axes[2].set_title('Pearson Correlation', fontsize=13)
axes[2].set_xticks(range(embed_dim))
axes[2].set_xticklabels(channel_labels)
axes[2].set_ylim([-0.1, 1.0])
axes[2].grid(True, alpha=0.3, axis='y')

plt.suptitle(f'Decoding Performance by Channel\n' + 
             f'hd_dim={hd_dim}, v_th_std={v_th_std}, g_std={g_std}',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

#%% ==========================================================================
# OPTIONAL: SAVE RESULTS
# =============================================================================

# Uncomment to save
# output_file = f'encoding_viz_session{session_id}_hd{hd_dim}_vth{v_th_std}_g{g_std}.pkl'
# with open(output_file, 'wb') as f:
#     pickle.dump(result, f)
# print(f"Results saved to: {output_file}")
```
