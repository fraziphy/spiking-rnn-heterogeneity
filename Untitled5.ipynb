```python
import pickle
import numpy as np

sessions = [1, 2, 3, 4, 5]
filename_template = 'results/data/encoding_session_{}_filter_common_tonic_common_tonic_uniform_k10.pkl'

# Parameter grids
v_th_std_list = np.linspace(0.01, 4, 10)
g_std_list = np.linspace(0.01, 4, 10)
hd_dim_list = np.linspace(1, 9, 3, dtype=int)
static_input_rates = [30, 50]  # add known static input rates here

# Initialize data structure to hold the metrics by static_input_rate
metrics = ['test_rmse_mean', 'test_r2_mean', 'test_correlation_mean', 'weight_participation_ratio_mean',
           'weight_effective_dim_mean', 'decoded_participation_ratio_mean', 'decoded_effective_dim_mean']

results_grid = {}
for rate in static_input_rates:
    for v_th_std in v_th_std_list:
        for g_std in g_std_list:
            for hd_dim in hd_dim_list:
                results_grid[(rate, v_th_std, g_std, hd_dim)] = {m: [] for m in metrics}

# Aggregate results across sessions
for session in sessions:
    filename = filename_template.format(session)
    with open(filename, 'rb') as f:
        results = pickle.load(f)
    for r in results:
        key = (r['static_input_rate'], r['v_th_std'], r['g_std'], r['hd_dim'])
        if key in results_grid:
            for m in metrics:
                results_grid[key][m].append(r['decoding'][m])

# Compute mean over sessions for each metric
averaged_grid = {}
for key, metric_lists in results_grid.items():
    averaged_grid[key] = {m: np.mean(metric_lists[m]) if len(metric_lists[m]) > 0 else np.nan for m in metrics}
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec


static_input_rate_to_plot = 30
hd_dim_plot = hd_dim_list[1]
data_plot = np.full((len(v_th_std_list), len(g_std_list)), np.nan)

for i, v in enumerate(v_th_std_list):
    for j, g in enumerate(g_std_list):
        mval = averaged_grid.get((static_input_rate_to_plot, v, g, hd_dim_plot), {}).get('test_rmse_mean', np.nan)
        data_plot[i, j] = mval

# Assuming data_plot, v_th_std_list, g_std_list, hd_dim_plot from earlier code
# data_plot: shape (len(v_th_std_list), len(g_std_list))

# Marginals
marginal_vth = np.nanmean(data_plot, axis=1)  # average over g_std (for each v_th_std)
marginal_g = np.nanmean(data_plot, axis=0)    # average over v_th_std (for each g_std)

fig = plt.figure(figsize=(7, 3))
gs = gridspec.GridSpec(2, 2, width_ratios=[1, 1], height_ratios=[1, 1], wspace=0.6, hspace=1)

# Main heatmap
ax_main = fig.add_subplot(gs[:, 0])
im = ax_main.imshow(data_plot, origin='lower', aspect='auto',
                    extent=(g_std_list.min(), g_std_list.max(), v_th_std_list.min(), v_th_std_list.max()))
ax_main.set_xlabel(r'$g_{std}$')
ax_main.set_ylabel(r'$v^{th}_{std}$')
ax_main.set_title(f'Intrinsic Dim={hd_dim_plot}\n Embedding Dim=10')
ax_main.set_xticks([1,2,3,4])
ax_main.set_yticks([1,2,3,4])
cbar = fig.colorbar(im, ax=ax_main, fraction=0.045)
cbar.set_label(r"$R^{2}$", labelpad=0)
cbar.set_label("DIM", labelpad=0)

# Marginal for v_th_std (top right)
ax_marginal_vth = fig.add_subplot(gs[0, 1])
ax_marginal_vth.plot(v_th_std_list, marginal_vth, marker='o')
ax_marginal_vth.set_xlabel(r'$v^{th}_{std}$',labelpad=0)
ax_marginal_vth.set_ylabel("R²")
ax_marginal_vth.set_ylabel("DIM")
ax_marginal_vth.set_title('Marginal')
ax_marginal_vth.grid(True)

# Marginal for g_std (bottom right)
ax_marginal_g = fig.add_subplot(gs[1, 1])
ax_marginal_g.plot(g_std_list, marginal_g, marker='o')
ax_marginal_g.set_xlabel(r'$g_{std}$',labelpad=0)
ax_marginal_g.set_ylabel("R²")
# ax_marginal_g.set_title('Marginal over v_th_std')
ax_marginal_g.grid(True)

fig.savefig('./pres/HD_input_3.pdf')
# plt.tight_layout()
plt.show()
```

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

sessions = [1, 2, 3]
combinations = [(2.005, 2.005), (2.005, 4.0), (4.0, 2.005)]
static_input_rate = 30.0

# Initialize dictionaries to hold session results per combination
all_test_rmse = {combo: [] for combo in combinations}
all_hd_dim = {combo: [] for combo in combinations}

for session in sessions:
    filename = f'results/data/encoding_session_{session}_filter_common_tonic_common_tonic_uniform_k10.pkl'
    with open(filename, 'rb') as f:
        results = pickle.load(f)
        
    for combo in combinations:
        v_th_std, g_std = combo
        
        test_rmse_session = []
        hd_dim_session = []
        
        for ind in range(len(results)):
            if (
                results[ind]["v_th_std"] == v_th_std and
                results[ind]["g_std"] == g_std and
                results[ind]["static_input_rate"] == static_input_rate
            ):
                test_rmse_session.append(results[ind]["decoding"]["test_r2_mean"])
                hd_dim_session.append(results[ind]["hd_dim"])
        
        # Collect session data for this combo, if available
        if test_rmse_session and hd_dim_session:
            all_test_rmse[combo].append(test_rmse_session)
            all_hd_dim[combo].append(hd_dim_session)

# Now average across sessions for each combination
plt.figure(figsize=(8, 6))
for combo in combinations:
    # Convert session lists (list of lists) to arrays for averaging
    test_rmse_array = np.array(all_test_rmse[combo])
    hd_dim_array = np.array(all_hd_dim[combo])
    
    # Average across sessions (axis=0)
    test_rmse_mean_across = test_rmse_array.mean(axis=0)
    hd_dim_mean_across = hd_dim_array.mean(axis=0)
    
    plt.plot(hd_dim_mean_across, test_rmse_mean_across, label=f'v_th_std={combo[0]}, g_std={combo[1]}')

plt.xlabel('HD Dimension')
plt.ylabel('Test RMSE Mean')
plt.title('Mean Test RMSE vs HD Dimension across Sessions')
plt.legend()
plt.grid(True)
plt.show()
```
