```python
import pickle
import numpy as np

filename = 'results/data/encoding_session_3_filter_common_tonic_common_tonic_uniform_k10.pkl'
with open(filename, 'rb') as f:
    results = pickle.load(f)

# Basic structure
print(f"Type: {type(results)}")
print(f"Number of parameter combinations: {len(results)}")
print()

# Inspect first combination
print("="*60)
print("FIRST PARAMETER COMBINATION:")
print("="*60)
combo = results[0]

# Top-level keys
print("\nTop-level keys:")
for key in combo.keys():
    print(f"  - {key}")

# Parameters
print("\n--- Parameters ---")
print(f"v_th_std: {combo['v_th_std']}")
print(f"g_std: {combo['g_std']}")
print(f"hd_dim: {combo['hd_dim']}")
print(f"embed_dim: {combo['embed_dim']}")
print(f"static_input_rate: {combo['static_input_rate']}")

# Decoding results
print("\n--- Decoding Performance ---")
decoding = combo['decoding']
print(f"Test RMSE: {decoding['test_rmse_mean']:.4f} ± {decoding['test_rmse_std']:.4f}")
print(f"Test R²: {decoding['test_r2_mean']:.4f} ± {decoding['test_r2_std']:.4f}")
print(f"Test Correlation: {decoding['test_correlation_mean']:.4f}")

# Check if trial_results exists (contains spike times)
if 'trial_results' in combo:
    print("\n--- Trial Results (HEAVY DATA) ---")
    trial_results = combo['trial_results']
    print(f"Number of trials: {len(trial_results)}")
    
    if len(trial_results) > 0:
        trial = trial_results[0]
        print(f"\nFirst trial keys: {list(trial.keys())}")
        
        if 'spike_times' in trial:
            print(f"  Spike times: {len(trial['spike_times'])} spikes")
            print(f"  First 5 spikes: {trial['spike_times'][:5]}")
        
        if 'readout_history' in trial:
            print(f"  Readout history: {len(trial['readout_history'])} timesteps")
        
        if 'hd_input_patterns' in trial:
            patterns = trial['hd_input_patterns']
            print(f"  HD input patterns shape: {patterns.shape}")

# Check decoding details
print("\n--- Decoding Details ---")
print(f"Number of folds: {decoding['n_folds']}")
print(f"Number of neurons: {decoding['n_neurons']}")
print(f"Number of channels: {decoding['n_channels']}")

if 'decoder_weights' in decoding:
    print(f"Decoder weights stored: YES (per fold)")
    print(f"  First weight matrix shape: {decoding['decoder_weights'][0].shape}")
else:
    print(f"Decoder weights stored: NO")

if 'weight_svd_analysis' in decoding:
    svd = decoding['weight_svd_analysis'][0]
    print(f"\nWeight SVD (first fold):")
    print(f"  Effective dim (95%): {svd['effective_dim_95']}")
    print(f"  Participation ratio: {svd['participation_ratio']:.3f}")

# File size estimate
import os
file_size_mb = os.path.getsize(filename) / (1024**2)
print(f"\n{'='*60}")
print(f"FILE SIZE: {file_size_mb:.2f} MB")
print(f"{'='*60}")
```

```python
import sys

def get_size_mb(obj):
    """Get approximate size in MB"""
    return sys.getsizeof(obj) / (1024**2)

combo = results[0]
print("Storage breakdown for first combination:")
print(f"{'='*60}")

for key, value in combo.items():
    size_mb = get_size_mb(value)
    print(f"{key:30s}: {size_mb:8.3f} MB")

if 'trial_results' in combo:
    print(f"\n  Breaking down trial_results:")
    for i, trial in enumerate(combo['trial_results'][:3]):  # First 3 trials
        print(f"  Trial {i}:")
        for k, v in trial.items():
            print(f"    {k:25s}: {get_size_mb(v):8.3f} MB")
```

```python
# Create lightweight version with only summary statistics
lightweight_results = []

for combo in results:
    lightweight = {
        'v_th_std': combo['v_th_std'],
        'g_std': combo['g_std'],
        'hd_dim': combo['hd_dim'],
        'embed_dim': combo['embed_dim'],
        'static_input_rate': combo['static_input_rate'],
        
        # Keep only aggregated decoding metrics
        'test_rmse_mean': combo['decoding']['test_rmse_mean'],
        'test_rmse_std': combo['decoding']['test_rmse_std'],
        'test_r2_mean': combo['decoding']['test_r2_mean'],
        'test_r2_std': combo['decoding']['test_r2_std'],
        'test_correlation_mean': combo['decoding']['test_correlation_mean'],
        
        # Keep spikes per trial if needed
        'n_spikes_mean': combo['n_spikes_mean'],
        'n_spikes_std': combo['n_spikes_std'],
    }
    lightweight_results.append(lightweight)

# Save lightweight version
with open('encoding_lightweight.pkl', 'wb') as f:
    pickle.dump(lightweight_results, f)
```

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

sessions = [1, 2, 3]
combinations = [(2.005, 2.005), (2.005, 4.0), (4.0, 2.005)]
static_input_rate = 30.0

# Initialize dictionaries to hold session results per combination
all_test_rmse = {combo: [] for combo in combinations}
all_hd_dim = {combo: [] for combo in combinations}

for session in sessions:
    filename = f'results/data/encoding_session_{session}_filter_common_tonic_common_tonic_uniform_k10.pkl'
    with open(filename, 'rb') as f:
        results = pickle.load(f)
        
    for combo in combinations:
        v_th_std, g_std = combo
        
        test_rmse_session = []
        hd_dim_session = []
        
        for ind in range(len(results)):
            if (
                results[ind]["v_th_std"] == v_th_std and
                results[ind]["g_std"] == g_std and
                results[ind]["static_input_rate"] == static_input_rate
            ):
                test_rmse_session.append(results[ind]["decoding"]["test_r2_mean"])
                hd_dim_session.append(results[ind]["hd_dim"])
        
        # Collect session data for this combo, if available
        if test_rmse_session and hd_dim_session:
            all_test_rmse[combo].append(test_rmse_session)
            all_hd_dim[combo].append(hd_dim_session)

# Now average across sessions for each combination
plt.figure(figsize=(8, 6))
for combo in combinations:
    # Convert session lists (list of lists) to arrays for averaging
    test_rmse_array = np.array(all_test_rmse[combo])
    hd_dim_array = np.array(all_hd_dim[combo])
    
    # Average across sessions (axis=0)
    test_rmse_mean_across = test_rmse_array.mean(axis=0)
    hd_dim_mean_across = hd_dim_array.mean(axis=0)
    
    plt.plot(hd_dim_mean_across, test_rmse_mean_across, label=f'v_th_std={combo[0]}, g_std={combo[1]}')

plt.xlabel('HD Dimension')
plt.ylabel('Test RMSE Mean')
plt.title('Mean Test RMSE vs HD Dimension across Sessions')
plt.legend()
plt.grid(True)
plt.show()
```
