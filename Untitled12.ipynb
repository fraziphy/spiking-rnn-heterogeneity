```python
import os

def print_tree(start_path='.', max_depth=2, _prefix=''):
    if max_depth < 0:
        return
    for item in os.listdir(start_path):
        path = os.path.join(start_path, item)
        print(_prefix + '|-- ' + item)
        if os.path.isdir(path):
            print_tree(path, max_depth-1, _prefix + '|   ')

print_tree('.', 2)
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import pandas as pd
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 10

# ============================================================================
# CONFIGURATION
# ============================================================================

# Path to your results
RESULTS_DIR = "results/categorical_sweep/data"

# SESSION FILTERING - CHANGE THESE!
SESSION_START = 0      # First session to include
SESSION_END = 4        # Last session to include
# Set SESSION_END = None to include all available sessions

# Which accuracy metric to use
ACCURACY_METRIC = 'test_accuracy_bayesian_mean'  # or 'test_accuracy_timeaveraged_mean'

# ============================================================================
# LOAD RESULTS
# ============================================================================

def load_all_results(results_dir, session_start=None, session_end=None):
    """Load all pickle files from results directory with optional session filtering."""
    results_path = Path(results_dir)
    
    if not results_path.exists():
        print(f"ERROR: Results directory not found: {results_dir}")
        return []
    
    pkl_files = list(results_path.glob("*.pkl"))
    print(f"Found {len(pkl_files)} result files")
    
    all_results = []
    for pkl_file in pkl_files:
        try:
            with open(pkl_file, 'rb') as f:
                data = pickle.load(f)
                # Handle both single dict and list of dicts
                if isinstance(data, list):
                    all_results.extend(data)
                else:
                    all_results.append(data)
        except Exception as e:
            print(f"Warning: Could not load {pkl_file.name}: {e}")
    
    print(f"Loaded {len(all_results)} total results before filtering")
    
    # Filter by session if specified
    if session_start is not None or session_end is not None:
        filtered_results = []
        for r in all_results:
            session_id = r.get('session_id', None)
            if session_id is not None:
                include = True
                if session_start is not None and session_id < session_start:
                    include = False
                if session_end is not None and session_id > session_end:
                    include = False
                if include:
                    filtered_results.append(r)
        
        print(f"Filtered to sessions {session_start} to {session_end}: {len(filtered_results)} results")
        all_results = filtered_results
    else:
        print(f"No filtering - using all {len(all_results)} results")
    
    return all_results

# Load data with session filtering
results = load_all_results(RESULTS_DIR, session_start=SESSION_START, session_end=SESSION_END)

if len(results) == 0:
    print("ERROR: No results found! Check your RESULTS_DIR path and session range.")
else:
    print(f"\n=== DATA LOADED SUCCESSFULLY ===")
    print(f"Number of results: {len(results)}")
    
    # Show session range in data
    sessions_in_data = sorted(set([r.get('session_id', -1) for r in results]))
    print(f"Sessions in loaded data: {sessions_in_data}")

# ============================================================================
# CONVERT TO DATAFRAME
# ============================================================================

def results_to_dataframe(results, accuracy_metric):
    """Convert results list to pandas DataFrame."""
    df_data = []
    
    for r in results:
        # Determine which std to use
        if accuracy_metric == 'test_accuracy_bayesian_mean':
            std_metric = 'test_accuracy_bayesian_std'
        else:
            std_metric = 'test_accuracy_timeaveraged_std'
        
        row = {
            'session_id': r.get('session_id', np.nan),
            'v_th_std': r.get('v_th_std', np.nan),
            'g_std': r.get('g_std', np.nan),
            'static_input_rate': r.get('static_input_rate', np.nan),
            'input_hd_dim': r.get('input_hd_dim', np.nan),
            'output_hd_dim': r.get('output_hd_dim', np.nan),
            'embed_dim_input': r.get('embed_dim_input', np.nan),
            'embed_dim_output': r.get('embed_dim_output', np.nan),
            'n_input_patterns': r.get('n_input_patterns', np.nan),
            'accuracy': r.get(accuracy_metric, np.nan),
            'accuracy_std': r.get(std_metric, np.nan),
            'train_accuracy': r.get('train_accuracy_timeaveraged_mean', np.nan),
            'test_accuracy_bayesian': r.get('test_accuracy_bayesian_mean', np.nan),
            'test_accuracy_timeaveraged': r.get('test_accuracy_timeaveraged_mean', np.nan),
            'confidence': r.get('test_confidence_bayesian_mean', np.nan),
            'uncertainty': r.get('test_uncertainty_bayesian_mean', np.nan),
            'methods_agreement': r.get('test_methods_agreement_rate_mean', np.nan),
            'methods_agreement_std': r.get('test_methods_agreement_rate_std', np.nan),
        }
        df_data.append(row)
    
    df = pd.DataFrame(df_data)
    return df

if len(results) > 0:
    df = results_to_dataframe(results, ACCURACY_METRIC)
    print(f"\n=== DATAFRAME CREATED ===")
    print(f"Shape: {df.shape}")
    print(f"\nUnique values:")
    print(f"  Sessions: {sorted(df['session_id'].unique())}")
    print(f"  Embed dims: {sorted(df['embed_dim_input'].unique())}")
    print(f"  HD dims: {sorted(df['input_hd_dim'].unique())}")
    print(f"\nFirst few rows:")
    print(df[['session_id', 'embed_dim_input', 'input_hd_dim', 'accuracy']].head(10))
    print(f"\nAccuracy statistics:")
    print(df['accuracy'].describe())
else:
    print("Cannot create dataframe - no results loaded!")
    df = pd.DataFrame()

# ============================================================================
# PLOT 1: Accuracy vs HD Dimension (Main Result)
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 1: Accuracy vs HD Dimension ===")
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    embed_dims = sorted(df['embed_dim_input'].unique())
    print(f"Plotting for embed_dims: {embed_dims}")
    
    # Group by embed_dim_input and input_hd_dim
    for embed_dim in embed_dims:
        subset = df[df['embed_dim_input'] == embed_dim]
        print(f"\nEmbed={embed_dim}: {len(subset)} data points")
        
        # Aggregate across sessions (mean and SEM)
        grouped = subset.groupby('input_hd_dim').agg({
            'accuracy': ['mean', 'sem', 'count'],
            'confidence': ['mean', 'sem']
        }).reset_index()
        
        hd_dims = grouped['input_hd_dim'].values
        acc_mean = grouped['accuracy']['mean'].values
        acc_sem = grouped['accuracy']['sem'].values
        acc_count = grouped['accuracy']['count'].values
        conf_mean = grouped['confidence']['mean'].values
        conf_sem = grouped['confidence']['sem'].values
        
        # Plot accuracy
        label = f'Embed={int(embed_dim)} (n={int(acc_count[0])} sessions)'
        axes[0].plot(hd_dims, acc_mean, marker='o', label=label, linewidth=2, markersize=8)
        axes[0].fill_between(hd_dims, acc_mean - acc_sem, acc_mean + acc_sem, alpha=0.2)
        
        # Plot confidence
        axes[1].plot(hd_dims, conf_mean, marker='o', label=label, linewidth=2, markersize=8)
        axes[1].fill_between(hd_dims, conf_mean - conf_sem, conf_mean + conf_sem, alpha=0.2)
    
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    metric_name = "Bayesian" if "bayesian" in ACCURACY_METRIC else "Time-Averaged"
    
    axes[0].set_xlabel('HD Input Dimension', fontsize=12)
    axes[0].set_ylabel(f'Classification Accuracy ({metric_name})', fontsize=12)
    axes[0].set_title(f'Accuracy vs HD Dimension\n({session_range_str})', fontsize=14, fontweight='bold')
    axes[0].legend(fontsize=10)
    axes[0].grid(True, alpha=0.3)
    # axes[0].set_ylim([0, 1.05])
    
    axes[1].set_xlabel('HD Input Dimension', fontsize=12)
    axes[1].set_ylabel('Confidence', fontsize=12)
    axes[1].set_title(f'Confidence vs HD Dimension\n({session_range_str})', fontsize=14, fontweight='bold')
    axes[1].legend(fontsize=10)
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    filename = f'categorical_hd_dimension_effect_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"\nSaved: {filename}")
    plt.show()

# ============================================================================
# PLOT 2: Heatmap - Accuracy vs HD dim and Embed dim
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 2: Heatmap ===")
    
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Create pivot table
    pivot = df.groupby(['embed_dim_input', 'input_hd_dim'])['accuracy'].mean().unstack()
    
    # Sort index in reverse order (largest embed_dim at bottom)
    pivot = pivot.sort_index(ascending=False)
    
    print(f"Pivot table:\n{pivot}")
    
    # Plot heatmap with plasma colormap
    sns.heatmap(pivot, annot=True, fmt='.3f', cmap='plasma', 
                vmin=np.nanmin(pivot.values), vmax=np.nanmax(pivot.values), cbar_kws={'label': 'Accuracy'},
                linewidths=0.5, ax=ax, annot_kws={'size': 10})
    
    ax.set_xlabel('HD Input Dimension', fontsize=12)
    ax.set_ylabel('Embedding Dimension', fontsize=12)
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    metric_name = "Bayesian" if "bayesian" in ACCURACY_METRIC else "Time-Averaged"
    ax.set_title(f'Classification Accuracy Heatmap ({metric_name})\n({session_range_str})', fontsize=14, fontweight='bold')
    
    plt.tight_layout()
    filename = f'categorical_accuracy_heatmap_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 3: Comparison of Bayesian vs Time-Averaged
# ============================================================================

if len(df) > 0 and 'test_accuracy_bayesian' in df.columns and 'test_accuracy_timeaveraged' in df.columns:
    print("\n=== CREATING PLOT 3: Method Comparison ===")
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Plot 1: Bayesian vs Time-Averaged scatter
    axes[0].scatter(df['test_accuracy_bayesian'], df['test_accuracy_timeaveraged'], 
                   alpha=0.6, s=50)
    axes[0].plot([0, 1], [0, 1], 'k--', alpha=0.3, label='y=x')
    axes[0].set_xlabel('Bayesian Accuracy', fontsize=12)
    axes[0].set_ylabel('Time-Averaged Accuracy', fontsize=12)
    axes[0].set_title('Bayesian vs Time-Averaged Methods', fontsize=14, fontweight='bold')
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)
    axes[0].set_xlim([0, 1])
    axes[0].set_ylim([0, 1])
    
    # Plot 2: Histogram of methods agreement rate
    agreement_data = df['methods_agreement'].dropna()
    axes[1].hist(agreement_data, bins=50, alpha=0.7, color='steelblue', edgecolor='black')
    axes[1].axvline(agreement_data.mean(), color='red', linestyle='--', linewidth=2, 
                   label=f'Mean: {agreement_data.mean():.3f}')
    axes[1].axvline(agreement_data.median(), color='orange', linestyle='--', linewidth=2,
                   label=f'Median: {agreement_data.median():.3f}')
    
    axes[1].set_xlabel('Methods Agreement Rate', fontsize=12)
    axes[1].set_ylabel('Count', fontsize=12)
    axes[1].set_title('Distribution of Methods Agreement', fontsize=14, fontweight='bold')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3, axis='y')
    axes[1].set_xlim([0.95, 1.02])
    
    plt.tight_layout()
    filename = f'categorical_method_comparison_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 4: Session Variability (only if multiple sessions)
# ============================================================================

if len(df) > 0 and len(df['session_id'].unique()) > 1:
    print("\n=== CREATING PLOT 4: Session Variability ===")
    
    embed_dims_to_plot = sorted(df['embed_dim_input'].unique())[:4]
    n_plots = len(embed_dims_to_plot)
    
    if n_plots == 1:
        fig, axes = plt.subplots(1, 1, figsize=(8, 6))
        axes = [axes]
    elif n_plots == 2:
        fig, axes = plt.subplots(1, 2, figsize=(15, 6))
        axes = axes.flatten()
    else:
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        axes = axes.flatten()
    
    for idx, embed_dim in enumerate(embed_dims_to_plot):
        ax = axes[idx]
        subset = df[df['embed_dim_input'] == embed_dim]
        
        # Box plot for each HD dimension
        hd_dims = sorted(subset['input_hd_dim'].unique())
        data_to_plot = [subset[subset['input_hd_dim'] == hd]['accuracy'].values 
                        for hd in hd_dims]
        
        bp = ax.boxplot(data_to_plot, labels=hd_dims, patch_artist=True)
        
        # Color boxes
        for patch in bp['boxes']:
            patch.set_facecolor('lightblue')
        
        n_sessions = len(subset['session_id'].unique())
        ax.set_xlabel('HD Input Dimension', fontsize=11)
        ax.set_ylabel('Accuracy', fontsize=11)
        ax.set_title(f'Session Variability (Embed={int(embed_dim)}, n={n_sessions})', 
                    fontsize=12, fontweight='bold')
        ax.grid(True, alpha=0.3, axis='y')
        ax.set_ylim([0, 1.05])
    
    # Hide extra subplots if any
    for idx in range(len(embed_dims_to_plot), len(axes)):
        axes[idx].axis('off')
    
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    plt.suptitle(f'{session_range_str}', fontsize=14, fontweight='bold', y=0.995)
    plt.tight_layout()
    filename = f'categorical_session_variability_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()
elif len(df) > 0:
    print("\n=== SKIPPING PLOT 4: Only one session, no variability to show ===")

# ============================================================================
# SUMMARY STATISTICS
# ============================================================================

if len(df) > 0:
    metric_name = "Bayesian" if "bayesian" in ACCURACY_METRIC else "Time-Averaged"
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    
    print("\n" + "="*80)
    print(f"SUMMARY STATISTICS - {metric_name} Accuracy ({session_range_str})")
    print("="*80)
    
    print("\nBest Performance by Embedding Dimension:")
    for embed_dim in sorted(df['embed_dim_input'].unique()):
        subset = df[df['embed_dim_input'] == embed_dim]
        n_sessions = len(subset['session_id'].unique())
        best_idx = subset['accuracy'].idxmax()
        best = subset.loc[best_idx]
        print(f"\nEmbed={int(embed_dim)} (n={n_sessions} sessions):")
        print(f"  Best accuracy: {best['accuracy']:.4f} +/- {best['accuracy_std']:.4f}")
        print(f"  HD dimension: {int(best['input_hd_dim'])}")
        if not np.isnan(best.get('confidence', np.nan)):
            print(f"  Confidence: {best['confidence']:.4f}")
        if not np.isnan(best.get('methods_agreement', np.nan)):
            print(f"  Methods agreement: {best['methods_agreement']:.4f}")
    
    print("\n" + "="*80)
    print("Overall Best Performance:")
    best_idx = df['accuracy'].idxmax()
    best = df.loc[best_idx]
    print(f"  Accuracy: {best['accuracy']:.4f} +/- {best['accuracy_std']:.4f}")
    print(f"  Embed dimension: {int(best['embed_dim_input'])}")
    print(f"  HD dimension: {int(best['input_hd_dim'])}")
    print(f"  Session: {int(best['session_id'])}")
    
    # Methods agreement statistics
    if 'methods_agreement' in df.columns:
        agreement_data = df['methods_agreement'].dropna()
        print("\n" + "="*80)
        print("Methods Agreement Statistics:")
        print(f"  Mean: {agreement_data.mean():.4f}")
        print(f"  Median: {agreement_data.median():.4f}")
        print(f"  Std: {agreement_data.std():.4f}")
        print(f"  Min: {agreement_data.min():.4f}")
        print(f"  Max: {agreement_data.max():.4f}")
    
    print("="*80)

print("\n=== PLOTTING COMPLETE ===")
```

```python
# ============================================================================
# COMPARING BAYESIAN VS TIME-AVERAGED METHODS
# ============================================================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# ============================================================================
# 1. Direct Accuracy Comparison (Scatter)
# ============================================================================
ax1 = fig.add_subplot(gs[0, 0])
ax1.scatter(df['test_accuracy_bayesian'], df['test_accuracy_timeaveraged'], 
           alpha=0.6, s=80, c=df['embed_dim_input'], cmap='viridis', edgecolors='black')
ax1.plot([0, 1], [0, 1], 'k--', alpha=0.5, linewidth=2)

# Add statistics
diff = df['test_accuracy_bayesian'] - df['test_accuracy_timeaveraged']
mean_diff = diff.mean()
ax1.text(0.05, 0.95, f'Mean difference: {mean_diff:.4f}\n' + 
         f'Bayesian wins: {(diff > 0).sum()}/{len(diff)} times',
         transform=ax1.transAxes, fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

ax1.set_xlabel('Bayesian Accuracy', fontsize=11)
ax1.set_ylabel('Time-Averaged Accuracy', fontsize=11)
ax1.set_title('Direct Accuracy Comparison', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.set_xlim([0, 1])
ax1.set_ylim([0, 1])
ax1.set_aspect('equal')

# ============================================================================
# 2. Accuracy Difference by HD Dimension
# ============================================================================
ax2 = fig.add_subplot(gs[0, 1])

for embed_dim in sorted(df['embed_dim_input'].unique()):
    subset = df[df['embed_dim_input'] == embed_dim]
    grouped = subset.groupby('input_hd_dim').agg({
        'test_accuracy_bayesian': ['mean', 'sem'],
        'test_accuracy_timeaveraged': ['mean', 'sem']
    }).reset_index()
    
    hd_dims = grouped['input_hd_dim'].values
    bay_mean = grouped['test_accuracy_bayesian']['mean'].values
    time_mean = grouped['test_accuracy_timeaveraged']['mean'].values
    diff = bay_mean - time_mean
    
    ax2.plot(hd_dims, diff, marker='o', label=f'Embed={int(embed_dim)}', 
            linewidth=2, markersize=8)

ax2.axhline(y=0, color='k', linestyle='--', alpha=0.5, linewidth=2)
ax2.set_xlabel('HD Input Dimension', fontsize=11)
ax2.set_ylabel('Accuracy Difference\n(Bayesian - Time-Averaged)', fontsize=11)
ax2.set_title('Method Difference by HD Dimension', fontsize=12, fontweight='bold')
ax2.legend(fontsize=9)
ax2.grid(True, alpha=0.3)

# ============================================================================
# 3. Paired t-test Results
# ============================================================================
ax3 = fig.add_subplot(gs[0, 2])

# Calculate paired t-test for each embed dimension
t_test_results = []
for embed_dim in sorted(df['embed_dim_input'].unique()):
    subset = df[df['embed_dim_input'] == embed_dim]
    bay = subset['test_accuracy_bayesian'].values
    time = subset['test_accuracy_timeaveraged'].values
    
    t_stat, p_val = stats.ttest_rel(bay, time)
    mean_bay = bay.mean()
    mean_time = time.mean()
    
    t_test_results.append({
        'embed': embed_dim,
        't_stat': t_stat,
        'p_val': p_val,
        'bay_mean': mean_bay,
        'time_mean': mean_time,
        'winner': 'Bayesian' if mean_bay > mean_time else 'Time-Avg'
    })

# Bar plot of mean accuracies
embed_dims = [r['embed'] for r in t_test_results]
bay_means = [r['bay_mean'] for r in t_test_results]
time_means = [r['time_mean'] for r in t_test_results]

x = np.arange(len(embed_dims))
width = 0.35

bars1 = ax3.bar(x - width/2, bay_means, width, label='Bayesian', alpha=0.8, color='steelblue')
bars2 = ax3.bar(x + width/2, time_means, width, label='Time-Averaged', alpha=0.8, color='coral')

# Add significance stars
for i, result in enumerate(t_test_results):
    if result['p_val'] < 0.001:
        sig = '***'
    elif result['p_val'] < 0.01:
        sig = '**'
    elif result['p_val'] < 0.05:
        sig = '*'
    else:
        sig = 'ns'
    
    y_pos = max(result['bay_mean'], result['time_mean']) + 0.02
    ax3.text(i, y_pos, sig, ha='center', fontsize=12, fontweight='bold')

ax3.set_xlabel('Embedding Dimension', fontsize=11)
ax3.set_ylabel('Mean Accuracy', fontsize=11)
ax3.set_title('Statistical Comparison\n(* p<0.05, ** p<0.01, *** p<0.001)', 
             fontsize=12, fontweight='bold')
ax3.set_xticks(x)
ax3.set_xticklabels([int(e) for e in embed_dims])
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3, axis='y')
ax3.set_ylim([0, 1.05])

# ============================================================================
# 4. ROC Space Comparison
# ============================================================================
ax4 = fig.add_subplot(gs[1, 0])

# Calculate ROC metrics
roc_data = []
for _, row in df.iterrows():
    for r in results:
        if (r['session_id'] == row['session_id'] and 
            r['embed_dim_input'] == row['embed_dim_input'] and
            r['input_hd_dim'] == row['input_hd_dim']):
            
            # Bayesian
            cm_bay = np.mean([np.array(cm) for cm in r['cv_confusion_matrices_bayesian']], axis=0)
            tpr_bay = np.mean([cm_bay[i,i] / cm_bay[i,:].sum() for i in range(len(cm_bay))])
            fpr_bay = np.mean([(cm_bay[:,i].sum() - cm_bay[i,i]) / 
                              (cm_bay.sum() - cm_bay[i,:].sum()) for i in range(len(cm_bay))])
            
            # Time-averaged
            cm_time = np.mean([np.array(cm) for cm in r['cv_confusion_matrices_timeaveraged']], axis=0)
            tpr_time = np.mean([cm_time[i,i] / cm_time[i,:].sum() for i in range(len(cm_time))])
            fpr_time = np.mean([(cm_time[:,i].sum() - cm_time[i,i]) / 
                               (cm_time.sum() - cm_time[i,:].sum()) for i in range(len(cm_time))])
            
            roc_data.append({
                'embed_dim': row['embed_dim_input'],
                'tpr_bay': tpr_bay, 'fpr_bay': fpr_bay,
                'tpr_time': tpr_time, 'fpr_time': fpr_time
            })
            break

roc_df = pd.DataFrame(roc_data)

# Plot both methods in ROC space
colors = plt.cm.viridis(np.linspace(0, 1, len(roc_df['embed_dim'].unique())))
for idx, embed_dim in enumerate(sorted(roc_df['embed_dim'].unique())):
    subset = roc_df[roc_df['embed_dim'] == embed_dim]
    ax4.scatter(subset['fpr_bay'], subset['tpr_bay'], 
               s=100, alpha=0.7, color=colors[idx], marker='o',
               edgecolors='black', linewidth=1, label=f'Bayes (E={int(embed_dim)})')
    ax4.scatter(subset['fpr_time'], subset['tpr_time'], 
               s=100, alpha=0.7, color=colors[idx], marker='s',
               edgecolors='black', linewidth=1, label=f'Time (E={int(embed_dim)})')

ax4.plot([0, 1], [0, 1], 'k--', alpha=0.5, linewidth=2)
ax4.plot(0, 1, 'r*', markersize=15, label='Ideal', zorder=10)
ax4.set_xlabel('False Positive Rate', fontsize=11)
ax4.set_ylabel('True Positive Rate', fontsize=11)
ax4.set_title('ROC Space Comparison', fontsize=12, fontweight='bold')
ax4.legend(fontsize=7, ncol=2)
ax4.grid(True, alpha=0.3)
ax4.set_xlim([-0.05, 1.05])
ax4.set_ylim([-0.05, 1.05])

# ============================================================================
# 5. Distribution of Accuracy Differences
# ============================================================================
ax5 = fig.add_subplot(gs[1, 1])

diff = df['test_accuracy_bayesian'] - df['test_accuracy_timeaveraged']
ax5.hist(diff, bins=20, alpha=0.7, color='steelblue', edgecolor='black')
ax5.axvline(0, color='k', linestyle='--', linewidth=2, label='No difference')
ax5.axvline(diff.mean(), color='r', linestyle='-', linewidth=2, 
           label=f'Mean: {diff.mean():.4f}')

# Add statistics
t_stat, p_val = stats.ttest_1samp(diff, 0)
ax5.text(0.05, 0.95, 
         f't-statistic: {t_stat:.3f}\n' + 
         f'p-value: {p_val:.4f}\n' +
         f'{"Bayesian better" if diff.mean() > 0 else "Time-Avg better"}',
         transform=ax5.transAxes, fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

ax5.set_xlabel('Accuracy Difference (Bayesian - Time-Averaged)', fontsize=11)
ax5.set_ylabel('Count', fontsize=11)
ax5.set_title('Distribution of Method Differences', fontsize=12, fontweight='bold')
ax5.legend(fontsize=10)
ax5.grid(True, alpha=0.3, axis='y')

# ============================================================================
# 6. Methods Agreement Rate
# ============================================================================
ax6 = fig.add_subplot(gs[1, 2])

for embed_dim in sorted(df['embed_dim_input'].unique()):
    subset = df[df['embed_dim_input'] == embed_dim]
    grouped = subset.groupby('input_hd_dim').agg({
        'methods_agreement': ['mean', 'sem']
    }).reset_index()
    
    hd_dims = grouped['input_hd_dim'].values
    agree_mean = grouped['methods_agreement']['mean'].values
    agree_sem = grouped['methods_agreement']['sem'].values
    
    ax6.plot(hd_dims, agree_mean, marker='o', label=f'Embed={int(embed_dim)}', 
            linewidth=2, markersize=8)
    ax6.fill_between(hd_dims, agree_mean - agree_sem, agree_mean + agree_sem, alpha=0.2)

ax6.set_xlabel('HD Input Dimension', fontsize=11)
ax6.set_ylabel('Agreement Rate', fontsize=11)
ax6.set_title('How Often Methods Agree', fontsize=12, fontweight='bold')
ax6.legend(fontsize=9)
ax6.grid(True, alpha=0.3)
ax6.set_ylim([0, 1.05])

# ============================================================================
# 7. Win Rate by Condition
# ============================================================================
ax7 = fig.add_subplot(gs[2, :])

# Calculate win rates for each embed_dim and hd_dim combination
win_data = []
for embed_dim in sorted(df['embed_dim_input'].unique()):
    for hd_dim in sorted(df[df['embed_dim_input'] == embed_dim]['input_hd_dim'].unique()):
        subset = df[(df['embed_dim_input'] == embed_dim) & (df['input_hd_dim'] == hd_dim)]
        
        bay_wins = (subset['test_accuracy_bayesian'] > subset['test_accuracy_timeaveraged']).sum()
        time_wins = (subset['test_accuracy_bayesian'] < subset['test_accuracy_timeaveraged']).sum()
        ties = (subset['test_accuracy_bayesian'] == subset['test_accuracy_timeaveraged']).sum()
        total = len(subset)
        
        win_data.append({
            'condition': f'E{int(embed_dim)}_H{int(hd_dim)}',
            'bayesian': bay_wins / total * 100,
            'timeaveraged': time_wins / total * 100,
            'ties': ties / total * 100
        })

win_df = pd.DataFrame(win_data)
x = np.arange(len(win_df))
width = 0.25

ax7.bar(x - width, win_df['bayesian'], width, label='Bayesian Wins', 
       alpha=0.8, color='steelblue')
ax7.bar(x, win_df['ties'], width, label='Ties', alpha=0.8, color='gray')
ax7.bar(x + width, win_df['timeaveraged'], width, label='Time-Averaged Wins', 
       alpha=0.8, color='coral')

ax7.axhline(y=50, color='k', linestyle='--', alpha=0.5, linewidth=1)
ax7.set_xlabel('Condition (Embed_HDdim)', fontsize=11)
ax7.set_ylabel('Win Rate (%)', fontsize=11)
ax7.set_title('Win Rate by Condition (across sessions)', fontsize=12, fontweight='bold')
ax7.set_xticks(x)
ax7.set_xticklabels(win_df['condition'], rotation=45, ha='right')
ax7.legend(fontsize=10)
ax7.grid(True, alpha=0.3, axis='y')
ax7.set_ylim([0, 105])

plt.savefig(f'method_comparison_complete_sessions_{SESSION_START}_{SESSION_END}.png', 
           dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# Print Summary
# ============================================================================
print("\n" + "="*80)
print("METHOD COMPARISON SUMMARY")
print("="*80)

overall_diff = df['test_accuracy_bayesian'] - df['test_accuracy_timeaveraged']
bay_wins = (overall_diff > 0).sum()
time_wins = (overall_diff < 0).sum()
ties = (overall_diff == 0).sum()

print(f"\nOverall Performance:")
print(f"  Bayesian mean accuracy: {df['test_accuracy_bayesian'].mean():.4f}")
print(f"  Time-Averaged mean accuracy: {df['test_accuracy_timeaveraged'].mean():.4f}")
print(f"  Mean difference: {overall_diff.mean():.4f}")

print(f"\nHead-to-Head:")
print(f"  Bayesian wins: {bay_wins}/{len(df)} ({bay_wins/len(df)*100:.1f}%)")
print(f"  Time-Averaged wins: {time_wins}/{len(df)} ({time_wins/len(df)*100:.1f}%)")
print(f"  Ties: {ties}/{len(df)} ({ties/len(df)*100:.1f}%)")

t_stat, p_val = stats.ttest_rel(df['test_accuracy_bayesian'], df['test_accuracy_timeaveraged'])
print(f"\nStatistical Test (paired t-test):")
print(f"  t-statistic: {t_stat:.3f}")
print(f"  p-value: {p_val:.4f}")
if p_val < 0.05:
    winner = "Bayesian" if overall_diff.mean() > 0 else "Time-Averaged"
    print(f"  >> {winner} is SIGNIFICANTLY better (p < 0.05)")
else:
    print(f"  No significant difference between methods")

print("\nBy Embedding Dimension:")
for result in t_test_results:
    print(f"  Embed={int(result['embed'])}: {result['winner']} " + 
          f"(Bay: {result['bay_mean']:.3f}, Time: {result['time_mean']:.3f}, p={result['p_val']:.4f})")

print("="*80)
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import pandas as pd
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 10

# ============================================================================
# CONFIGURATION
# ============================================================================

# Path to your results
RESULTS_DIR = "results/temporal_sweep/data"

# SESSION FILTERING - CHANGE THESE!
SESSION_START = 0      # First session to include
SESSION_END = 5        # Last session to include (change to 19 for all 20 sessions)
# Set SESSION_END = None to include all available sessions

# ============================================================================
# LOAD RESULTS
# ============================================================================

def load_all_results(results_dir, session_start=None, session_end=None):
    """Load all pickle files from results directory with optional session filtering."""
    results_path = Path(results_dir)
    
    if not results_path.exists():
        print(f"ERROR: Results directory not found: {results_dir}")
        return []
    
    pkl_files = list(results_path.glob("*.pkl"))
    print(f"Found {len(pkl_files)} result files")
    
    all_results = []
    for pkl_file in pkl_files:
        try:
            with open(pkl_file, 'rb') as f:
                data = pickle.load(f)
                # Handle both single dict and list of dicts
                if isinstance(data, list):
                    all_results.extend(data)
                else:
                    all_results.append(data)
        except Exception as e:
            print(f"Warning: Could not load {pkl_file.name}: {e}")
    
    print(f"Loaded {len(all_results)} total results before filtering")
    
    # Filter by session if specified
    if session_start is not None or session_end is not None:
        filtered_results = []
        for r in all_results:
            session_id = r.get('session_id', None)
            if session_id is not None:
                include = True
                if session_start is not None and session_id < session_start:
                    include = False
                if session_end is not None and session_id > session_end:
                    include = False
                if include:
                    filtered_results.append(r)
        
        print(f"Filtered to sessions {session_start} to {session_end}: {len(filtered_results)} results")
        all_results = filtered_results
    else:
        print(f"No filtering - using all {len(all_results)} results")
    
    return all_results

# Load data with session filtering
results = load_all_results(RESULTS_DIR, session_start=SESSION_START, session_end=SESSION_END)

if len(results) == 0:
    print("ERROR: No results found! Check your RESULTS_DIR path and session range.")
else:
    print(f"\n=== DATA LOADED SUCCESSFULLY ===")
    print(f"Number of results: {len(results)}")
    
    # Show session range in data
    sessions_in_data = sorted(set([r.get('session_id', -1) for r in results]))
    print(f"Sessions in loaded data: {sessions_in_data}")

# ============================================================================
# CONVERT TO DATAFRAME
# ============================================================================

def results_to_dataframe(results):
    """Convert results list to pandas DataFrame."""
    df_data = []
    
    for r in results:
        row = {
            'session_id': r.get('session_id', np.nan),
            'v_th_std': r.get('v_th_std', np.nan),
            'g_std': r.get('g_std', np.nan),
            'static_input_rate': r.get('static_input_rate', np.nan),
            'input_hd_dim': r.get('input_hd_dim', np.nan),
            'output_hd_dim': r.get('output_hd_dim', np.nan),
            'embed_dim_input': r.get('embed_dim_input', np.nan),
            'embed_dim_output': r.get('embed_dim_output', np.nan),
            'n_input_patterns': r.get('n_input_patterns', np.nan),
            # Regression metrics
            'train_rmse': r.get('train_rmse_mean', np.nan),
            'train_rmse_std': r.get('train_rmse_std', np.nan),
            'test_rmse': r.get('test_rmse_mean', np.nan),
            'test_rmse_std': r.get('test_rmse_std', np.nan),
            'train_r2': r.get('train_r2_mean', np.nan),
            'train_r2_std': r.get('train_r2_std', np.nan),
            'test_r2': r.get('test_r2_mean', np.nan),
            'test_r2_std': r.get('test_r2_std', np.nan),
            'train_correlation': r.get('train_correlation_mean', np.nan),
            'train_correlation_std': r.get('train_correlation_std', np.nan),
            'test_correlation': r.get('test_correlation_mean', np.nan),
            'test_correlation_std': r.get('test_correlation_std', np.nan),
        }
        df_data.append(row)
    
    df = pd.DataFrame(df_data)
    return df

if len(results) > 0:
    df = results_to_dataframe(results)
    print(f"\n=== DATAFRAME CREATED ===")
    print(f"Shape: {df.shape}")
    print(f"\nUnique values:")
    print(f"  Sessions: {sorted(df['session_id'].unique())}")
    print(f"  Embed dims input: {sorted(df['embed_dim_input'].unique())}")
    print(f"  Embed dims output: {sorted(df['embed_dim_output'].unique())}")
    print(f"  HD dims input: {sorted(df['input_hd_dim'].unique())}")
    print(f"  HD dims output: {sorted(df['output_hd_dim'].unique())}")
    print(f"\nPerformance statistics:")
    print(df[['test_r2', 'test_rmse', 'test_correlation']].describe())
else:
    print("Cannot create dataframe - no results loaded!")
    df = pd.DataFrame()

# ============================================================================
# PLOT 1: R vs HD Dimensions (Main Result)
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 1: R vs HD Dimensions ===")
    
    # Get unique embed dimension pairs
    embed_pairs = df.groupby(['embed_dim_input', 'embed_dim_output']).size().reset_index()[['embed_dim_input', 'embed_dim_output']]
    
    # Create subplots for different output embedding dimensions
    n_output_embeds = len(df['embed_dim_output'].unique())
    fig, axes = plt.subplots(1, min(3, n_output_embeds), figsize=(18, 5))
    if n_output_embeds == 1:
        axes = [axes]
    elif n_output_embeds == 2:
        axes = list(axes)
    else:
        axes = axes.flatten()
    
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    
    for idx, embed_out in enumerate(sorted(df['embed_dim_output'].unique())[:3]):
        ax = axes[idx]
        subset = df[df['embed_dim_output'] == embed_out]
        
        for embed_in in sorted(subset['embed_dim_input'].unique()):
            embed_subset = subset[subset['embed_dim_input'] == embed_in]
            
            # Aggregate across sessions
            grouped = embed_subset.groupby('input_hd_dim').agg({
                'test_r2': ['mean', 'sem', 'count']
            }).reset_index()
            
            hd_dims = grouped['input_hd_dim'].values
            r2_mean = grouped['test_r2']['mean'].values
            r2_sem = grouped['test_r2']['sem'].values
            n_sessions = grouped['test_r2']['count'].values[0]
            
            ax.plot(hd_dims, r2_mean, marker='o', 
                   label=f'Input Embed={int(embed_in)} (n={int(n_sessions)})', 
                   linewidth=2, markersize=8)
            ax.fill_between(hd_dims, r2_mean - r2_sem, r2_mean + r2_sem, alpha=0.2)
        
        ax.set_xlabel('HD Input Dimension', fontsize=12)
        ax.set_ylabel('R Score', fontsize=12)
        ax.set_title(f'Output Embed={int(embed_out)}\n({session_range_str})', 
                    fontsize=12, fontweight='bold')
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3)
        ax.set_ylim([-0.1, 1.05])
        ax.axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
    
    # Hide extra subplots
    for idx in range(min(3, n_output_embeds), len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    filename = f'temporal_r2_vs_hd_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 2: Heatmaps - R for different embed combinations
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 2: R Heatmaps ===")
    
    # Create heatmaps for each output embedding dimension
    n_output_embeds = len(df['embed_dim_output'].unique())
    fig, axes = plt.subplots(1, min(3, n_output_embeds), figsize=(18, 6))
    if n_output_embeds == 1:
        axes = [axes]
    elif n_output_embeds == 2:
        axes = list(axes)
    else:
        axes = axes.flatten()
    
    for idx, embed_out in enumerate(sorted(df['embed_dim_output'].unique())[:3]):
        ax = axes[idx]
        subset = df[df['embed_dim_output'] == embed_out]
        
        # Create pivot: rows=embed_input, columns=hd_input
        pivot = subset.groupby(['embed_dim_input', 'input_hd_dim'])['test_r2'].mean().unstack()
        pivot = pivot.sort_index(ascending=False)
        
        # Plot heatmap
        sns.heatmap(pivot, annot=True, fmt='.3f', cmap='plasma', 
                    vmin=np.nanmin(pivot.values), vmax=np.nanmax(pivot.values), cbar_kws={'label': 'R Score'},
                    linewidths=0.5, ax=ax, annot_kws={'size': 9})
        
        ax.set_xlabel('HD Input Dimension', fontsize=11)
        ax.set_ylabel('Input Embedding Dimension', fontsize=11)
        ax.set_title(f'R Heatmap (Output Embed={int(embed_out)})\n({session_range_str})', 
                    fontsize=12, fontweight='bold')
    
    # Hide extra subplots
    for idx in range(min(3, n_output_embeds), len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    filename = f'temporal_r2_heatmap_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 3: Multiple Metrics Comparison
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 3: Multiple Metrics ===")
    
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    # Select one output embedding dimension for detailed analysis
    main_embed_out = sorted(df['embed_dim_output'].unique())[len(df['embed_dim_output'].unique())//2]
    subset = df[df['embed_dim_output'] == main_embed_out]
    
    metrics = [
        ('test_r2', 'R Score', axes[0, 0]),
        ('test_rmse', 'RMSE', axes[0, 1]),
        ('test_correlation', 'Correlation', axes[1, 0]),
        ('train_r2', 'Train R', axes[1, 1])
    ]
    
    for metric_name, metric_label, ax in metrics:
        for embed_in in sorted(subset['embed_dim_input'].unique()):
            embed_subset = subset[subset['embed_dim_input'] == embed_in]
            
            grouped = embed_subset.groupby('input_hd_dim').agg({
                metric_name: ['mean', 'sem']
            }).reset_index()
            
            hd_dims = grouped['input_hd_dim'].values
            metric_mean = grouped[metric_name]['mean'].values
            metric_sem = grouped[metric_name]['sem'].values
            
            ax.plot(hd_dims, metric_mean, marker='o', 
                   label=f'Embed={int(embed_in)}', linewidth=2, markersize=8)
            ax.fill_between(hd_dims, metric_mean - metric_sem, 
                           metric_mean + metric_sem, alpha=0.2)
        
        ax.set_xlabel('HD Input Dimension', fontsize=11)
        ax.set_ylabel(metric_label, fontsize=11)
        ax.set_title(f'{metric_label} vs HD Dimension\n(Output Embed={int(main_embed_out)})', 
                    fontsize=12, fontweight='bold')
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3)
        
        if 'r2' in metric_name or 'correlation' in metric_name:
            ax.set_ylim([-0.1, 1.05])
            ax.axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
    
    plt.suptitle(f'{session_range_str}', fontsize=14, fontweight='bold', y=1.00)
    plt.tight_layout()
    filename = f'temporal_multiple_metrics_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 4: 2D Heatmap - Input vs Output HD dimensions
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 4: Input vs Output HD Heatmap ===")
    
    # Select middle embedding dimensions
    mid_embed_in = sorted(df['embed_dim_input'].unique())[len(df['embed_dim_input'].unique())//2]
    mid_embed_out = sorted(df['embed_dim_output'].unique())[len(df['embed_dim_output'].unique())//2]
    
    subset = df[(df['embed_dim_input'] == mid_embed_in) & (df['embed_dim_output'] == mid_embed_out)]
    
    if len(subset) > 0:
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        
        metrics = [
            ('test_r2', 'R Score', axes[0]),
            ('test_rmse', 'RMSE', axes[1]),
            ('test_correlation', 'Correlation', axes[2])
        ]
        
        for metric_name, metric_label, ax in metrics:
            # Create pivot: rows=output_hd, columns=input_hd
            pivot = subset.groupby(['output_hd_dim', 'input_hd_dim'])[metric_name].mean().unstack()
            pivot = pivot.sort_index(ascending=False)
            
            # Determine color scale
            if 'rmse' in metric_name:
                cmap = 'plasma_r'  # Lower is better
                vmin, vmax = 0, pivot.max().max() * 1.1
            else:
                cmap = 'plasma'
                vmin, vmax = -0.2, 1.0
            
            sns.heatmap(pivot, annot=True, fmt='.3f', cmap=cmap, 
                       vmin=vmin, vmax=vmax, cbar_kws={'label': metric_label},
                       linewidths=0.5, ax=ax, annot_kws={'size': 9})
            
            ax.set_xlabel('HD Input Dimension', fontsize=11)
            ax.set_ylabel('HD Output Dimension', fontsize=11)
            ax.set_title(f'{metric_label}\n(Input Embed={int(mid_embed_in)}, Output Embed={int(mid_embed_out)})', 
                        fontsize=12, fontweight='bold')
        
        plt.suptitle(f'{session_range_str}', fontsize=14, fontweight='bold', y=1.00)
        plt.tight_layout()
        filename = f'temporal_input_output_hd_heatmap_sessions_{SESSION_START}_{SESSION_END}.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"Saved: {filename}")
        plt.show()

# ============================================================================
# PLOT 5: Session Variability (if multiple sessions)
# ============================================================================

if len(df) > 0 and len(df['session_id'].unique()) > 1:
    print("\n=== CREATING PLOT 5: Session Variability ===")
    
    # Select one embedding pair for visualization
    main_embed_in = sorted(df['embed_dim_input'].unique())[0]
    main_embed_out = sorted(df['embed_dim_output'].unique())[0]
    subset = df[(df['embed_dim_input'] == main_embed_in) & (df['embed_dim_output'] == main_embed_out)]
    
    if len(subset) > 0:
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        # Get HD dimension pairs
        hd_pairs = subset.groupby(['input_hd_dim', 'output_hd_dim']).size().reset_index()[['input_hd_dim', 'output_hd_dim']]
        hd_labels = [f'I{int(row.input_hd_dim)}_O{int(row.output_hd_dim)}' for _, row in hd_pairs.iterrows()]
        
        # Box plot for R
        data_r2 = []
        data_corr = []
        for _, pair in hd_pairs.iterrows():
            pair_data = subset[(subset['input_hd_dim'] == pair.input_hd_dim) & 
                              (subset['output_hd_dim'] == pair.output_hd_dim)]
            data_r2.append(pair_data['test_r2'].values)
            data_corr.append(pair_data['test_correlation'].values)
        
        bp1 = axes[0].boxplot(data_r2, labels=hd_labels, patch_artist=True)
        for patch in bp1['boxes']:
            patch.set_facecolor('lightblue')
        
        axes[0].set_xlabel('HD Dimension Pair (Input_Output)', fontsize=11)
        axes[0].set_ylabel('R Score', fontsize=11)
        axes[0].set_title(f'R Variability Across Sessions\n(Input Embed={int(main_embed_in)}, Output Embed={int(main_embed_out)})', 
                         fontsize=12, fontweight='bold')
        axes[0].grid(True, alpha=0.3, axis='y')
        axes[0].axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
        plt.setp(axes[0].xaxis.get_majorticklabels(), rotation=45, ha='right')
        
        bp2 = axes[1].boxplot(data_corr, labels=hd_labels, patch_artist=True)
        for patch in bp2['boxes']:
            patch.set_facecolor('lightcoral')
        
        axes[1].set_xlabel('HD Dimension Pair (Input_Output)', fontsize=11)
        axes[1].set_ylabel('Correlation', fontsize=11)
        axes[1].set_title(f'Correlation Variability Across Sessions\n(Input Embed={int(main_embed_in)}, Output Embed={int(main_embed_out)})', 
                         fontsize=12, fontweight='bold')
        axes[1].grid(True, alpha=0.3, axis='y')
        axes[1].axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
        plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=45, ha='right')
        
        plt.tight_layout()
        filename = f'temporal_session_variability_sessions_{SESSION_START}_{SESSION_END}.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"Saved: {filename}")
        plt.show()

# ============================================================================
# SUMMARY STATISTICS
# ============================================================================

if len(df) > 0:
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    
    print("\n" + "="*80)
    print(f"SUMMARY STATISTICS - Temporal Transformation ({session_range_str})")
    print("="*80)
    
    print("\nBest Performance by Embedding Dimensions:")
    for embed_out in sorted(df['embed_dim_output'].unique()):
        subset = df[df['embed_dim_output'] == embed_out]
        best_idx = subset['test_r2'].idxmax()
        best = subset.loc[best_idx]
        n_sessions = len(subset['session_id'].unique())
        
        print(f"\nOutput Embed={int(embed_out)} (n={n_sessions} sessions):")
        print(f"  Best R: {best['test_r2']:.4f} +/- {best['test_r2_std']:.4f}")
        print(f"  Input Embed: {int(best['embed_dim_input'])}, HD Input: {int(best['input_hd_dim'])}, HD Output: {int(best['output_hd_dim'])}")
        print(f"  RMSE: {best['test_rmse']:.4f}")
        print(f"  Correlation: {best['test_correlation']:.4f}")
    
    print("\n" + "="*80)
    print("Overall Best Performance:")
    best_idx = df['test_r2'].idxmax()
    best = df.loc[best_idx]
    print(f"  R: {best['test_r2']:.4f} +/- {best['test_r2_std']:.4f}")
    print(f"  RMSE: {best['test_rmse']:.4f}")
    print(f"  Correlation: {best['test_correlation']:.4f}")
    print(f"  Input Embed: {int(best['embed_dim_input'])}, Output Embed: {int(best['embed_dim_output'])}")
    print(f"  HD Input: {int(best['input_hd_dim'])}, HD Output: {int(best['output_hd_dim'])}")
    print(f"  Session: {int(best['session_id'])}")
    print("="*80)

print("\n=== PLOTTING COMPLETE ===")
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import pandas as pd
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 10

# ============================================================================
# CONFIGURATION
# ============================================================================

# Path to your results
RESULTS_DIR = "results/autoencoding_sweep/data"

# SESSION FILTERING - CHANGE THESE!
SESSION_START = 0      # First session to include
SESSION_END = 10        # Last session to include (change to 19 for all 20 sessions)
# Set SESSION_END = None to include all available sessions

# ============================================================================
# LOAD RESULTS
# ============================================================================

def load_all_results(results_dir, session_start=None, session_end=None):
    """Load all pickle files from results directory with optional session filtering."""
    results_path = Path(results_dir)
    
    if not results_path.exists():
        print(f"ERROR: Results directory not found: {results_dir}")
        return []
    
    pkl_files = list(results_path.glob("*.pkl"))
    print(f"Found {len(pkl_files)} result files")
    
    all_results = []
    for pkl_file in pkl_files:
        try:
            with open(pkl_file, 'rb') as f:
                data = pickle.load(f)
                # Handle both single dict and list of dicts
                if isinstance(data, list):
                    all_results.extend(data)
                else:
                    all_results.append(data)
        except Exception as e:
            print(f"Warning: Could not load {pkl_file.name}: {e}")
    
    print(f"Loaded {len(all_results)} total results before filtering")
    
    # Filter by session if specified
    if session_start is not None or session_end is not None:
        filtered_results = []
        for r in all_results:
            session_id = r.get('session_id', None)
            if session_id is not None:
                include = True
                if session_start is not None and session_id < session_start:
                    include = False
                if session_end is not None and session_id > session_end:
                    include = False
                if include:
                    filtered_results.append(r)
        
        print(f"Filtered to sessions {session_start} to {session_end}: {len(filtered_results)} results")
        all_results = filtered_results
    else:
        print(f"No filtering - using all {len(all_results)} results")
    
    return all_results

# Load data with session filtering
results = load_all_results(RESULTS_DIR, session_start=SESSION_START, session_end=SESSION_END)

if len(results) == 0:
    print("ERROR: No results found! Check your RESULTS_DIR path and session range.")
else:
    print(f"\n=== DATA LOADED SUCCESSFULLY ===")
    print(f"Number of results: {len(results)}")
    
    # Show session range in data
    sessions_in_data = sorted(set([r.get('session_id', -1) for r in results]))
    print(f"Sessions in loaded data: {sessions_in_data}")

# ============================================================================
# CONVERT TO DATAFRAME
# ============================================================================

def results_to_dataframe(results):
    """Convert results list to pandas DataFrame."""
    df_data = []
    
    for r in results:
        row = {
            'session_id': r.get('session_id', np.nan),
            'v_th_std': r.get('v_th_std', np.nan),
            'g_std': r.get('g_std', np.nan),
            'static_input_rate': r.get('static_input_rate', np.nan),
            'input_hd_dim': r.get('input_hd_dim', np.nan),
            'output_hd_dim': r.get('output_hd_dim', np.nan),
            'embed_dim_input': r.get('embed_dim_input', np.nan),
            'n_input_patterns': r.get('n_input_patterns', np.nan),
            # Regression metrics
            'train_rmse': r.get('train_rmse_mean', np.nan),
            'train_rmse_std': r.get('train_rmse_std', np.nan),
            'test_rmse': r.get('test_rmse_mean', np.nan),
            'test_rmse_std': r.get('test_rmse_std', np.nan),
            'train_r2': r.get('train_r2_mean', np.nan),
            'train_r2_std': r.get('train_r2_std', np.nan),
            'test_r2': r.get('test_r2_mean', np.nan),
            'test_r2_std': r.get('test_r2_std', np.nan),
            'train_correlation': r.get('train_correlation_mean', np.nan),
            'train_correlation_std': r.get('train_correlation_std', np.nan),
            'test_correlation': r.get('test_correlation_mean', np.nan),
            'test_correlation_std': r.get('test_correlation_std', np.nan),
        }
        df_data.append(row)
    
    df = pd.DataFrame(df_data)
    return df

if len(results) > 0:
    df = results_to_dataframe(results)
    print(f"\n=== DATAFRAME CREATED ===")
    print(f"Shape: {df.shape}")
    print(f"\nUnique values:")
    print(f"  Sessions: {sorted(df['session_id'].unique())}")
    print(f"  Embed dims: {sorted(df['embed_dim_input'].unique())}")
    print(f"  HD dims: {sorted(df['input_hd_dim'].unique())}")
    print(f"\nPerformance statistics:")
    print(df[['test_r2', 'test_rmse', 'test_correlation']].describe())
else:
    print("Cannot create dataframe - no results loaded!")
    df = pd.DataFrame()

# ============================================================================
# PLOT 1: R vs HD Dimension (Main Result)
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 1: Reconstruction Performance ===")
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    
    # Plot R, RMSE, and Correlation
    for embed_dim in sorted(df['embed_dim_input'].unique()):
        subset = df[df['embed_dim_input'] == embed_dim]
        
        # Aggregate across sessions
        grouped = subset.groupby('input_hd_dim').agg({
            'test_r2': ['mean', 'sem', 'count'],
            'test_rmse': ['mean', 'sem'],
            'test_correlation': ['mean', 'sem']
        }).reset_index()
        
        hd_dims = grouped['input_hd_dim'].values
        n_sessions = grouped['test_r2']['count'].values[0]
        
        # R
        r2_mean = grouped['test_r2']['mean'].values
        r2_sem = grouped['test_r2']['sem'].values
        axes[0].plot(hd_dims, r2_mean, marker='o', 
                    label=f'Embed={int(embed_dim)} (n={int(n_sessions)})', 
                    linewidth=2, markersize=8)
        axes[0].fill_between(hd_dims, r2_mean - r2_sem, r2_mean + r2_sem, alpha=0.2)
        
        # RMSE
        rmse_mean = grouped['test_rmse']['mean'].values
        rmse_sem = grouped['test_rmse']['sem'].values
        axes[1].plot(hd_dims, rmse_mean, marker='o', 
                    label=f'Embed={int(embed_dim)}', linewidth=2, markersize=8)
        axes[1].fill_between(hd_dims, rmse_mean - rmse_sem, rmse_mean + rmse_sem, alpha=0.2)
        
        # Correlation
        corr_mean = grouped['test_correlation']['mean'].values
        corr_sem = grouped['test_correlation']['sem'].values
        axes[2].plot(hd_dims, corr_mean, marker='o', 
                    label=f'Embed={int(embed_dim)}', linewidth=2, markersize=8)
        axes[2].fill_between(hd_dims, corr_mean - corr_sem, corr_mean + corr_sem, alpha=0.2)
    
    # R plot
    axes[0].set_xlabel('HD Dimension', fontsize=12)
    axes[0].set_ylabel('R Score', fontsize=12)
    axes[0].set_title(f'Reconstruction Quality (R)\n({session_range_str})', 
                     fontsize=14, fontweight='bold')
    axes[0].legend(fontsize=10)
    axes[0].grid(True, alpha=0.3)
    # axes[0].set_ylim([-0.1, 1.05])
    # axes[0].axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
    
    # RMSE plot
    axes[1].set_xlabel('HD Dimension', fontsize=12)
    axes[1].set_ylabel('RMSE', fontsize=12)
    axes[1].set_title(f'Reconstruction Error (RMSE)\n({session_range_str})', 
                     fontsize=14, fontweight='bold')
    axes[1].legend(fontsize=10)
    axes[1].grid(True, alpha=0.3)
    
    # Correlation plot
    axes[2].set_xlabel('HD Dimension', fontsize=12)
    axes[2].set_ylabel('Correlation', fontsize=12)
    axes[2].set_title(f'Reconstruction Correlation\n({session_range_str})', 
                     fontsize=14, fontweight='bold')
    axes[2].legend(fontsize=10)
    axes[2].grid(True, alpha=0.3)
    # axes[2].set_ylim([-0.1, 1.05])
    # axes[2].axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
    
    plt.tight_layout()
    filename = f'autoencoding_performance_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 2: Heatmap - R vs Embed and HD dimensions
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 2: Performance Heatmap ===")
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    
    metrics = [
        ('test_r2', 'R Score', 'plasma', -0.2, 1.0),
        ('test_rmse', 'RMSE', 'plasma_r', None, None),
        ('test_correlation', 'Correlation', 'plasma', -0.2, 1.0)
    ]
    
    for idx, (metric, label, cmap, vmin, vmax) in enumerate(metrics):
        # Create pivot: rows=embed_dim, columns=hd_dim
        pivot = df.groupby(['embed_dim_input', 'input_hd_dim'])[metric].mean().unstack()
        pivot = pivot.sort_index(ascending=False)
        
        
        sns.heatmap(pivot, annot=True, fmt='.3f', cmap=cmap, 
                    vmin=np.nanmin(pivot.values), vmax=np.nanmax(pivot.values), cbar_kws={'label': label},
                    linewidths=0.5, ax=axes[idx], annot_kws={'size': 10})
        
        axes[idx].set_xlabel('HD Dimension', fontsize=12)
        axes[idx].set_ylabel('Embedding Dimension', fontsize=12)
        axes[idx].set_title(f'{label} Heatmap\n({session_range_str})', 
                           fontsize=14, fontweight='bold')
    
    plt.tight_layout()
    filename = f'autoencoding_heatmap_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 3: HD Dimension = Embedding Dimension Analysis
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 3: Diagonal Analysis (HD = Embed) ===")
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    # Extract diagonal: where HD dim = Embed dim
    diagonal_data = []
    for _, row in df.iterrows():
        if row['input_hd_dim'] == row['embed_dim_input']:
            diagonal_data.append(row)
    
    if len(diagonal_data) > 0:
        diag_df = pd.DataFrame(diagonal_data)
        
        # Aggregate across sessions
        diag_grouped = diag_df.groupby('embed_dim_input').agg({
            'test_r2': ['mean', 'sem'],
            'test_rmse': ['mean', 'sem'],
            'test_correlation': ['mean', 'sem']
        }).reset_index()
        
        embed_dims = diag_grouped['embed_dim_input'].values
        
        # Plot 1: R and Correlation at diagonal
        ax1 = axes[0]
        r2_mean = diag_grouped['test_r2']['mean'].values
        r2_sem = diag_grouped['test_r2']['sem'].values
        corr_mean = diag_grouped['test_correlation']['mean'].values
        corr_sem = diag_grouped['test_correlation']['sem'].values
        
        ax1.plot(embed_dims, r2_mean, marker='o', label='R', 
                linewidth=2, markersize=10, color='steelblue')
        ax1.fill_between(embed_dims, r2_mean - r2_sem, r2_mean + r2_sem, 
                        alpha=0.2, color='steelblue')
        
        ax1.plot(embed_dims, corr_mean, marker='s', label='Correlation', 
                linewidth=2, markersize=10, color='coral')
        ax1.fill_between(embed_dims, corr_mean - corr_sem, corr_mean + corr_sem, 
                        alpha=0.2, color='coral')
        
        ax1.set_xlabel('Dimension (HD = Embed)', fontsize=12)
        ax1.set_ylabel('Score', fontsize=12)
        ax1.set_title('Performance at HD = Embed Dimension\n(Diagonal)', 
                     fontsize=14, fontweight='bold')
        ax1.legend(fontsize=11)
        ax1.grid(True, alpha=0.3)
        ax1.set_ylim([-0.1, 1.05])
        ax1.axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
        
        # Plot 2: RMSE at diagonal
        ax2 = axes[1]
        rmse_mean = diag_grouped['test_rmse']['mean'].values
        rmse_sem = diag_grouped['test_rmse']['sem'].values
        
        ax2.plot(embed_dims, rmse_mean, marker='o', 
                linewidth=2, markersize=10, color='crimson')
        ax2.fill_between(embed_dims, rmse_mean - rmse_sem, rmse_mean + rmse_sem, 
                        alpha=0.2, color='crimson')
        
        ax2.set_xlabel('Dimension (HD = Embed)', fontsize=12)
        ax2.set_ylabel('RMSE', fontsize=12)
        ax2.set_title('Reconstruction Error at HD = Embed\n(Diagonal)', 
                     fontsize=14, fontweight='bold')
        ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    filename = f'autoencoding_diagonal_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 4: Below vs At vs Above Diagonal
# ============================================================================

if len(df) > 0:
    print("\n=== CREATING PLOT 4: Performance by HD vs Embed Relationship ===")
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    # Categorize each result
    df['hd_embed_relation'] = 'Equal'
    df.loc[df['input_hd_dim'] < df['embed_dim_input'], 'hd_embed_relation'] = 'HD < Embed'
    df.loc[df['input_hd_dim'] > df['embed_dim_input'], 'hd_embed_relation'] = 'HD > Embed'
    
    relations = ['HD < Embed', 'Equal', 'HD > Embed']
    colors = ['steelblue', 'orange', 'crimson']
    
    metrics = [
        ('test_r2', 'R Score', axes[0]),
        ('test_rmse', 'RMSE', axes[1]),
        ('test_correlation', 'Correlation', axes[2])
    ]
    
    for metric_name, metric_label, ax in metrics:
        data_to_plot = []
        for relation in relations:
            subset = df[df['hd_embed_relation'] == relation]
            if len(subset) > 0:
                data_to_plot.append(subset[metric_name].values)
            else:
                data_to_plot.append([])
        
        bp = ax.boxplot(data_to_plot, labels=relations, patch_artist=True)
        
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
            patch.set_alpha(0.7)
        
        ax.set_xlabel('HD vs Embedding Dimension', fontsize=11)
        ax.set_ylabel(metric_label, fontsize=11)
        ax.set_title(f'{metric_label} by Relationship', fontsize=12, fontweight='bold')
        ax.grid(True, alpha=0.3, axis='y')
        
        if 'r2' in metric_name or 'correlation' in metric_name:
            ax.axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
    
    plt.tight_layout()
    filename = f'autoencoding_relationship_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()

# ============================================================================
# PLOT 5: Session Variability (if multiple sessions)
# ============================================================================

if len(df) > 0 and len(df['session_id'].unique()) > 1:
    print("\n=== CREATING PLOT 5: Session Variability ===")
    
    # For each embedding dimension, show variability across HD dimensions
    n_embeds = len(df['embed_dim_input'].unique())
    n_rows = (n_embeds + 1) // 2
    
    fig, axes = plt.subplots(n_rows, 2, figsize=(15, 5*n_rows))
    if n_rows == 1:
        axes = axes.reshape(1, -1)
    axes = axes.flatten()
    
    for idx, embed_dim in enumerate(sorted(df['embed_dim_input'].unique())):
        ax = axes[idx]
        subset = df[df['embed_dim_input'] == embed_dim]
        
        hd_dims = sorted(subset['input_hd_dim'].unique())
        data_to_plot = [subset[subset['input_hd_dim'] == hd]['test_r2'].values 
                       for hd in hd_dims]
        
        bp = ax.boxplot(data_to_plot, labels=hd_dims, patch_artist=True)
        
        for patch in bp['boxes']:
            patch.set_facecolor('lightblue')
        
        n_sessions = len(subset['session_id'].unique())
        ax.set_xlabel('HD Dimension', fontsize=11)
        ax.set_ylabel('R Score', fontsize=11)
        ax.set_title(f'R Variability (Embed={int(embed_dim)}, n={n_sessions} sessions)', 
                    fontsize=12, fontweight='bold')
        ax.grid(True, alpha=0.3, axis='y')
        ax.axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
        ax.set_ylim([-0.2, 1.05])
    
    # Hide extra subplots
    for idx in range(n_embeds, len(axes)):
        axes[idx].axis('off')
    
    plt.suptitle(f'{session_range_str}', fontsize=14, fontweight='bold', y=1.00)
    plt.tight_layout()
    filename = f'autoencoding_session_variability_sessions_{SESSION_START}_{SESSION_END}.png'
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    print(f"Saved: {filename}")
    plt.show()


# ============================================================================
# PLOT 6: Dimensionality Heatmaps (2ms bin only)
# ============================================================================

if len(df) > 0 and 'dimensionality_summary' in results[0]:
    print("\n=== CREATING PLOT 6: Dimensionality Analysis (2ms bin) ===")
    
    # Extract dimensionality data for 2ms bin ONLY
    dim_data = []
    for r in results:
        if (r['session_id'] >= SESSION_START and 
            (SESSION_END is None or r['session_id'] <= SESSION_END)):
            
            dim_summary = r.get('dimensionality_summary', {})
            
            # Only extract 2ms bin data
            if 'bin_2.0ms' in dim_summary:
                metrics = dim_summary['bin_2.0ms']
                dim_data.append({
                    'session_id': r['session_id'],
                    'embed_dim': r['embed_dim_input'],
                    'hd_dim': r['input_hd_dim'],
                    'participation_ratio': metrics.get('participation_ratio_mean', np.nan),
                    'participation_ratio_std': metrics.get('participation_ratio_std', np.nan),
                    'effective_dim': metrics.get('effective_dimensionality_mean', np.nan),
                    'effective_dim_std': metrics.get('effective_dimensionality_std', np.nan),
                })
    
    if len(dim_data) > 0:
        dim_df = pd.DataFrame(dim_data)
        print(f"Found {len(dim_df)} results with 2ms bin data")
        
        fig, axes = plt.subplots(2, 2, figsize=(16, 14))
        
        # ================================================================
        # Plot 1: Participation Ratio Heatmap
        # ================================================================
        ax = axes[0, 0]
        pivot = dim_df.groupby(['embed_dim', 'hd_dim'])['participation_ratio'].mean().unstack()
        pivot = pivot.sort_index(ascending=False)
        
        sns.heatmap(pivot, annot=True, fmt='.2f', cmap='viridis', 
                    cbar_kws={'label': 'Participation Ratio'},
                    linewidths=0.5, ax=ax, annot_kws={'size': 9})
        
        ax.set_xlabel('HD Dimension', fontsize=11)
        ax.set_ylabel('Embedding Dimension', fontsize=11)
        ax.set_title(f'Participation Ratio Heatmap\n(2ms bin, {session_range_str})', 
                    fontsize=12, fontweight='bold')
        
        # ================================================================
        # Plot 2: Effective Dimensionality Heatmap
        # ================================================================
        ax = axes[0, 1]
        pivot = dim_df.groupby(['embed_dim', 'hd_dim'])['effective_dim'].mean().unstack()
        pivot = pivot.sort_index(ascending=False)
        
        sns.heatmap(pivot, annot=True, fmt='.2f', cmap='viridis', 
                    cbar_kws={'label': 'Effective Dimensionality'},
                    linewidths=0.5, ax=ax, annot_kws={'size': 9})
        
        ax.set_xlabel('HD Dimension', fontsize=11)
        ax.set_ylabel('Embedding Dimension', fontsize=11)
        ax.set_title(f'Effective Dimensionality Heatmap\n(2ms bin, {session_range_str})', 
                    fontsize=12, fontweight='bold')
        
        # ================================================================
        # Plot 3: Participation Ratio vs HD Dimension
        # ================================================================
        ax = axes[1, 0]
        
        for embed_dim in sorted(dim_df['embed_dim'].unique()):
            subset = dim_df[dim_df['embed_dim'] == embed_dim]
            
            grouped = subset.groupby('hd_dim').agg({
                'participation_ratio': ['mean', 'sem']
            }).reset_index()
            
            hd_dims = grouped['hd_dim'].values
            pr_mean = grouped['participation_ratio']['mean'].values
            pr_sem = grouped['participation_ratio']['sem'].values
            
            ax.plot(hd_dims, pr_mean, marker='o', 
                   label=f'Embed={int(embed_dim)}', linewidth=2, markersize=8)
            ax.fill_between(hd_dims, pr_mean - pr_sem, pr_mean + pr_sem, alpha=0.2)
        
        # Add diagonal reference line (HD = PR in ideal case)
        max_hd = dim_df['hd_dim'].max()
        ax.plot([0, max_hd], [0, max_hd], 'k--', alpha=0.3, linewidth=2, label='y=x (ideal)')
        
        ax.set_xlabel('HD Dimension', fontsize=11)
        ax.set_ylabel('Participation Ratio', fontsize=11)
        ax.set_title('Participation Ratio vs HD Dimension\n(2ms bin)', 
                    fontsize=12, fontweight='bold')
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3)
        
        # ================================================================
        # Plot 4: Dimensionality vs R Performance
        # ================================================================
        ax = axes[1, 1]
        
        # Merge with performance data
        perf_dim = df.merge(
            dim_df[['session_id', 'embed_dim', 'hd_dim', 'participation_ratio', 'effective_dim']],
            left_on=['session_id', 'embed_dim_input', 'input_hd_dim'],
            right_on=['session_id', 'embed_dim', 'hd_dim'],
            how='inner'
        )
        
        # Scatter plot colored by embedding dimension
        colors = plt.cm.viridis(np.linspace(0, 1, len(perf_dim['embed_dim_input'].unique())))
        
        for idx, embed_dim in enumerate(sorted(perf_dim['embed_dim_input'].unique())):
            subset = perf_dim[perf_dim['embed_dim_input'] == embed_dim]
            ax.scatter(subset['participation_ratio'], subset['test_r2'],
                      label=f'Embed={int(embed_dim)}', s=80, alpha=0.7,
                      color=colors[idx], edgecolors='black', linewidth=0.5)
        
        ax.set_xlabel('Participation Ratio', fontsize=11)
        ax.set_ylabel('R Score', fontsize=11)
        ax.set_title('Dimensionality vs Performance\n(Does higher PR lead to better reconstruction?)', 
                    fontsize=12, fontweight='bold')
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3)
        ax.axhline(y=0, color='k', linestyle='--', alpha=0.3, linewidth=1)
        
        # Add correlation
        valid = perf_dim[['participation_ratio', 'test_r2']].dropna()
        if len(valid) > 2:
            corr, p_val = stats.pearsonr(valid['participation_ratio'], valid['test_r2'])
            ax.text(0.05, 0.95, f'Correlation: {corr:.3f}\np-value: {p_val:.4f}',
                   transform=ax.transAxes, fontsize=10, verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
        
        plt.tight_layout()
        filename = f'autoencoding_dimensionality_2ms_sessions_{SESSION_START}_{SESSION_END}.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"Saved: {filename}")
        plt.show()
        
        # ================================================================
        # Additional: Diagonal Analysis with Dimensionality
        # ================================================================
        print("\n=== CREATING PLOT 7: Diagonal Dimensionality (2ms bin) ===")
        
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        # Extract diagonal data (HD = Embed)
        diag_dim = dim_df[dim_df['hd_dim'] == dim_df['embed_dim']]
        
        if len(diag_dim) > 0:
            # Aggregate
            diag_grouped = diag_dim.groupby('embed_dim').agg({
                'participation_ratio': ['mean', 'sem'],
                'effective_dim': ['mean', 'sem']
            }).reset_index()
            
            embed_dims = diag_grouped['embed_dim'].values
            
            # Plot 1: Participation Ratio at diagonal
            ax = axes[0]
            pr_mean = diag_grouped['participation_ratio']['mean'].values
            pr_sem = diag_grouped['participation_ratio']['sem'].values
            
            ax.plot(embed_dims, pr_mean, marker='o', linewidth=2, markersize=10, 
                   color='steelblue', label='Participation Ratio')
            ax.fill_between(embed_dims, pr_mean - pr_sem, pr_mean + pr_sem, 
                           alpha=0.2, color='steelblue')
            
            # Add ideal line (PR = dimension)
            ax.plot(embed_dims, embed_dims, 'k--', alpha=0.5, linewidth=2, label='Ideal (PR = Dim)')
            
            ax.set_xlabel('Dimension (HD = Embed)', fontsize=12)
            ax.set_ylabel('Participation Ratio', fontsize=12)
            ax.set_title('Neural Dimensionality at Diagonal\n(HD = Embed, 2ms bin)', 
                        fontsize=14, fontweight='bold')
            ax.legend(fontsize=11)
            ax.grid(True, alpha=0.3)
            
            # Plot 2: Effective Dimensionality at diagonal
            ax = axes[1]
            ed_mean = diag_grouped['effective_dim']['mean'].values
            ed_sem = diag_grouped['effective_dim']['sem'].values
            
            ax.plot(embed_dims, ed_mean, marker='s', linewidth=2, markersize=10, 
                   color='coral', label='Effective Dimensionality')
            ax.fill_between(embed_dims, ed_mean - ed_sem, ed_mean + ed_sem, 
                           alpha=0.2, color='coral')
            
            # Add ideal line
            ax.plot(embed_dims, embed_dims, 'k--', alpha=0.5, linewidth=2, label='Ideal (Eff Dim = Dim)')
            
            ax.set_xlabel('Dimension (HD = Embed)', fontsize=12)
            ax.set_ylabel('Effective Dimensionality', fontsize=12)
            ax.set_title('Effective Dimensionality at Diagonal\n(HD = Embed, 2ms bin)', 
                        fontsize=14, fontweight='bold')
            ax.legend(fontsize=11)
            ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        filename = f'autoencoding_diagonal_dimensionality_2ms_sessions_{SESSION_START}_{SESSION_END}.png'
        plt.savefig(filename, dpi=300, bbox_inches='tight')
        print(f"Saved: {filename}")
        plt.show()
        
        # ================================================================
        # Print dimensionality statistics
        # ================================================================
        print("\n" + "="*80)
        print("DIMENSIONALITY STATISTICS (2ms bin)")
        print("="*80)
        print("\nDiagonal (HD = Embed) Performance:")
        if len(diag_dim) > 0:
            for dim in sorted(diag_dim['embed_dim'].unique()):
                subset = diag_dim[diag_dim['embed_dim'] == dim]
                pr_mean = subset['participation_ratio'].mean()
                print(f"  Dimension {int(dim)}:")
                print(f"    Participation Ratio: {pr_mean:.3f}  {subset['participation_ratio'].std():.3f}")
                print(f"    Effective Dim: {subset['effective_dim'].mean():.3f}  {subset['effective_dim'].std():.3f}")
                print(f"    Efficiency: {pr_mean / dim * 100:.1f}% (using {pr_mean:.1f} of {int(dim)} dimensions)")
        
        print("\nOverall Statistics:")
        print(f"  Mean Participation Ratio: {dim_df['participation_ratio'].mean():.3f}")
        print(f"  Mean Effective Dimensionality: {dim_df['effective_dim'].mean():.3f}")
        print("="*80)
    
    else:
        print("No 2ms bin data found in results")
else:
    print("No dimensionality_summary found in results")
# ============================================================================
# SUMMARY STATISTICS
# ============================================================================

if len(df) > 0:
    session_range_str = f"Sessions {SESSION_START}-{SESSION_END}" if SESSION_END else f"Sessions {SESSION_START}+"
    
    print("\n" + "="*80)
    print(f"SUMMARY STATISTICS - Autoencoding ({session_range_str})")
    print("="*80)
    
    print("\nBest Performance by Embedding Dimension:")
    for embed_dim in sorted(df['embed_dim_input'].unique()):
        subset = df[df['embed_dim_input'] == embed_dim]
        best_idx = subset['test_r2'].idxmax()
        best = subset.loc[best_idx]
        n_sessions = len(subset['session_id'].unique())
        
        print(f"\nEmbed={int(embed_dim)} (n={n_sessions} sessions):")
        print(f"  Best R: {best['test_r2']:.4f} +/- {best['test_r2_std']:.4f}")
        print(f"  HD dimension: {int(best['input_hd_dim'])}")
        print(f"  RMSE: {best['test_rmse']:.4f}")
        print(f"  Correlation: {best['test_correlation']:.4f}")
        print(f"  HD vs Embed: {'Equal' if best['input_hd_dim'] == embed_dim else ('Below' if best['input_hd_dim'] < embed_dim else 'Above')}")
    
    print("\n" + "="*80)
    print("Overall Best Performance:")
    best_idx = df['test_r2'].idxmax()
    best = df.loc[best_idx]
    print(f"  R: {best['test_r2']:.4f} +/- {best['test_r2_std']:.4f}")
    print(f"  RMSE: {best['test_rmse']:.4f}")
    print(f"  Correlation: {best['test_correlation']:.4f}")
    print(f"  Embedding dimension: {int(best['embed_dim_input'])}")
    print(f"  HD dimension: {int(best['input_hd_dim'])}")
    print(f"  Session: {int(best['session_id'])}")
    
    # Diagonal performance
    diagonal_df = df[df['input_hd_dim'] == df['embed_dim_input']]
    if len(diagonal_df) > 0:
        print("\n" + "="*80)
        print("Performance at HD = Embed (Diagonal):")
        print(f"  Mean R: {diagonal_df['test_r2'].mean():.4f}")
        print(f"  Mean RMSE: {diagonal_df['test_rmse'].mean():.4f}")
        print(f"  Mean Correlation: {diagonal_df['test_correlation'].mean():.4f}")
    
    print("="*80)

print("\n=== PLOTTING COMPLETE ===")
```
