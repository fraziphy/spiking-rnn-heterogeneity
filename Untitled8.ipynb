```python
#!/usr/bin/env python3
"""
Plot comparison of temporal task vs auto-encoding task.
Shows test R as a function of input HD dimensionality.
"""

import numpy as np
import matplotlib.pyplot as plt
import pickle
import os
from glob import glob
from collections import defaultdict

def load_results(filename):
    """Load pickle results file."""
    with open(filename, 'rb') as f:
        return pickle.load(f)

def collect_results(base_dir, pattern, sessions, hd_dims):
    """
    Collect results for given pattern across sessions and HD dims.
    
    Args:
        base_dir: Base directory (e.g., 'results/temporal/data')
        pattern: Filename pattern with {sess} and {hdin} placeholders
        sessions: List of session IDs
        hd_dims: List of HD dimensions
    
    Returns:
        Dictionary: {hd_dim: [test_r2_values_across_sessions]}
    """
    results_by_hd = defaultdict(list)
    
    for hd_dim in hd_dims:
        for sess in sessions:
            # Format filename
            filename = pattern.format(sess=sess, hdin=hd_dim)
            filepath = os.path.join(base_dir, filename)
            
            if os.path.exists(filepath):
                try:
                    data = load_results(filepath)
                    # Results are stored as list with single dict
                    if isinstance(data, list) and len(data) > 0:
                        result = data[0]
                    else:
                        result = data
                    
                    test_r2 = result.get('test_r2_mean', np.nan)
                    results_by_hd[hd_dim].append(test_r2)
                except Exception as e:
                    print(f"Warning: Could not load {filepath}: {e}")
            else:
                print(f"Warning: File not found: {filepath}")
    
    return results_by_hd

def compute_statistics(results_by_hd):
    """
    Compute mean and std across sessions for each HD dim.
    
    Returns:
        hd_dims, means, stds, n_sessions
    """
    hd_dims = sorted(results_by_hd.keys())
    means = []
    stds = []
    n_sessions_list = []
    
    for hd_dim in hd_dims:
        values = np.array(results_by_hd[hd_dim])
        # Remove NaN values
        values = values[~np.isnan(values)]
        
        if len(values) > 0:
            means.append(np.mean(values))
            stds.append(np.std(values))
            n_sessions_list.append(len(values))
        else:
            means.append(np.nan)
            stds.append(np.nan)
            n_sessions_list.append(0)
    
    return np.array(hd_dims), np.array(means), np.array(stds), np.array(n_sessions_list)

def plot_comparison(temporal_data, autoencoding_data, save_path=None):
    """
    Plot temporal vs auto-encoding comparison.
    
    Args:
        temporal_data: Tuple of (hd_dims, means, stds, n_sessions) for temporal task
        autoencoding_data: Tuple of (hd_dims, means, stds, n_sessions) for auto-encoding
        save_path: Path to save figure (optional)
    """
    temp_hd, temp_mean, temp_std, temp_n = temporal_data
    auto_hd, auto_mean, auto_std, auto_n = autoencoding_data
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot temporal task
    ax.errorbar(temp_hd, temp_mean, yerr=temp_std, 
                marker='o', linestyle='-', linewidth=2, markersize=8,
                capsize=5, capthick=2, label='Temporal (1D output)',
                color='#2E86AB', ecolor='#2E86AB', alpha=0.8)
    
    # Plot auto-encoding
    ax.errorbar(auto_hd, auto_mean, yerr=auto_std,
                marker='s', linestyle='--', linewidth=2, markersize=8,
                capsize=5, capthick=2, label='Auto-encoding (reconstruct input)',
                color='#A23B72', ecolor='#A23B72', alpha=0.8)
    
    # Styling
    ax.set_xlabel('Input HD Dimensionality', fontsize=14, fontweight='bold')
    ax.set_ylabel('Test R (mean  std across sessions)', fontsize=14, fontweight='bold')
    ax.set_title('Temporal Task vs Auto-Encoding: Reconstruction Performance\n' +
                 'v_th=0.0, g=1.0, rate=30 Hz',
                 fontsize=15, fontweight='bold', pad=20)
    
    ax.legend(fontsize=12, frameon=True, shadow=True, loc='best')
    ax.grid(True, alpha=0.3, linestyle='--')
    
    # Set x-axis to show actual HD dimensions
    all_hd = sorted(set(list(temp_hd) + list(auto_hd)))
    ax.set_xticks(all_hd)
    
    # # Add horizontal line at R=0
    # ax.axhline(y=0, color='gray', linestyle=':', linewidth=1, alpha=0.5)
    
    # # Add horizontal line at R=1 (perfect)
    # ax.axhline(y=1, color='gray', linestyle=':', linewidth=1, alpha=0.5)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Figure saved to: {save_path}")
    
    plt.show()

def print_summary_table(temporal_data, autoencoding_data):
    """Print summary statistics table."""
    temp_hd, temp_mean, temp_std, temp_n = temporal_data
    auto_hd, auto_mean, auto_std, auto_n = autoencoding_data
    
    print("\n" + "="*80)
    print("SUMMARY TABLE: Test R (mean  std)")
    print("="*80)
    print(f"{'HD Dim':<10} {'Temporal Task':<30} {'Auto-Encoding':<30}")
    print("-"*80)
    
    # Find common HD dimensions
    all_hd = sorted(set(list(temp_hd) + list(auto_hd)))
    
    for hd in all_hd:
        temp_idx = np.where(temp_hd == hd)[0]
        auto_idx = np.where(auto_hd == hd)[0]
        
        temp_str = f"{temp_mean[temp_idx[0]]:.4f}  {temp_std[temp_idx[0]]:.4f} (n={temp_n[temp_idx[0]]})" if len(temp_idx) > 0 else "N/A"
        auto_str = f"{auto_mean[auto_idx[0]]:.4f}  {auto_std[auto_idx[0]]:.4f} (n={auto_n[auto_idx[0]]})" if len(auto_idx) > 0 else "N/A"
        
        print(f"{int(hd):<10} {temp_str:<30} {auto_str:<30}")
    
    print("="*80)
    
    # Print comparison for overlapping HD dimensions
    print("\nPerformance Comparison (for overlapping HD dimensions):")
    for hd in all_hd:
        temp_idx = np.where(temp_hd == hd)[0]
        auto_idx = np.where(auto_hd == hd)[0]
        
        if len(temp_idx) > 0 and len(auto_idx) > 0:
            t_mean = temp_mean[temp_idx[0]]
            a_mean = auto_mean[auto_idx[0]]
            
            if not np.isnan(t_mean) and not np.isnan(a_mean):
                if a_mean > t_mean:
                    pct_change = (a_mean - t_mean) / abs(t_mean) * 100
                    print(f"  HD={int(hd)}: Auto-encoding is {pct_change:.1f}% better (R: {a_mean:.4f} vs {t_mean:.4f})")
                elif a_mean < t_mean:
                    pct_change = (t_mean - a_mean) / abs(a_mean) * 100
                    print(f"  HD={int(hd)}: Temporal is {pct_change:.1f}% better (R: {t_mean:.4f} vs {a_mean:.4f})")
                else:
                    print(f"  HD={int(hd)}: Equal performance (R: {t_mean:.4f})")

def main():
    """Main execution function."""
    
    # Configuration - UPDATED based on your files
    sessions = list(range(0, 8))  # 0 to 24
    
    # HD dimensions - DIFFERENT for each task!
    temporal_hd_dims = list(range(1, 6))  # 10, 11, 12, 13 (based on your files)
    autoencoding_hd_dims = list(range(1, 6))  # 1, 2, 3, 4, 5
    
    # File patterns - FIXED (no .0 after rate)
    temporal_pattern = "task_temporal_session_{sess}_vth_0.000_g_1.000_rate_30_hdin_{hdin}_hdout_1.pkl"
    autoencoding_pattern = "task_autoencoding_session_{sess}_vth_0.000_g_1.000_rate_30_hd_{hdin}.pkl"
    
    print("Collecting temporal task results...")
    temporal_results = collect_results(
        "results/temporal/data",
        temporal_pattern,
        sessions,
        temporal_hd_dims
    )
    
    print("Collecting auto-encoding results...")
    autoencoding_results = collect_results(
        "results/autoencoding/data",
        autoencoding_pattern,
        sessions,
        autoencoding_hd_dims
    )
    
    # Compute statistics
    print("\nComputing statistics...")
    temporal_data = compute_statistics(temporal_results)
    autoencoding_data = compute_statistics(autoencoding_results)
    
    # Print summary
    print_summary_table(temporal_data, autoencoding_data)
    
    # Plot
    print("\nGenerating plot...")
    plot_comparison(
        temporal_data,
        autoencoding_data,
        save_path="temporal_vs_autoencoding_comparison.png"
    )
    
    # Print data availability
    print("\n" + "="*80)
    print("DATA AVAILABILITY:")
    print("="*80)
    print("\nTemporal Task:")
    for hd in temporal_hd_dims:
        count = len(temporal_results[hd])
        print(f"  HD={hd}: {count}/25 sessions")
    
    print("\nAuto-Encoding Task:")
    for hd in autoencoding_hd_dims:
        count = len(autoencoding_results[hd])
        print(f"  HD={hd}: {count}/25 sessions")

if __name__ == "__main__":
    main()
```

```python
"""
Identity Mapping Test - Using Your Actual Code
Tests if network can learn identity mapping (output = input)
"""

#%% Cell 1: Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

project_root = Path('/Users/frazi/github_projects/spiking-rnn-heterogeneity')
sys.path.insert(0, str(project_root))

from src.spiking_network import SpikingRNN
from src.hd_input import HDInputGenerator
from sklearn.linear_model import Ridge
from sklearn.model_selection import KFold

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

print(" Imports successful")
```

```python
#%% Cell 2: Configuration
N_NEURONS = 500
N_READOUT = 10
V_TH_STD = 1.0
G_STD = 1.0
STATIC_INPUT_RATE = 40.0
DT = 0.1

SIGNAL_DIM = 3  # HD intrinsic dimensionality
N_INPUT_CHANNELS = 8  # HD embedding dimensionality
STIMULUS_DURATION = 300.0
N_TRIALS = 30

LAMBDA_REG = 1e-3
N_CV_FOLDS = 5

print(f"Config: {N_NEURONS} neurons, {SIGNAL_DIM}D signal, {N_TRIALS} trials")
```

```python
#%% Cell 3: Create Signals (Identity Mapping)

def create_identity_signals(signal_dim, duration_ms, dt):
    """Create input = output for identity test."""
    n_timesteps = int(duration_ms / dt)
    time = np.arange(n_timesteps) * dt
    
    input_signal = np.zeros((n_timesteps, signal_dim))
    for dim in range(signal_dim):
        freq = (dim + 1) * 2 * np.pi / duration_ms
        input_signal[:, dim] = 0.5 + 0.5 * np.sin(freq * time)
    
    output_signal = input_signal.copy()  # IDENTITY
    
    return input_signal, output_signal, time

input_signal, output_signal, time = create_identity_signals(SIGNAL_DIM, STIMULUS_DURATION, DT)

print(f" Signals: {input_signal.shape}")
```

```python
#%% Cell 4: Visualize Signals

fig, ax = plt.subplots(figsize=(14, 6))
for dim in range(SIGNAL_DIM):
    ax.plot(time, input_signal[:, dim], '--', label=f'Input {dim+1}', alpha=0.5, linewidth=2)
    ax.plot(time, output_signal[:, dim], '-', label=f'Target {dim+1}', linewidth=2)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Value')
ax.set_title('Identity Mapping Test: Output = Input')
ax.legend(ncol=2)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('identity_signals.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
#%% Cell 5: Create Network

# Create network WITHOUT readout synapses (we'll decode offline)
network = SpikingRNN(
    n_neurons=N_NEURONS,
    n_readout_neurons=N_READOUT,
    dt=DT,
    synaptic_mode='filter',
    static_input_mode='independent',
    hd_input_mode='independent',
    n_hd_channels=N_INPUT_CHANNELS,
    use_readout_synapses=False  # Don't need readout synapses for offline decoding
)

# Create HD input generator
hd_gen = HDInputGenerator(embed_dim=N_INPUT_CHANNELS, dt=DT, signal_cache_dir='test_hd_signals')

print(" Network created")
```

```python
#%% Cell 6: Initialize Network Structure

SESSION_ID = 0

# Initialize network parameters (structure - fixed across trials)
network.initialize_network(
    session_id=SESSION_ID,
    v_th_std=V_TH_STD,
    g_std=G_STD,
    v_th_distribution='normal',
    hd_dim=SIGNAL_DIM,
    embed_dim=N_INPUT_CHANNELS,
    hd_connection_prob=0.3,
    hd_input_strength=3.0,
    static_input_strength=10.0
)

# Initialize HD signal base (the embedding that maps signal to channels)
hd_gen.initialize_base_input(
    session_id=SESSION_ID,
    hd_dim=SIGNAL_DIM,
    pattern_id=0,
    rate_rnn_params={'n_neurons': 1000, 'T': 500.0, 'g': 2.0}
)

print(" Network initialized")
```

```python
#%% Cell 7: Generate HD Input Patterns

# The HD generator creates a random embedding from d-dimensional to k-dimensional space
# We need to map our custom signal through this embedding

# Get the base HD pattern
base_hd = hd_gen.Y_base  # Shape: (n_timesteps_base, N_INPUT_CHANNELS)

# We need to create input rates that when decoded give our desired signal
# Simple approach: scale the base pattern to match our signal shape
n_signal_timesteps = len(input_signal)
n_base_timesteps = base_hd.shape[0]

# Resample/interpolate base pattern to match signal length
from scipy.interpolate import interp1d

resampled_hd = np.zeros((n_signal_timesteps, N_INPUT_CHANNELS))
for ch in range(N_INPUT_CHANNELS):
    f = interp1d(np.linspace(0, 1, n_base_timesteps), base_hd[:, ch], kind='linear')
    resampled_hd[:, ch] = f(np.linspace(0, 1, n_signal_timesteps))

# Normalize to rates (0-100 Hz)
hd_input_rates = resampled_hd - resampled_hd.min()
hd_input_rates = hd_input_rates / (hd_input_rates.max() + 1e-10) * 2.0

print(f" HD input rates: {hd_input_rates.shape}")
```

```python
#%% Cell 8: Simulate Trials

print(f"\nSimulating {N_TRIALS} trials...")

all_spike_trains = []
all_outputs = []

for trial_idx in range(N_TRIALS):
    if (trial_idx + 1) % 10 == 0:
        print(f"  Trial {trial_idx + 1}/{N_TRIALS}")
    
    # Reset for new trial
    network.reset_simulation(SESSION_ID, V_TH_STD, G_STD, trial_idx)
    
    # Simulate
    n_steps = int(STIMULUS_DURATION / DT)
    spike_times_dict = {i: [] for i in range(N_NEURONS)}
    
    for step in range(n_steps):
        # Get HD rates for this timestep
        hd_rates = hd_input_rates[step] if step < len(hd_input_rates) else np.zeros(N_INPUT_CHANNELS)
        
        # Step network
        spike_indices, _ = network.step(
            session_id=SESSION_ID,
            v_th_std=V_TH_STD,
            g_std=G_STD,
            trial_id=trial_idx,
            static_input_rate=STATIC_INPUT_RATE,
            hd_input_rates=hd_rates,
            hd_dim=SIGNAL_DIM,
            embed_dim=N_INPUT_CHANNELS,
            time_step=step
        )
        
        # Record spikes
        for neuron_id in spike_indices:
            spike_times_dict[neuron_id].append(step * DT)
    
    # Convert to list of arrays
    spike_train = [np.array(spike_times_dict[i]) for i in range(N_NEURONS)]
    
    all_spike_trains.append(spike_train)
    all_outputs.append(output_signal)

print(" Simulation complete")
```

```python
#%% Cell 10: Convert to Rate Matrices

X_all, y_all = [], []
rate_window = 20.0

for trial_idx in range(N_TRIALS):
    spike_train = all_spike_trains[trial_idx]
    target = all_outputs[trial_idx]
    
    rate_matrix = np.zeros((len(target), N_NEURONS))
    
    for t_idx in range(len(target)):
        t_center = t_idx * DT
        t_start = max(0, t_center - rate_window/2)
        t_end = min(STIMULUS_DURATION, t_center + rate_window/2)
        
        for neuron_idx in range(N_NEURONS):
            spike_count = np.sum((spike_train[neuron_idx] >= t_start) & (spike_train[neuron_idx] < t_end))
            rate_matrix[t_idx, neuron_idx] = spike_count / (rate_window / 1000.0)
    
    X_all.append(rate_matrix)
    y_all.append(target)

X = np.vstack(X_all)
y = np.vstack(y_all)

print(f" Data: X={X.shape}, y={y.shape}")
```

```python
#%% Cell 11: Train Decoder

decoder = Ridge(alpha=LAMBDA_REG)
decoder.fit(X, y)
y_pred = decoder.predict(X)

train_rmse = np.sqrt(np.mean((y - y_pred)**2))
train_r2 = 1 - np.sum((y - y_pred)**2) / np.sum((y - y.mean(axis=0))**2)

print(f"\n Training: RMSE={train_rmse:.4f}, R={train_r2:.4f}")

#%% Cell 12: Cross-Validation

kfold = KFold(n_splits=N_CV_FOLDS, shuffle=True, random_state=42)
cv_rmse, cv_r2 = [], []

for fold_idx, (train_idx, test_idx) in enumerate(kfold.split(X)):
    fold_decoder = Ridge(alpha=LAMBDA_REG)
    fold_decoder.fit(X[train_idx], y[train_idx])
    y_pred_fold = fold_decoder.predict(X[test_idx])
    
    rmse = np.sqrt(np.mean((y[test_idx] - y_pred_fold)**2))
    r2 = 1 - np.sum((y[test_idx] - y_pred_fold)**2) / np.sum((y[test_idx] - y[test_idx].mean(axis=0))**2)
    
    cv_rmse.append(rmse)
    cv_r2.append(r2)
    print(f"  Fold {fold_idx+1}: RMSE={rmse:.4f}, R={r2:.4f}")

print(f"\n CV: RMSE={np.mean(cv_rmse):.4f}{np.std(cv_rmse):.4f}, R={np.mean(cv_r2):.4f}{np.std(cv_r2):.4f}")

#%% Cell 13: Visualize Results

trial_length = len(time)
y_pred_trials = y_pred.reshape(N_TRIALS, trial_length, SIGNAL_DIM)
y_true_trials = y.reshape(N_TRIALS, trial_length, SIGNAL_DIM)

fig, axes = plt.subplots(SIGNAL_DIM, 1, figsize=(14, 4*SIGNAL_DIM))
if SIGNAL_DIM == 1:
    axes = [axes]

for dim in range(SIGNAL_DIM):
    ax = axes[dim]
    
    target_mean = y_true_trials[:, :, dim].mean(axis=0)
    target_std = y_true_trials[:, :, dim].std(axis=0)
    decoded_mean = y_pred_trials[:, :, dim].mean(axis=0)
    decoded_std = y_pred_trials[:, :, dim].std(axis=0)
    
    ax.plot(time, target_mean, 'k-', linewidth=2, label='Target')
    ax.fill_between(time, target_mean - target_std, target_mean + target_std, color='gray', alpha=0.2)
    ax.plot(time, decoded_mean, 'r-', linewidth=2, label='Decoded')
    ax.fill_between(time, decoded_mean - decoded_std, decoded_mean + decoded_std, color='red', alpha=0.2)
    
    ax.set_ylabel(f'Dim {dim+1}')
    ax.set_title(f'Dimension {dim+1}: Identity Mapping')
    ax.legend()
    ax.grid(True, alpha=0.3)
    if dim == SIGNAL_DIM - 1:
        ax.set_xlabel('Time (ms)')

plt.tight_layout()
plt.savefig('identity_results.png', dpi=150, bbox_inches='tight')
plt.show()

#%% Cell 14: Summary

print("\n" + "="*60)
print("IDENTITY MAPPING TEST")
print("="*60)
print(f"Train: RMSE={train_rmse:.4f}, R={train_r2:.4f}")
print(f"CV:    RMSE={np.mean(cv_rmse):.4f}{np.std(cv_rmse):.4f}")
print(f"       R={np.mean(cv_r2):.4f}{np.std(cv_r2):.4f}")

if np.mean(cv_r2) > 0.7:
    print("\n EXCELLENT - Identity learned!")
elif np.mean(cv_r2) > 0.5:
    print("\n GOOD - Reasonable learning")
elif np.mean(cv_r2) > 0.3:
    print("\n MODERATE - Partial learning")
else:
    print("\n WEAK - Limited learning")
print("="*60)
```

```python
"""
Liquid State Machine (LSM) - Online-Compatible Implementation
Maass et al. (2002) approach with exact synaptic dynamics

Key: Synaptic traces computed with iterative exponential decay
     (not FFT convolution) to match online simulation exactly
"""

#%% Cell 1: Imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

project_root = Path('/Users/frazi/github_projects/spiking-rnn-heterogeneity')
sys.path.insert(0, str(project_root))

from src.spiking_network import SpikingRNN
from src.hd_input import HDInputGenerator
from analysis.encoding_analysis import LinearDecoder
from sklearn.model_selection import StratifiedKFold

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (14, 10)

print(" Imports successful")
print("\n" + "="*70)
print("LIQUID STATE MACHINE - ONLINE-COMPATIBLE")
print("Synaptic traces computed iteratively (matches online dynamics)")
print("="*70)

#%% Cell 2: Configuration

# Reservoir
N_NEURONS = 500
V_TH_STD = 1.0
G_STD = 1.0
STATIC_INPUT_RATE = 30.0
DT = 0.1

# Task - IMPORTANT: N_READOUT must match output HD dimensionality
OUTPUT_DIM_INTRINSIC = 3  # HD output dimensionality
OUTPUT_DIM_EMBEDDING = 8  # Number of output neurons (readout layer size)
N_READOUT = OUTPUT_DIM_INTRINSIC  # Match output dim!

INPUT_DIM_INTRINSIC = 3
INPUT_DIM_EMBEDDING = 8

STIMULUS_DURATION = 300.0
N_TRIALS_TRAIN = 50
N_TRIALS_TEST = 10

# Synaptic dynamics
TAU_SYN = 10.0  # Synaptic time constant (ms) - same as your Synapse class

# Training
LAMBDA_REG = 1e-3
N_CV_FOLDS = 20

print(f"\nLSM Configuration:")
print(f"  Reservoir: {N_NEURONS} spiking neurons (FIXED)")
print(f"  Output: {OUTPUT_DIM_INTRINSIC}D HD signal")
print(f"  Readout neurons: {N_READOUT} (= output_dim_intrinsic)")
print(f"  Synaptic tau: {TAU_SYN} ms")
print(f"  Training trials: {N_TRIALS_TRAIN}")
print(f"  Test trials: {N_TRIALS_TEST}")

#%% Cell 3: Helper Function - Compute Synaptic Traces (ONLINE-COMPATIBLE)

def compute_synaptic_traces_online(spike_matrix, dt, tau):
    """
    Compute synaptic traces using iterative exponential decay.
    This EXACTLY matches what happens during online simulation.
    
    Args:
        spike_matrix: (T, N_NEURONS) binary spike matrix (0 or 1)
        dt: timestep (ms)
        tau: synaptic time constant (ms)
        
    Returns:
        traces: (T, N_NEURONS) synaptic traces (the ABAB matrix)
                traces[t, i] = filtered activity of neuron i at time t
    """
    T, N = spike_matrix.shape
    traces = np.zeros((T, N))
    decay_factor = np.exp(-dt / tau)
    
    for t in range(1, T):
        # Exact synaptic dynamics: trace[t] = trace[t-1] * exp(-dt/tau) + spike[t]
        traces[t] = traces[t-1] * decay_factor + spike_matrix[t]
    
    return traces

print("\n Online synaptic trace function defined")
print(f"  Dynamics: trace[t] = trace[t-1] * exp(-{DT}/{TAU_SYN}) + spike[t]")

#%% Cell 4: Create Target Signals

def create_identity_signals(signal_dim, duration_ms, dt):
    """Create input = output for identity mapping test."""
    n_timesteps = int(duration_ms / dt)
    time = np.arange(n_timesteps) * dt
    
    input_signal = np.zeros((n_timesteps, signal_dim))
    for dim in range(signal_dim):
        freq = (dim + 1) * 2 * np.pi / duration_ms
        input_signal[:, dim] = 0.5 + 0.5 * np.sin(freq * time)
    
    output_signal = input_signal.copy()  # Identity mapping
    
    return input_signal, output_signal, time

input_signal, output_signal, time = create_identity_signals(
    OUTPUT_DIM_INTRINSIC, STIMULUS_DURATION, DT
)

print(f"\n Signals created: {input_signal.shape}")

# Visualize
fig, ax = plt.subplots(figsize=(14, 6))
for dim in range(OUTPUT_DIM_INTRINSIC):
    ax.plot(time, input_signal[:, dim], '--', label=f'Input {dim+1}', alpha=0.5, linewidth=2)
    ax.plot(time, output_signal[:, dim], '-', label=f'Target {dim+1}', linewidth=2)
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Value')
ax.set_title('LSM Task: Identity Mapping (Online-Compatible)')
ax.legend(ncol=2)
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('lsm_online_signals.png', dpi=150, bbox_inches='tight')
plt.show()

#%% Cell 5: Initialize Fixed Random Reservoir

print("\n" + "="*70)
print("PHASE 1: Initialize Fixed Random Reservoir")
print("="*70)

SESSION_ID = 0

# Create reservoir WITHOUT built-in readout (we'll train our own)
network = SpikingRNN(
    n_neurons=N_NEURONS,
    n_readout_neurons=0,  # No built-in readout
    dt=DT,
    synaptic_mode='filter',
    static_input_mode='independent',
    hd_input_mode='independent',
    n_hd_channels=INPUT_DIM_EMBEDDING,
    use_readout_synapses=False  # We'll create auxiliary layer manually
)

# Initialize reservoir (FIXED)
network.initialize_network(
    session_id=SESSION_ID,
    v_th_std=V_TH_STD,
    g_std=G_STD,
    v_th_distribution='normal',
    hd_dim=INPUT_DIM_INTRINSIC,
    embed_dim=INPUT_DIM_EMBEDDING,
    hd_connection_prob=0.3,
    hd_input_strength=1.0,
    static_input_strength=10.0
)

print(f" Reservoir initialized (structure FIXED)")
print(f"  All weights and thresholds are random and will NOT be trained")

# HD input generator
hd_gen = HDInputGenerator(embed_dim=INPUT_DIM_EMBEDDING, dt=DT, signal_cache_dir='lsm_online_hd')
hd_gen.initialize_base_input(
    SESSION_ID, INPUT_DIM_INTRINSIC, pattern_id=0,
    rate_rnn_params={'n_neurons': 1000, 'T': 500.0, 'g': 2.0}
)

# Prepare HD input rates
base_hd = hd_gen.Y_base
n_signal_timesteps = len(input_signal)
n_base_timesteps = base_hd.shape[0]

from scipy.interpolate import interp1d
resampled_hd = np.zeros((n_signal_timesteps, INPUT_DIM_EMBEDDING))
for ch in range(INPUT_DIM_EMBEDDING):
    f = interp1d(np.linspace(0, 1, n_base_timesteps), base_hd[:, ch], kind='linear')
    resampled_hd[:, ch] = f(np.linspace(0, 1, n_signal_timesteps))

hd_input_rates = resampled_hd - resampled_hd.min()
hd_input_rates = hd_input_rates - hd_input_rates.min()

print(f" HD input rates: {hd_input_rates.shape}")

#%% Cell 6: Collect Spikes from Training Trials

print("\n" + "="*70)
print("PHASE 2: Run Reservoir and Collect Spikes")
print("="*70)

print(f"\nRunning {N_TRIALS_TRAIN} training trials...")

spike_matrices_train = []
targets_train = []

for trial_idx in range(N_TRIALS_TRAIN):
    if (trial_idx + 1) % 10 == 0:
        print(f"  Training trial {trial_idx + 1}/{N_TRIALS_TRAIN}")
    
    # Reset network
    network.reset_simulation(SESSION_ID, V_TH_STD, G_STD, trial_idx)
    
    # Collect spikes as binary matrix
    n_steps = int(STIMULUS_DURATION / DT)
    spike_matrix = np.zeros((n_steps, N_NEURONS), dtype=np.float32)
    
    for step in range(n_steps):
        hd_rates = hd_input_rates[step] if step < len(hd_input_rates) else np.zeros(INPUT_DIM_EMBEDDING)
        
        spike_indices, _ = network.step(
            session_id=SESSION_ID,
            v_th_std=V_TH_STD,
            g_std=G_STD,
            trial_id=trial_idx,
            static_input_rate=STATIC_INPUT_RATE,
            hd_input_rates=hd_rates,
            hd_dim=INPUT_DIM_INTRINSIC,
            embed_dim=INPUT_DIM_EMBEDDING,
            time_step=step
        )
        
        # Mark spikes in matrix
        if len(spike_indices) > 0:
            spike_matrix[step, spike_indices] = 1.0
    
    spike_matrices_train.append(spike_matrix)
    targets_train.append(output_signal)

print(f"\n Collected {len(spike_matrices_train)} spike matrices")
print(f"  Each matrix: {spike_matrices_train[0].shape}")

#%% Cell 7: Compute ABAB Matrix (Auxiliary Layer Activity)

print("\n" + "="*70)
print("PHASE 3: Compute Synaptic Traces (ABAB Matrix)")
print("="*70)

print(f"\nComputing synaptic traces with online dynamics...")
print(f"  tau = {TAU_SYN} ms, dt = {DT} ms")

ABAB_train = []

for trial_idx, spike_matrix in enumerate(spike_matrices_train):
    if (trial_idx + 1) % 10 == 0:
        print(f"  Processing trial {trial_idx + 1}/{N_TRIALS_TRAIN}")
    
    # Compute traces using online-compatible method
    traces = compute_synaptic_traces_online(spike_matrix, DT, TAU_SYN)
    ABAB_train.append(traces)

# Stack: (n_trials, T, N_NEURONS)
ABAB_train = np.array(ABAB_train)

print(f"\n ABAB matrix (auxiliary layer activity):")
print(f"  Shape: {ABAB_train.shape}")
print(f"  Mean activity: {ABAB_train.mean():.4f}")
print(f"  Max activity: {ABAB_train.max():.4f}")
```

```python


# Visualize auxiliary layer activity for one trial
trial_vis = 0
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Spike raster
ax = axes[0]
spike_vis = spike_matrices_train[trial_vis]
for i in range(min(100, N_NEURONS)):
    spike_times = np.where(spike_vis[:, i] > 0)[0] * DT
    if len(spike_times) > 0:
        ax.scatter(spike_times, [i]*len(spike_times), s=1, c='black', alpha=0.6)
ax.set_ylabel('Neuron')
ax.set_title('RNN Spike Raster (first 100 neurons)')
ax.set_xlim([0, STIMULUS_DURATION])

# Auxiliary layer traces
ax = axes[1]
traces_vis = ABAB_train[trial_vis]
for i in range(min(10, N_NEURONS)):
    ax.plot(time, traces_vis[:, i], alpha=0.7, linewidth=1)
ax.set_ylabel('Synaptic Trace')
ax.set_title('Auxiliary Layer Activity (first 10 neurons)')
ax.set_xlim([0, STIMULUS_DURATION])
ax.grid(True, alpha=0.3)

# Target signal
ax = axes[2]
for dim in range(OUTPUT_DIM_INTRINSIC):
    ax.plot(time, output_signal[:, dim], linewidth=2, label=f'Target {dim+1}')
ax.set_xlabel('Time (ms)')
ax.set_ylabel('Target Value')
ax.set_title('Target Output Signal')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('lsm_auxiliary_layer.png', dpi=150, bbox_inches='tight')
plt.show()

#%% Cell 8: Prepare Data for Training

print("\n" + "="*70)
print("PHASE 4: Prepare Training Data")
print("="*70)

# Stack targets: (n_trials, T, OUTPUT_DIM_INTRINSIC)
targets_train_array = np.array(targets_train)

print(f"\nTraining data:")
print(f"  ABAB (auxiliary activity): {ABAB_train.shape}")
print(f"  Targets: {targets_train_array.shape}")

# Reshape for stratified CV
# LinearDecoder expects: (n_trials, n_channels, T)
ABAB_for_cv = ABAB_train.transpose(0, 2, 1)  # (n_trials, N_NEURONS, T)
targets_for_cv = targets_train_array.transpose(0, 2, 1)  # (n_trials, OUTPUT_DIM_INTRINSIC, T)

print(f"\nReshaped for CV:")
print(f"  ABAB: {ABAB_for_cv.shape}")
print(f"  Targets: {targets_for_cv.shape}")

#%% Cell 9: Train Readout Weights with Stratified CV

print("\n" + "="*70)
print("PHASE 5: Train Readout Weights (Stratified Cross-Validation)")
print("="*70)

# Create decoder (using same interface as encoding)
decoder = LinearDecoder(
    dt=DT,
    tau=TAU_SYN,
    lambda_reg=LAMBDA_REG,
    random_state=42
)

print(f"\nTraining readout: {N_NEURONS} auxiliary neurons  {OUTPUT_DIM_INTRINSIC}D output")
print(f"Using {N_CV_FOLDS}-fold stratified cross-validation")

# Train with stratified CV
# Note: We already computed ABAB, so we pass use_cached=True conceptually
# The decoder will use ridge regression to learn weights

# Manually do stratified CV since we have pre-computed ABAB
from sklearn.linear_model import Ridge

# Flatten for sklearn
X_flat = ABAB_train.reshape(-1, N_NEURONS)  # (n_trials*T, N_NEURONS)
y_flat = targets_train_array.reshape(-1, OUTPUT_DIM_INTRINSIC)  # (n_trials*T, OUTPUT_DIM_INTRINSIC)

print(f"\nFlattened data:")
print(f"  X: {X_flat.shape}")
print(f"  y: {y_flat.shape}")

# Simple train/test split for now (we'll do proper CV next)
ridge = Ridge(alpha=LAMBDA_REG)
ridge.fit(X_flat, y_flat)

# Get learned weights
readout_weights = ridge.coef_  # (OUTPUT_DIM_INTRINSIC, N_NEURONS)
readout_bias = ridge.intercept_  # (OUTPUT_DIM_INTRINSIC,)

print(f"\n Readout weights learned:")
print(f"  Weights shape: {readout_weights.shape}")
print(f"  Bias shape: {readout_bias.shape}")

# Evaluate on training data
y_pred_train = ridge.predict(X_flat)

train_rmse = np.sqrt(np.mean((y_flat - y_pred_train)**2))
train_r2 = 1 - np.sum((y_flat - y_pred_train)**2) / np.sum((y_flat - y_flat.mean(axis=0))**2)

print(f"\nTraining performance:")
print(f"  RMSE: {train_rmse:.4f}")
print(f"  R: {train_r2:.4f}")

#%% Cell 10: Proper Cross-Validation

print("\n" + "="*70)
print("Cross-Validation (Trial-Level Splits)")
print("="*70)

# Split by trials (not by timesteps)
cv_rmse = []
cv_r2 = []

from sklearn.model_selection import KFold
kfold = KFold(n_splits=N_CV_FOLDS, shuffle=True, random_state=42)

for fold_idx, (train_trials, test_trials) in enumerate(kfold.split(range(N_TRIALS_TRAIN))):
    
    # Get training trials
    X_train_fold = ABAB_train[train_trials].reshape(-1, N_NEURONS)
    y_train_fold = targets_train_array[train_trials].reshape(-1, OUTPUT_DIM_INTRINSIC)
    
    # Get test trials
    X_test_fold = ABAB_train[test_trials].reshape(-1, N_NEURONS)
    y_test_fold = targets_train_array[test_trials].reshape(-1, OUTPUT_DIM_INTRINSIC)
    
    # Train
    fold_ridge = Ridge(alpha=LAMBDA_REG)
    fold_ridge.fit(X_train_fold, y_train_fold)
    
    # Predict
    y_pred_fold = fold_ridge.predict(X_test_fold)
    
    # Evaluate
    fold_rmse = np.sqrt(np.mean((y_test_fold - y_pred_fold)**2))
    fold_r2 = 1 - np.sum((y_test_fold - y_pred_fold)**2) / np.sum((y_test_fold - y_test_fold.mean(axis=0))**2)
    
    cv_rmse.append(fold_rmse)
    cv_r2.append(fold_r2)
    
    print(f"  Fold {fold_idx+1}: RMSE={fold_rmse:.4f}, R={fold_r2:.4f}")

print(f"\n Cross-validation complete:")
print(f"  Mean CV RMSE: {np.mean(cv_rmse):.4f}  {np.std(cv_rmse):.4f}")
print(f"  Mean CV R: {np.mean(cv_r2):.4f}  {np.std(cv_r2):.4f}")

#%% Cell 11: Test on New Trials (ONLINE)

print("\n" + "="*70)
print("PHASE 6: Test with Learned Weights (ONLINE)")
print("="*70)

print(f"\nRunning {N_TRIALS_TEST} test trials with ONLINE readout...")

test_predictions = []
test_targets = []

decay_factor = np.exp(-DT / TAU_SYN)

for trial_idx in range(N_TRIALS_TEST):
    test_trial_id = N_TRIALS_TRAIN + trial_idx
    
    if (trial_idx + 1) % 5 == 0:
        print(f"  Test trial {trial_idx + 1}/{N_TRIALS_TEST}")
    
    # Reset network
    network.reset_simulation(SESSION_ID, V_TH_STD, G_STD, test_trial_id)
    
    # Initialize auxiliary traces
    auxiliary_traces = np.zeros(N_NEURONS)
    
    # Storage for this trial
    trial_output = []
    n_steps = int(STIMULUS_DURATION / DT)
    
    for step in range(n_steps):
        hd_rates = hd_input_rates[step] if step < len(hd_input_rates) else np.zeros(INPUT_DIM_EMBEDDING)
        
        # Step network
        spike_indices, _ = network.step(
            session_id=SESSION_ID,
            v_th_std=V_TH_STD,
            g_std=G_STD,
            trial_id=test_trial_id,
            static_input_rate=STATIC_INPUT_RATE,
            hd_input_rates=hd_rates,
            hd_dim=INPUT_DIM_INTRINSIC,
            embed_dim=INPUT_DIM_EMBEDDING,
            time_step=step
        )
        
        # Update auxiliary traces (ONLINE)
        auxiliary_traces = auxiliary_traces * decay_factor
        if len(spike_indices) > 0:
            auxiliary_traces[spike_indices] += 1.0
        
        # Apply learned readout weights (ONLINE)
        output_t = auxiliary_traces @ readout_weights.T + readout_bias
        trial_output.append(output_t)
    
    trial_output = np.array(trial_output)  # (T, OUTPUT_DIM_INTRINSIC)
    
    test_predictions.append(trial_output)
    test_targets.append(output_signal)

# Evaluate
y_pred_test = np.vstack(test_predictions)
y_test = np.vstack(test_targets)

test_rmse = np.sqrt(np.mean((y_test - y_pred_test)**2))
test_r2 = 1 - np.sum((y_test - y_pred_test)**2) / np.sum((y_test - y_test.mean(axis=0))**2)

print(f"\n Test completed (ONLINE readout):")
print(f"  Test RMSE: {test_rmse:.4f}")
print(f"  Test R: {test_r2:.4f}")

#%% Cell 12: Visualize Results

print("\nVisualizing results...")

trial_length = len(time)

# Training
y_pred_train_trials = y_pred_train.reshape(N_TRIALS_TRAIN, trial_length, OUTPUT_DIM_INTRINSIC)
y_train_trials = y_flat.reshape(N_TRIALS_TRAIN, trial_length, OUTPUT_DIM_INTRINSIC)

# Test
y_pred_test_trials = y_pred_test.reshape(N_TRIALS_TEST, trial_length, OUTPUT_DIM_INTRINSIC)
y_test_trials = y_test.reshape(N_TRIALS_TEST, trial_length, OUTPUT_DIM_INTRINSIC)

fig, axes = plt.subplots(OUTPUT_DIM_INTRINSIC, 2, figsize=(16, 4*OUTPUT_DIM_INTRINSIC))
if OUTPUT_DIM_INTRINSIC == 1:
    axes = axes.reshape(1, 2)

for dim in range(OUTPUT_DIM_INTRINSIC):
    # Training
    ax = axes[dim, 0]
    target_mean = y_train_trials[:, :, dim].mean(axis=0)
    pred_mean = y_pred_train_trials[:, :, dim].mean(axis=0)
    
    ax.plot(time, target_mean, 'k-', linewidth=2, label='Target')
    ax.plot(time, pred_mean, 'b-', linewidth=2, label='LSM Output', alpha=0.8)
    ax.set_ylabel(f'Dim {dim+1}')
    ax.set_title(f'Training - Dimension {dim+1}')
    ax.legend()
    ax.grid(True, alpha=0.3)
    if dim == OUTPUT_DIM_INTRINSIC - 1:
        ax.set_xlabel('Time (ms)')
    
    # Test (ONLINE)
    ax = axes[dim, 1]
    target_mean = y_test_trials[:, :, dim].mean(axis=0)
    pred_mean = y_pred_test_trials[:, :, dim].mean(axis=0)
    
    ax.plot(time, target_mean, 'k-', linewidth=2, label='Target')
    ax.plot(time, pred_mean, 'r-', linewidth=2, label='LSM Output (ONLINE)', alpha=0.8)
    ax.set_ylabel(f'Dim {dim+1}')
    ax.set_title(f'Test (ONLINE) - Dimension {dim+1}')
    ax.legend()
    ax.grid(True, alpha=0.3)
    if dim == OUTPUT_DIM_INTRINSIC - 1:
        ax.set_xlabel('Time (ms)')

plt.tight_layout()
plt.savefig('lsm_online_results.png', dpi=150, bbox_inches='tight')
plt.show()

#%% Cell 13: Analyze Readout Weights

print("\n" + "="*70)
print("Readout Weight Analysis")
print("="*70)

fig, axes = plt.subplots(1, OUTPUT_DIM_INTRINSIC, figsize=(5*OUTPUT_DIM_INTRINSIC, 4))
if OUTPUT_DIM_INTRINSIC == 1:
    axes = [axes]

for dim in range(OUTPUT_DIM_INTRINSIC):
    ax = axes[dim]
    weights = readout_weights[dim, :]
    
    ax.hist(weights, bins=50, alpha=0.7, edgecolor='black')
    ax.set_xlabel('Weight Value')
    ax.set_ylabel('Count')
    ax.set_title(f'Output Dim {dim+1} - Readout Weights')
    ax.axvline(0, color='red', linestyle='--', linewidth=1)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('lsm_online_weights.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"\nReadout weight statistics:")
for dim in range(OUTPUT_DIM_INTRINSIC):
    weights = readout_weights[dim, :]
    n_strong_pos = np.sum(weights > 0.01)
    n_strong_neg = np.sum(weights < -0.01)
    
    print(f"\n  Output dimension {dim+1}:")
    print(f"    Mean: {weights.mean():.4f}, Std: {weights.std():.4f}")
    print(f"    Range: [{weights.min():.4f}, {weights.max():.4f}]")
    print(f"    Strong positive: {n_strong_pos}/{N_NEURONS} neurons")
    print(f"    Strong negative: {n_strong_neg}/{N_NEURONS} neurons")

#%% Cell 14: Final Summary

print("\n" + "="*70)
print("LIQUID STATE MACHINE - FINAL SUMMARY")
print("="*70)

print(f"\nArchitecture:")
print(f"  1. Fixed random reservoir: {N_NEURONS} spiking neurons")
print(f"  2. Auxiliary layer: {N_NEURONS} synaptic traces (weight=1.0)")
print(f"  3. Readout layer: {N_READOUT} neurons with learned weights")
print(f"  4. Output: {OUTPUT_DIM_INTRINSIC}D HD signal")

print(f"\nSynaptic Dynamics:")
print(f"  tau_syn = {TAU_SYN} ms")
print(f"  Update: trace[t] = trace[t-1] * exp(-dt/tau) + spike[t]")
print(f"  ONLINE-COMPATIBLE ")

print(f"\nPerformance:")
print(f"  Training:")
print(f"    RMSE: {train_rmse:.4f}")
print(f"    R: {train_r2:.4f}")
print(f"  Cross-Validation:")
print(f"    RMSE: {np.mean(cv_rmse):.4f}  {np.std(cv_rmse):.4f}")
print(f"    R: {np.mean(cv_r2):.4f}  {np.std(cv_r2):.4f}")
print(f"  Test (ONLINE):")
print(f"    RMSE: {test_rmse:.4f}")
print(f"    R: {test_r2:.4f}")

print(f"\nKey Features:")
print(f"   Synaptic traces match online dynamics exactly")
print(f"   Training and testing use identical computation")
print(f"   Readout can be computed online (no need to collect all spikes)")
print(f"   Same approach as Maass et al. (2002) LSM")

print(f"\nInterpretation:")
if test_r2 > 0.7:
    print("   EXCELLENT - LSM successfully learned identity mapping!")
    print("      Fixed reservoir + trained readout works very well")
    print("      Online computation is accurate")
elif test_r2 > 0.5:
    print("   GOOD - LSM learned reasonably well")
    print("     Reservoir provides useful dynamics")
elif test_r2 > 0.3:
    print("   MODERATE - Partial learning")
    print("    Consider: larger reservoir, tuning parameters")
else:
    print("   WEAK - Limited learning")
    print("    Try: more neurons, different G_STD, higher input strength")

print("\n" + "="*70)
print("This is Maass's LSM with ONLINE-COMPATIBLE synaptic dynamics")
print("Training and testing use identical iterative computation")
print("="*70)
```
