```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D



# Function to extract data for specific input rate
def get_data_grid(results, metric_name, input_rate, v_th_stds, g_stds):
    """Extract data into 2D grid for specific input rate"""
    grid = np.full((len(g_stds), len(v_th_stds)), np.nan)
    
    for r in results:
        if np.isclose(r['static_input_rate'], input_rate):
            i = g_stds.index(r['g_std'])
            j = v_th_stds.index(r['v_th_std'])
            grid[i, j] = r[metric_name]
    
    return grid


# Function to get dimensionality across bin sizes
def get_dimensionality_vs_binsize(results, metric_prefix):
    """Average dimensionality across all parameters for each bin size"""
    bin_sizes = [0.1, 2.0, 5.0, 20.0, 50.0, 100.0]
    means = []
    stds = []
    
    for bin_size in bin_sizes:
        metric_name = f'{metric_prefix}_bin_{bin_size}ms_mean'
        values = [r[metric_name] for r in results if not np.isnan(r[metric_name])]
        means.append(np.mean(values))
        stds.append(np.std(values))
    
    return bin_sizes, np.array(means), np.array(stds)
```

```python


fig = plt.figure(figsize=(16, 8))

metrics_3d = [
    ('percent_active_mean', 'Active Neurons (%)'),
    ('mean_firing_rate_mean', 'Firing Rate (Hz)'),
    ('participation_ratio_bin_50.0ms_mean', 'Participation Ratio')
]

# Define outer grid with 1 row, 2 columns
outer = gridspec.GridSpec(1, 2, width_ratios=[1, 1], wspace=0.4)

# Left 22 block
left = gridspec.GridSpecFromSubplotSpec(2, 2, subplot_spec=outer[0], wspace=0.8, hspace=0.0)

# Right 22 block
right = gridspec.GridSpecFromSubplotSpec(2, 2, subplot_spec=outer[1], wspace=0.8, hspace=0.0)

# Example: fill left block
ax_left_00 = fig.add_subplot(left[0, 0], projection='3d')   # top-left
ax_left_01 = fig.add_subplot(left[0, 1], projection='3d')   # top-right
ax_left_10 = fig.add_subplot(left[1, 0])                    # bottom-left
pos = ax_left_10.get_position()
ax_left_10.set_position([
    pos.x0,               # left stays the same
    pos.y0 + pos.height*0.25,  # move bottom up
    pos.width,            # same width
    pos.height * 0.5      # shrink height
])
ax_left_11 = fig.add_subplot(left[1, 1], projection='3d')   # bottom-right

# Example: fill right block
ax_right_00 = fig.add_subplot(right[0, 0], projection='3d')
ax_right_01 = fig.add_subplot(right[0, 1], projection='3d')
ax_right_10 = fig.add_subplot(right[1, 0])
pos = ax_right_10.get_position()
ax_right_10.set_position([
    pos.x0,               # left stays the same
    pos.y0 + pos.height*0.25,  # move bottom up
    pos.width,            # same width
    pos.height * 0.5      # shrink height
])
ax_right_11 = fig.add_subplot(right[1, 1], projection='3d')



axleft = [ax_left_00,ax_left_01,ax_left_11]
filename = 'results/data/spontaneous_session_2_immediate.pkl'  # Adjust as needed
with open(filename, 'rb') as f:
    results = pickle.load(f)
v_th_stds = sorted(list(set([r['v_th_std'] for r in results])))
g_stds = sorted(list(set([r['g_std'] for r in results])))
input_rates = sorted(list(set([r['static_input_rate'] for r in results])))
selected_rate = input_rates[0]
V_TH, G = np.meshgrid(v_th_stds, g_stds)





mask = (V_TH > 0) & (G > 0)  # True where both conditions are met



for idx, (metric, title) in enumerate(metrics_3d):
    
    Z = get_data_grid(results, metric, selected_rate, v_th_stds, g_stds)
    # Mask values outside your desired range
    Z_masked = np.where(mask, Z, np.nan)
    
    surf = axleft[idx].plot_surface(G, V_TH, Z_masked, cmap=cm.viridis)
    
    # Set view angle so origin corner is at lowest values
    axleft[idx].view_init(elev=30, azim=225)  # Adjust angles as needed
    
    axleft[idx].set_ylabel('v_th_std', fontsize=10)
    axleft[idx].set_xlabel('g_std', fontsize=10)
    axleft[idx].set_zlabel(title, fontsize=10)
    axleft[idx].set_title(f'{title}', fontsize=11)

    
    axleft[idx].set_xticks([0,4,8])
    axleft[idx].set_yticks([0,2,4])
    
    fig.colorbar(surf, ax=axleft[idx], shrink=0.5, aspect=5)
    
axleft[2].set_title('Dimensionality (bin = 50 ms)', fontsize=11)

bin_sizes, means, stds = get_dimensionality_vs_binsize(results, 'participation_ratio')
ax_left_10.errorbar(bin_sizes, means, yerr=stds, fmt='o-', capsize=5, 
            linewidth=2, markersize=6, color='purple')
ax_left_10.set_xlabel('Bin Size (ms)', fontsize=11)
ax_left_10.set_ylabel('Participation Ratio', fontsize=11)
ax_left_10.set_title('Dimensionality vs Bin Size', fontsize=11)
# ax.set_xscale('log')
ax_left_10.grid(True, alpha=0.3)





axright = [ax_right_00,ax_right_01,ax_right_11]
filename = 'results/data/spontaneous_session_2_dynamic.pkl'  # Adjust as needed
with open(filename, 'rb') as f:
    results = pickle.load(f)
v_th_stds = sorted(list(set([r['v_th_std'] for r in results])))
g_stds = sorted(list(set([r['g_std'] for r in results])))
input_rates = sorted(list(set([r['static_input_rate'] for r in results])))
selected_rate = input_rates[0]
V_TH, G = np.meshgrid(v_th_stds, g_stds)
for idx, (metric, title) in enumerate(metrics_3d):
    
    Z = get_data_grid(results, metric, selected_rate, v_th_stds, g_stds)
    # Mask values outside your desired range
    Z_masked = np.where(mask, Z, np.nan)
    
    surf = axright[idx].plot_surface(G, V_TH, Z_masked, cmap=cm.viridis)
    
    # Set view angle so origin corner is at lowest values
    axright[idx].view_init(elev=30, azim=225)  # Adjust angles as needed
    
    axright[idx].set_ylabel('v_th_std', fontsize=10)
    axright[idx].set_xlabel('g_std', fontsize=10)
    axright[idx].set_zlabel(title, fontsize=10)
    axright[idx].zaxis.label.set_rotation(180)
    axright[idx].set_title(f'{title}', fontsize=11)

    axright[idx].set_xticks([0,4,8])
    axright[idx].set_yticks([0,2,4])
    
    fig.colorbar(surf, ax=axright[idx], shrink=0.5, aspect=5)

axright[2].set_title('Dimensionality (bin = 50 ms)', fontsize=11)


bin_sizes, means, stds = get_dimensionality_vs_binsize(results, 'participation_ratio')
ax_right_10.errorbar(bin_sizes, means, yerr=stds, fmt='o-', capsize=5, 
            linewidth=2, markersize=6, color='purple')
ax_right_10.set_xlabel('Bin Size (ms)', fontsize=11)
ax_right_10.set_ylabel('Participation Ratio', fontsize=11)
ax_right_10.set_title('Dimensionality vs Bin Size', fontsize=11)
# ax.set_xscale('log')
ax_right_10.grid(True, alpha=0.3)

fig.text(0.25, 0.85, "Pulse Synapses", ha="center", fontsize=16, fontweight="bold")
fig.text(0.72, 0.85, "Filter Synapses", ha="center", fontsize=16, fontweight="bold")


plt.savefig('spontaneous_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
