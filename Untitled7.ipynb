```python
import matplotlib.pyplot as plt
import numpy as np

# Suppose chosen static_input_rate values are [30, 35, 40, 45] (choose any 4)
static_input_rates_to_plot = [30, 35, 40, 45]
v_th_options = [1, 2]
g_std_options = [1, 2]
hd_dim_options = [1, 4, 7, 10, 13]

fig, axs = plt.subplots(2, 2, figsize=(12, 10), sharex=True, sharey=True)

for idx, rate in enumerate(static_input_rates_to_plot):
    ax = axs[idx // 2, idx % 2]
    for v_th in v_th_options:
        for g_std in g_std_options:
            # Gather data for all hd_dim values
            means = []
            stds = []
            x = []
            for hd_dim in hd_dim_options:
                # Get matching entries
                matches = [res for res in results if res['static_input_rate'] == rate and
                           res['v_th_std'] == v_th and res['g_std'] == g_std and res['hd_dim'] == hd_dim]
                if matches:
                    mean = matches[0]['encoding_metrics']['test_r2']['mean']
                    std = matches[0]['encoding_metrics']['test_r2']['std']
                    means.append(mean)
                    stds.append(std)
                    x.append(hd_dim)
            # Plot if any data found
            if means:
                label = f'v_th={v_th}, g_std={g_std}'
                ax.errorbar(x, means, yerr=stds, marker='o', capsize=4, label=label)
    ax.set_title(f'static_input_rate={rate}')
    ax.set_xlabel('hd_dim')
    ax.set_ylabel('test_rmse')
    ax.legend()
plt.tight_layout()
plt.show()
```
