```python
# Jupyter Notebook Cell 1: Import and Setup
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# Add your project directories to Python path
sys.path.insert(0, 'src')
sys.path.insert(0, 'analysis') 
sys.path.insert(0, 'experiments')

# Import your existing modules
from spiking_network import SpikingRNN
from rng_utils import get_rng

print("Modules imported successfully!")

# Jupyter Notebook Cell 2: Run Network Simulation
def create_network_with_ei_types(n_neurons=1000, inhibitory_fraction=0.2):
    """Create network and define E/I neuron types"""
    
    # Create your existing network
    network = SpikingRNN(n_neurons=n_neurons, dt=0.1)
    
    # Define which neurons are excitatory vs inhibitory
    # (Your current network doesn't distinguish, so we'll assign types for visualization)
    n_inhibitory = int(n_neurons * inhibitory_fraction)
    
    # Create neuron type arrays
    inhibitory_indices = np.arange(n_inhibitory)  # First 20% are inhibitory
    excitatory_indices = np.arange(n_inhibitory, n_neurons)  # Rest are excitatory
    
    return network, excitatory_indices, inhibitory_indices

# Create and initialize network
print("Creating network...")
network, exc_indices, inh_indices = create_network_with_ei_types(n_neurons=1000)

# Initialize network with your parameters
network.initialize_network(
    session_id=1, 
    block_id=1, 
    v_th_std=1,  # Spike threshold heterogeneity
    g_std=0.01,     # Synaptic weight heterogeneity
    static_input_strength=1.0,
    connection_prob=0.1
)

print(f"Network initialized:")
print(f"  Excitatory neurons: {len(exc_indices)}")
print(f"  Inhibitory neurons: {len(inh_indices)}")

# Jupyter Notebook Cell 3: Run Simulation
# Run simulation using your existing method
print("Running simulation...")
duration = 500.0  # ms

spike_times = network.simulate_network_dynamics(
    session_id=1,
    block_id=1, 
    trial_id=1,
    duration=duration,
    static_input_rate=10.0  # Background Poisson rate in Hz
)

print(f"Simulation completed!")
print(f"Total spikes generated: {len(spike_times)}")
print(f"Average firing rate: {len(spike_times)/(duration/1000)/network.n_neurons:.2f} Hz")

# Show first few spikes
print(f"First 10 spikes: {spike_times[:10]}")

# Jupyter Notebook Cell 4: Create Raster Plot
def plot_raster_with_ei_colors(spike_times, exc_indices, inh_indices, duration=500.0):
    """Create raster plot with red for inhibitory, blue for excitatory"""
    
    if not spike_times:
        print("No spikes to plot!")
        return None, None
    
    # Separate spike times by neuron type
    exc_times = []
    exc_neurons = []
    inh_times = []
    inh_neurons = []
    
    for spike_time, neuron_id in spike_times:
        if neuron_id in exc_indices:
            exc_times.append(spike_time)
            exc_neurons.append(neuron_id)
        elif neuron_id in inh_indices:
            inh_times.append(spike_time)
            inh_neurons.append(neuron_id)
    
    # Create the plot
    fig, ax = plt.subplots(figsize=(14, 8))
    
    # Plot excitatory spikes in blue
    if exc_times:
        ax.scatter(exc_times, exc_neurons, c='blue', s=2, alpha=0.7, 
                  label=f'Excitatory (n={len(exc_indices)})')
    
    # Plot inhibitory spikes in red
    if inh_times:
        ax.scatter(inh_times, inh_neurons, c='red', s=2, alpha=0.7, 
                  label=f'Inhibitory (n={len(inh_indices)})')
    
    # Formatting
    ax.set_xlabel('Time (ms)', fontsize=14)
    ax.set_ylabel('Neuron ID', fontsize=14)
    ax.set_title('Spiking Network Raster Plot\n(Blue: Excitatory, Red: Inhibitory)', 
                fontsize=16, fontweight='bold')
    ax.legend(loc='upper right', fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, duration)
    ax.set_ylim(-5, max(exc_indices[-1], inh_indices[-1]) + 5)
    
    # Add statistics text box
    total_spikes = len(spike_times)
    exc_spikes = len(exc_times)
    inh_spikes = len(inh_times)
    exc_rate = exc_spikes / (duration/1000) / len(exc_indices) if exc_spikes > 0 else 0
    inh_rate = inh_spikes / (duration/1000) / len(inh_indices) if inh_spikes > 0 else 0
    
    stats_text = (f'Total spikes: {total_spikes}\n'
                 f'Excitatory: {exc_spikes} ({exc_spikes/total_spikes*100:.1f}%)\n'
                 f'Inhibitory: {inh_spikes} ({inh_spikes/total_spikes*100:.1f}%)\n'
                 f'Exc. rate: {exc_rate:.1f} Hz\n'
                 f'Inh. rate: {inh_rate:.1f} Hz')
    
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, fontsize=11,
            verticalalignment='top', bbox=dict(boxstyle='round', 
            facecolor='lightgray', alpha=0.8))
    
    plt.tight_layout()
    return fig, ax

# Create the raster plot
fig, ax = plot_raster_with_ei_colors(spike_times, exc_indices, inh_indices, duration)
plt.show()

# Jupyter Notebook Cell 5: Population Firing Rate Analysis
def plot_population_rates(spike_times, exc_indices, inh_indices, duration=500.0, bin_size=20.0):
    """Plot population firing rates over time"""
    
    # Create time bins
    time_bins = np.arange(0, duration + bin_size, bin_size)
    time_centers = time_bins[:-1] + bin_size/2
    
    # Initialize spike counts
    exc_counts = np.zeros(len(time_bins)-1)
    inh_counts = np.zeros(len(time_bins)-1)
    
    # Count spikes in each time bin
    for spike_time, neuron_id in spike_times:
        bin_idx = int(spike_time // bin_size)
        if bin_idx < len(exc_counts):
            if neuron_id in exc_indices:
                exc_counts[bin_idx] += 1
            elif neuron_id in inh_indices:
                inh_counts[bin_idx] += 1
    
    # Convert to firing rates (Hz)
    exc_rates = exc_counts / (bin_size/1000) / len(exc_indices)
    inh_rates = inh_counts / (bin_size/1000) / len(inh_indices)
    
    # Create plot
    fig, ax = plt.subplots(figsize=(14, 6))
    
    ax.plot(time_centers, exc_rates, 'b-', linewidth=2.5, 
           label='Excitatory Population', alpha=0.8)
    ax.plot(time_centers, inh_rates, 'r-', linewidth=2.5, 
           label='Inhibitory Population', alpha=0.8)
    
    ax.set_xlabel('Time (ms)', fontsize=14)
    ax.set_ylabel('Population Firing Rate (Hz)', fontsize=14)
    ax.set_title(f'Population Firing Rates (bin size: {bin_size:.0f}ms)', 
                fontsize=16, fontweight='bold')
    ax.legend(fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, duration)
    
    # Add mean rate lines
    mean_exc_rate = np.mean(exc_rates)
    mean_inh_rate = np.mean(inh_rates)
    ax.axhline(mean_exc_rate, color='blue', linestyle='--', alpha=0.6, 
              label=f'Mean Exc: {mean_exc_rate:.1f} Hz')
    ax.axhline(mean_inh_rate, color='red', linestyle='--', alpha=0.6, 
              label=f'Mean Inh: {mean_inh_rate:.1f} Hz')
    
    plt.tight_layout()
    return fig, ax

# Create population rate plot
fig2, ax2 = plot_population_rates(spike_times, exc_indices, inh_indices, duration)
plt.show()

# Jupyter Notebook Cell 6: Network Analysis
# Analyze the network structure using your existing methods
network_info = network.get_network_info()

print("Network Analysis:")
print(f"  Total neurons: {network_info['n_neurons']}")
print(f"  Non-zero synaptic connections: {network_info['weight_matrix_nnz']}")
print(f"  Connection density: {network_info['weight_matrix_nnz']/(network_info['n_neurons']**2)*100:.2f}%")
print(f"  Spike threshold range: {np.min(network_info['spike_thresholds']):.3f} to {np.max(network_info['spike_thresholds']):.3f} mV")
print(f"  Spike threshold std: {np.std(network_info['spike_thresholds']):.3f} mV")

# Calculate firing statistics
if spike_times:
    spike_counts_per_neuron = np.zeros(network.n_neurons)
    for _, neuron_id in spike_times:
        spike_counts_per_neuron[neuron_id] += 1
    
    active_neurons = np.sum(spike_counts_per_neuron > 0)
    mean_rate = np.mean(spike_counts_per_neuron) / (duration/1000)
    
    print(f"  Active neurons: {active_neurons}/{network.n_neurons} ({active_neurons/network.n_neurons*100:.1f}%)")
    print(f"  Mean firing rate: {mean_rate:.2f} Hz")
    print(f"  Max spikes from single neuron: {int(np.max(spike_counts_per_neuron))}")

# Jupyter Notebook Cell 7: Perturbation Analysis (Optional)
# Run a perturbation experiment using your chaos analysis
print("\nRunning perturbation analysis...")

# Run control simulation
control_spikes = network.simulate_network_dynamics(
    session_id=2,  # Different session for new random state
    block_id=1, 
    trial_id=1,
    duration=duration,
    static_input_rate=100.0
)

# Run perturbed simulation (inject spike at t=50ms into neuron 10)
perturbed_spikes = network.simulate_network_dynamics(
    session_id=2,
    block_id=1, 
    trial_id=1,
    duration=duration,
    static_input_rate=100.0,
    perturbation_time=50.0,
    perturbation_neuron=10
)

# Analyze perturbation response using your analysis tools
from spike_analysis import analyze_perturbation_response

lz_complexity, hamming_slope = analyze_perturbation_response(
    spikes_control=control_spikes,
    spikes_perturbed=perturbed_spikes,
    num_neurons=network.n_neurons,
    perturbation_time=50.0,
    simulation_end=duration,
    perturbed_neuron=10
)

print(f"Perturbation Analysis Results:")
print(f"  Lempel-Ziv Complexity: {lz_complexity}")
print(f"  Hamming Distance Slope: {hamming_slope:.4f}")
print(f"  Chaos Level: {'High' if hamming_slope > 0.01 else 'Low'}")

print("\nVisualization complete! You now have:")
print("1. Raster plot with E/I color coding")
print("2. Population firing rate analysis") 
print("3. Network structure statistics")
print("4. Perturbation response analysis")
```
